# Публичная жизнь и личные мечты: как сделать поле недоступным
Обратите внимание на появление слова private перед объявлением каждого поля в классе Account. Это ключевое слово языка Java. Его наличие в объявлении означает, что поле объявлено как закрытое. Модификатор public создает открытое поле. Закрытое поле отличается от открытого тем, что никакой код, расположенный за пределами класса, не имеет к нему непосредственного доступа. Если вставить в класс UseAccount (см. листинг 7.9) инструкцию myAccount.name = "Барри", компилятор сообщит о том, что поле name имеет тип доступа private и к нему нельзя обращаться.
Вместо обращения к полю myAccount.name программист, создающий класс UseAccount, должен вы3ывать меTод myAccount.getName() или myAccount. setName (). Эти методы называются методами доступа, потому что с их помощью класс Account предоставляет доступ к закрытому полю name. Формально термин "метод доступа" в синтаксисе Java не определен. Это всего лишь термин, который программисты используют в отношении методов, решающих задачи такого рода. Методы категорий get и set по отдельности называют методами-получателями и методами-установщиками. Часто их называют также get-методами и set-методами.

- совет - Во многих интегрированных средах разработки можно не вводить код методов доступа вручную, потому что в них есть средства автоматизации этой задачи. Сначала введите объявление поля, например private String name. Затем выберите в меню команду Source Generate Getter or Setter (Исходный код- Сrенерировать получатель или установщик). В некоторых средах команда называется Code Insert Code-Setter (Код- Вставить код-Установщик). После запуска соответствующей команды среда разработки автоматически создает методы доступа и добавляет их в код программы.

Обратите внимание на то, что в листинге 7.8 все методы доступа объявлены как public. Это гарантирует возможность их вызова из любого кода. Лежащая в основе такого подхода идея заключается в том, что непосредственное манипулирование фактическими полями извне класса Account невозможно, однако это не препятствует их использованию посредством разрешенных методов доступа.
