# Создание платежного чека
В листинге 8.1 есть три вызова метода printf. В каждом вызове используются строка форматирования (например, " (%s) ***$") и отображаемое значение (например, jobTitle). В каждой строке форматирования есть заполнитель (наподобие 8s), который определяет, в каком месте и как выводится значение. К примеру, во втором вызове printf строка форматирования содержит заполнитель 8s, который при выводе заменяется значением переменной jobTitle. Согласно правилам Java пара символов &s всегда служит заполни- телем для значения типа String. Это означает, что при вызове метода printf символы &s заменяются строкой, хранящейся в переменной jobTitle, которая должна быть объявлена с типом String. Скобки, окружающие заполнитель &s, выводятся без изменений, поэтому на экране значение jobTitle заключено в скобки (рис. 8.2).

    Выплатить служащему Барри Берд (Директор) $***\$$5 000,00
    Выплатить служашему Гарриет Риттер (Капитан) $***\$$7 000,00
    Выплатить служащему Ваше имя (Лучший менеджер дня) ***$10 000,00

В листинге 8.1 обратите внимание на запятую в заполнителе 8,.2f. Она сообщает компьютеру о том, что при выводе значения нужно применить разделитель групп -- символ, разделяющий в длинном числе группы из трех цифр. # Работа с файлами (небольшое отступление)
В предыдущих главах программы читали символы и строки с клавиатуры. Но код листинга 8.2 читает строки из существующего файла. Файл Employee Info.txt находится на жестком диске компьютера.
Файл EmployeeInfo.txt сохранен в текстовом формате и может содержать буквы, цифры и любые другие символы. В отличие от документа Word, файл в текстовом формате не может содержать элементы форматирования, такие как курсивный шрифт, маркированные и нумерованные списки и т.п. Кроме того, он не может содержать изображения, таблицы, гиперссылки и пр.
В файле в текстовом формате могут храниться только обычные символы. В главах 5 и б была рассмотрена программа игры в угадывание, которая читала символы с клавиатуры. Важное различие между чтением символов из файла и с клавиатуры состоит в том, что во втором случае происходит диалог пользователя с компьютером. Сначала компьютер отображает на экране приглашение ввести символ, и только после этого пользователь вводит его. При чтении из файла никакого диалога нет. Программа DoPayroll (см. листинг 8.2) читает символы с жесткого диска, не предлагая пользователю что-либо сделать и даже не сообщая ему об этом. Данная глава посвящена повторному использованию кода. Однако листинг 8.2 отражает другую важную идею, не связанную непосредственно с повторным использованием кода: чтение информации из файла. Поэтому в нескольких следующих разделах я немного отвлекусь от главной темы главы, чтобы коротко рассказать о работе с файлами на жестком диске.

# Хранение данных в файле
Код листинга 8.2 не может быть выполнен, если на диске нет файла EmployeeInfo.txt, содержащего данные о служащих. Поэтому перед выполнением листинга 8.2 я создал небольшой текстовый файл, приведенный на рис. 8.3. Результирующий вывод программы был показан на рис. 8.2.

в интернете - Файл EmployeeInfo.txt можно загрузить на сайте книги, адрес которого указан во введении.

    EmployeeInfo.txt - Блокнот - Файл - Правка - Формат - Вид - Справка -
    Барри Берд
    Директор
    5000,00
    Гарриет Риттер
    Капитан
    7000,00
    Ваше имя Лучший менеджер дня
    10000,00
    
    Рис. 8.3. Содержимое файла EmployeeInfo. txt

- внимание - Ради простоты листинга 8.2 необходимо, чтобы файл EmployeeInfo. txt заканчивался пустой строкой. Поэтому, введя последнее число 10000, 00, не забудьте нажать клавишу $<Enter>$ для завершения строки. Обратите внимание на символ курсора после числа 10000, 00 на рис. 8.3. Он свидетельствует о том, что после этого числа есть пустая строка. Если ее не будет, выполнение кода листинга 8.2 завершится аварийно.
