# Определение класса
Как хорошо было бы, если бы каждая часть программного обеспечения делала то, что нам нужно! В идеальном мире вы могли бы просто купить программу, установить ее и сразу же запустить в работу, имея при этом возможность адаптировать ее к любой ситуации и легко обновлять, как только в этом возникает необходимость. К сожалению, мечты остаются мечтами. (Идеала в нашем мире не существует!) Правда в том, что любая программа способна выполнить лишь часть того, что вам нужно в данный момент.
Это одна из причин ошеломляющего успеха концепции объектно- ориентированного программирования. На протяжении многих лет компании покупали дорогостоящие программные продукты и обнаруживали, что программы делают не то, что им нужно. Как действовали компании, оказавшись в подобной ситуации? Они начинали переделывать код. Их программисты изменяли имена переменных, перетасовывали подпрограммы, перерабатывали формулы, но ни к чему хорошему это не приводило. Реальность была таковой, что если программа не делала то, что нужно (даже если она делала нечто очень близкое к тому, что нужно), то улучшить ее поведение было почти невозможно.
Практически всегда наилучшим решением было выбросить дорогостоящую программу и купить новую или написать нужную программу самому. С появлением инструментов объектно-ориентированного программирования ситуация радикально изменилась. Объектно-ориентированная программа с самого начала создается с учетом возможной необходимости вносить в нее изменения в будущем. Когда код правильно структурирован, можно использовать преимущества уже встроенных средств, добавлять новые средства и переопределять средства, которые не во всем удовлетворяют ваши нужды. Важнее всего то, что изменения могут быть "чистыми". Вам не придется взламывать чужие модули. Вместо этого вы сможете добавлять корректно сопрягаемые компоненты, не изменяя существующую структуру кода и его внутреннюю логику. Это идеальное решение.

# Что означает "быть служащим"
Приступая к написанию объектно-ориентированной программы, вы прежде всего продумываете, что собой должны представлять данные. Когда вы пишете программу обработки банковских счетов, вы думаете о том, что такое банковский счет. Кодируя реакцию на щелчок на кнопке, вы думаете о том, что такое кнопка. Когда вы пишете программу, создающую платежные чеки для служащих, вы думаете о том, кто такой служащий.
Для компьютерной программы служащий -- это некий объект, у которого есть два параметра: имя (name) и название занимаемой им должности (jobTitle). Конечно, в реальной жизни у служащего есть и другие параметры, но пока что нам этих двух вполне достаточно. Код, приведенный в листинге 8.1, дает исчерпывающее определение того, кто такой служащий (Employee)
в данном случае.

    Листинг 8.1. Класс Employee
    import static java.lang.System.out;
    public class Employee
    {
        private String name;
        private String jobTitle;
        public void setName (String nameIn) { name = nameIn; }
        public String getName () { return name;}
        public void setJobTitle(String jobTitleIn) { jobTitle = jobTitleIn;}
        public String getJobTitle() { return jobTitle; }
        public void cutCheck(double amountPaid) 
        {
            out.printf("Выплатить служащему ts ", name);
            out.printf("($s) ***$", jobTitle);
            out.printf("8,.2f\n", amountPaid);
        }
    }

Согласно листингу 8.1 каждому служащему (класс Employee) соответствуют семь компонентов. Два из них весьма просты: у каждого служащего есть имя (name) и название его должности (jobTitle).
Что еще есть у служащего? Каждому служащему присущи четыре метода доступа, позволяющих обращаться к его имени и названию должности. Это методы setName, getName, setJobTitle и getJobTitle. (Методы доступа рассматривались в главе 7.)
И наконец, у каждого служащего есть метод cutCheck. Идея состоит в том, что метод, выписывающий платежный чек, должен принадлежать одному из классов. Куда его лучше поместить? Большая часть информации о платежном чеке имеет отношение к конкретному служащему, поэтому лучше всего поместить метод cutCheck в класс Employee.
Вызовы метода prinf, используемые в методе cutCheck, подробнее рассматриваются в разделе "Создание платежного чека".

# Использование класса Employee
В классе Employee (см. листинг 8.1) отсутствует метод main, поэтому он не может служить стартовой точкой программы. Чтобы программу можно было запустить на выполнение, нужно написать класс, содержащий метод main и манипулирующий экземплярами класса Employee. Такой класс показан в листинге 8.2. Фактически он является тестовым модулем для класса Employee.

    Листинг 8.2. Класс, выписывающий платежные чеки для служащих
    import java.util.Scanner;
    import java.io.File;
    import java.io.IOException;
    public class DoPayroll
    {
        public static void main(String args[]) throws IOException
        {
            Scanner diskScanner = new Scanner (new File("EmployeeInfo.txt"));
            for (int empNum = 1; empNum <= 3; empNum++) { payOneEmployee (diskScanner); }
            diskScanner.close();
        }
        static void payOneEmployee (Scanner aScanner)
        {
            Employee anEmployee = new Employee();
            anEmployee.setName (aScanner.nextLine ());
            anEmployee.setJobTitle (aScanner.nextLine());
            anEmployee.cutCheck(aScanner.nextDouble());
            aScanner.nextLine();
        }
    }

Я провел этот сеансна компьютере, находящемся в США. Посколькустрана пребывания имеет значение, в ответ на первый вызов keyboard.nextDouble () я ввел 1000.00 (с точкой перед двумя последними нулями), и Java приняла этот ввод в качестве значения "одна тысяча". Но затем в сеансе JShell выполняется команда Locale.setDefault (Locale. FRANCE), которая указывает Java на то, что мой компьютер находится во Франции. После создания другого экземпляра класса Scanner и повторного вызова метода keyboard. nextDouble () Java принимает значение 1000,00 (с запятой перед последними двумя нулями) в качестве выражения, означающего "mille" (французский эквивалент слова"одна тысяча"). Более того, Java больше не допускает точки в числе 1000.00. Как только вы попробуете ввести 1000. 00 (с точкой), в ответ отобразится сообщение об ошибке InputMismatchException. По умолчанию экземпляр класса Scanner на компьютере воспринимает числа типа double так, как обычно их принято вводить в вашей стране. Если вы вводите числа в соответствии с соглашениями, принятыми в другой стране, то отобразится сообщение об ошибке InputMismatchException. Поэтому при запуске кода из листинга 8.2 числа в файле EmployeeInfo.txt должны иметь формат, принятый в вашей стране.
