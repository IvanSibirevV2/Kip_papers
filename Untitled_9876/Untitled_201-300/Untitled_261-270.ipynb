# ТИПЫ ПЕРЕЧИСЛЕНИЯ И ИНСТРУКЦИЯ SWITCH
В листингах 9.2 и 9.3 используются длинные имена TempScale. FAHRENHEIT и TempScale. CELSIUS. Короткие имена FAHRENHEIT и CELSIUS принадлежат типу перечисления TempScale, определенному в листинге 9.1. За пределами контекста TempScale они не имеют смысла. Попробуйте удалить префикс TempScale из выражения TempScale.FAHRENHEIT, и вы увидите, что компилятор выдаст сообщение об ошибке. Компилятор Java очень придирчив к именам типов и точкам. Однако в некоторых случаях вы сможете создать нужный контекст, например в инструкции switch. Если в его условии приведена переменная типа перечисления, компилятор считает тело инструкции switch контекстом данного типа и не требует указывать имя типа в выражениях case. Например, приведенный ниже код не приводит к ошибке компиляции.

    TempScale scale = TempScale.RANKINE;
    char letter;
    switch (scale)
    {
        case CELSIUS: letter = 'C'; break;
        case KELVIN: letter = 'K'; break;
        case RANKINE:
        case RÉAUMUR:
        case RøMER: letter = 'R'; break;
        default: letter = 'X'; break;
    }

В первой строке этого кода я пишу TempScale. RANKINE, поскольку эта строка находится вне тела инструкции switch. Однако в следующих строках используются выражения case CELSIUS, case KELVIN и case RANKINE без слова TempScale. Кстати, если вы теперь попытаетесь написать, например, TempScale.KELVIN вместо KELVIN, то компилятор выдаст сообщение об ошибке. Единственное правило, которое при этом должно соблюдаться, состоит в том, что количество параметров и их типы в объявлении и вызове конструктора должны совпадать. Например, в листинге 9.3 последний вызов конструктора имеет следующий вид:
>new Temperature(2.73, TempScale.KELVIN)

Java не возражает против такого вызова, поскольку он согласуется с объявлением этого конструктора в листинге 9.2:

>public Temperature (double number, TempScale scale)

# О ЧЕМ ВООБЩЕ ИДЕТ РЕЧЬ?
Рассмотрим первый из двух конструкторов, которые определены в листинге 9.4. Посредством параметра whatever конструктор получает число 32.0. Далее значение этого параметра присваивается полю number. Данный код работает безупречно. Но при его написании мы ввели для параметра отдельное имя -- whatever, и все это только ради того, чтобы передать нужное значение полю number. Что за расточительство? Чтобы отличить параметр от поля number, мы вводим дополнительное имя, которое необходимо лишь для того, чтобы обеспечить кратковременное сохранение значения number.
Выбор имен -- это искусство, а не наука. За долгие годы работы я перепробовал множество подходов. Много лет назад всякий раз, когда мне требовалось дать имя параметру, я получал его путем незначительного искажения имени исходной переменной (например, в данном случае я выбрал бы нечто наподобие numbr или nuhmber.) Затем я пытался превращать некоторые строчные буквы в прописные (например, так: Number или nUMBER). В главе 8 я получал имена параметров путем добавления суффикса In (переменной jobTitle соответствовал параметр jobTitleIn). Но ни одна из этих схем именования меня не устраивает полностью. Мне самомутрудно запоминать эти странные имена, которые я сам же и создал. Хорошая новость состоит в том, чтоотэтих мучений с придумыванием имен для параметров можно легко избавиться. Вы сможете использовать для параметра и переменной одинаковые имена. Чтобы различать их, достаточно воспользоваться ключевым словом this.

Переменная format имеет тип String и содержит строку форматирования, определенную с помощью следующей инструкции:

    final String format = "%5.2f градусов по $s\n";
Обратите внимание на ключевое слово final. В Java оно означает, что переменная format не может изменяться в коде, т.е. она не будет изменена нигде и никогда. Если в каком-либо месте кода встретится инструкция, изменяющая значение format, компилятор сообщит об ошибке и не допустит выполнения программы. В листинге 9.3 ключевое слово final необязательное. Оно добавлено лишь для защиты от неумышленного изменения переменной, которая не должна изменяться. Во многих случаях такая защита очень полезна. Когда программа содержит десятки тысяч строк кода и разрабатывается много месяцев, очень легко забыть, что данная переменная не должна изменяться. Когда же по неосторожности вы модифицируете ее, а потом обнаружите, что она изменилась, вам будет тяжело найти, какая инструкция изменила ее. Поэтому рекомендуется всегда добавлять ключевое слово final, если значение переменной не должно изменяться. Переменные такого вида часто называют константами.

        public void display() { out.printf("85.2f градусов по 8s\n", getNumber(), getScale()); }
    }

Обратите внимание на то, что для доступа к нужным значениям в методе display используются методы getNumber и getScale. Зачем они здесь нужны? Дело в том, что при непосредственном обращении кода класса TemperatureNice к закрытым полям базового класса Temperature компилятор сгенерирует ошибку. Hесмотря на то что в каждом объекте класса TemperatureNice имеются поля number и scale, обращаться к ним напрямую нельзя, поскольку в базовом классе они объявлены как закрытые (с помощью модификатора private). Непосредственное обращение к этим полям возможно только в коде класса
