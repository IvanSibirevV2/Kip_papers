# Аннотации Java
Запись @Override (переопределить) в листинге 8.7 - это аннотация. В Java аннотация сообщает компилятору дополнительную информацию о коде. В частности, аннотация @Override в листинге 8.7 говорит компилятору о том, что он должен проверить, есть ли в базовом классе метод с такими же именем и параметрами. Если такового нет, компилятор должен вывести сообщение об ошибке. Следовательно, если ошибочно ввести:
> public double findPaymentAmount (double hours) {

вместо int hours, как в листингах 8.5 и 8.7, компилятор напомнит, что метод
findPaymentAmount в действительности ничего не переопределяет. В Java есть много других аннотаций, например @Deprecated (устаревший) или @SuppressWarning (подавить предупреждения). Аннотация @ SuppressWarning рассматривается в главе 9.
- технические подробности - Добавлять аннотации в код Java необязательно. Если удалить слово @Override из листинга 8.7, код будет выполняться так же, как и с ним. Но при использовании аннотации уменьшается вероятность ошибки, потому что компилятор проверит, делаете ли вы то, что намеревались (т.е. действительно ли вы переопределяете метод базового класса). Однако некоторые аннотации не такие уж необязательные. Удалить определенные типы аннотаций можно, только заменив их соответствующим кодом Java, который может включать довольно много строк.

# Что такое температура
"Доброе утро! Вас приветствует Служба новостей объекта! Сейчас у нас довольно приятная температура - 73° по Фаренгейту".
Каждое обозначение температуры состоит из двух частей: числа и указателя температурной шкалы. Число - это значение типа double, например 32.0 или 70.52. Но что такое "температурная шкала"? Это строка символов, такая как "Fahrenheit" или "Celsius"? Не совсем. Например, температурной шкалы "Quelploof" не существует, и программа, которая отображает "73 градуса по Kу- эльплуфу", - это плохая программа. Как же ограничить доступные нам варианты небольшим количеством шкал, которые можно использовать? Очевидное решение состоит в том, чтобы объединить их список. В Java такой список называется перечислением и создается с помощью специального ключевого слова enum.
# Что такое температурная шкала
В Java предусмотрено множество способов группирования сущностей. В главе 11 рассматривается группирование объектов с помощью массивов, а в главе 12 -- с помощью коллекций. В данной главе мы рассмотрим группирование с помощью перечислений -- переменных типа enum. (Разумеется, вы не сможете ничего сгруппировать, если не научитесь правильно выговаривать это слово - "и-нум", с ударением на втором слоге; от англ. "enumeration" "перечисление".)
Создать сложный тип enum нелегко, но для создания простого перечисления достаточно написать ряд идентификаторов в фигурных скобках. Пример определения перечисления приведен в листинге 9.1. Имя типа -- TempScale. Тип TempScale содержит перечень доступных температурных шкал.

# Так что же такое температура?
Итак, обозначение температуры состоит из числа и указателя температурной шкалы. Если привести только число, то будет непонятно: это температура по Цельсию, Кельвину, Фаренгейту или Реомюру? Теперь, зная, что такое тип перечисления, мы можем определить температуру как класс, содержащий число и указатель температурной шкалы.

    Листинг 9.2. Класс Тemperature
    public class Temperature
    {
        private double number;
        private TempScale scale;
        public Temperature() { number = 0.0; scale = TempScale.FAHRENHEIT; }
        public Temperature(double number) { this.number = number; scale = TempScale.FAHRENHEIT; }
        public Temperature(TempScale scale) { number = 0.0; this.scale = scale; }
        public Temperature(double number, TempScale scale) {  this.number = number; this.scale = scale; }
        public void setNumber(double number) { this.number = number; }
        public double getNumber() { return number; }
        public void setScale (TempScale scale){ this.scale = scale; }
        public TempScale getScale() { return scale; }
    }
