# Переброска исключения из одного метода в другой
В главе 8, рассматривая чтение данных из файла, мы затронули тему исключений. В процессе работы с файлами на дисках нужно всегда допускать возможность возникновения исключения IOException. Это следовало бы усвоить после прочтения главы 8, и именно поэтому мы указываем в конструкторе, определенном в листинге 10.2, на возможность возникновения соответствующей исключительной ситуации, используя ключевое слово throws.

Но какое отношение к этому имеет метод main в листинге 10.3? В нем нет обращений к файлу, откуда же в нем может возникнуть исключение IOException? Дело в том, что исключения напоминают горячую картофелину, которую перебрасывают из рук в руки. Получив картофелину, вам следует либо съесть ее, либо быстренько перебросить кому-то другому, чтобы не обжечь руки. Примерно то же самое можно сказать и об исключениях. Метод, сгенерировавший или получивший исключение, должен либо обработать его в блоке try (как это делается, мы рассмотрим в главе 13), либо передать обработчику, находящемуся в другом месте программы. Список передаваемых исключений должен указываться в заголовке метода после ключевого слова throws.

В листинге 10.2 в заголовке конструктора TeamFrame с помощью ключевого слова throws указано, что в случае генерации исключения IOException оно будет передаваться внешнему обработчику. Но кто именно будет его получателем, на кого в этой цепочке кода будет возложена ответственность за обработку потенциальных исключений IOException? Ответ очень прост: за обработку передаваемых исключений отвечает код, который вызвал метод, сгенерировавший исключение. А что это за код, из которого вызывается конструктор TeamFrame? Конечно же, это код, приведенный в листинге 10.3. Таким образом, в данном случае "горячая картофелина" передается методу main, который, как видно из листинга 10.3, также объявляет исключение IOException как передаваемое. Именно так работает директива throws в программах на Java.

- запомни - Если в вызываемом методе установлена директива throws, вызывающий метод должен либо обработать исключение, либо передать его своему вызывающему методу. Но для этого в его заголовке также должна быть установлена директива throws. Обработка исключений подробно рассматривается в главе 13.
- технические подробности - В этот момент проницательный читатель наверняка спросит "А кто вызывает метод main и кто получает сгенерированное в нем исключение IOException?" Ответ таков: метод main вызывается виртуальной машиной Java (JVM), которая в конечном счете и обработает возможное исключение IOException в случае генерации его кодом из листинга 10.3. Сама же JVM и позаботится об отображении соответствующего сообщения об ошибке и завершении выполнения вашей программы. Чрезвычайно удобный механизм!

# Статическая инициализация
В листинге 10.4 поле decFormat статическое. Это имеет смысл, потому что данное поле определяет формат вывода отношения статических величин totalOfAverages/playerCount. Иными словами, коду нужна лишь одна переменная, форматирующая числа. Если требуется форматировать вывод нескольких чисел, это может быть сделано с помощью одного и того же объекта decFormat, принадлежащего классу в целом. Создание отдельного объекта decFormat для каждого игрока было бы сплошным расточительством.
Однако при объявлении поля decFormat как статического возникает небольшая проблема. Чтобы настроить форматирование, необходимо вызвать ряд методов объекта decFormat, таких как decFormat.setMaximumIntegerDigits (0). Вызовы этих методов нельзя вставлять как попало, в любое место класса PlayerPlus. Ниже приведен пример плохого, неправильного, неудачного, неверного кода, нарушающего правила, принятые в Java.

В листинге 10.5 объект GridLayout добавил две дополнительные строки: одну -- для отображения черточек, другую - для вывода среднего показателя команды. В каждой из этих двух строк размещены по два объекта Label.
- Первая строка содержит пустую надпись и надпись с черточками. Пустая надпись служит заполнителем. Компоненты добавляются в GridLayout построчно, начиная с левого конца строки и переходя к правому концу. Без пустой надписи черточки оказались бы слева, а справа появилась бы надпись Средний бэттинг команды.
- Во второй строке отображаются надпись Средний бэттинг команды и значение, возвращенное методом findTeamAverage String, равное , 106. Рассмотрим подробнее вызов метода, получающего число , 106. Вызов выглядит так:
> PlayerPlus.findTeamAverageString ()

Как видите, вызов выполняется в необычной форме: имя класса имя метода ()
Для вас это нечто новенькое. До сих пор мы добавляли в качестве префикса имя объекта, а не класса. Почему же здесь добавлено имя класса? Потому что вызывается статический метод, принадлежащий классу, а не конкретному объекту. То же самое справедливо и при обращении к статическим полям. Поэтомузапомните, что при обращении к статическому полю или методу необходимо указывать имя класса, а не объекта.
- совет - В некоторых случаях компилятор можно обмануть, обратившись к статическому полю или методу через имя объекта, а не класса. Например, после определенной реорганизации инструкций в листинге 10.5 можно было бы использовать выражение player.findTeam AverageString().

# Статический импорт
Мы уже много говорили о статических полях и методах, но статические компоненты фигурировали в нашем коде еще в начальных главах книги. Например, в главе 3 вы познакомились с методом System.out.println. Имя System ссылается на класс, а имя out - на статическое поле этого класса. Вот почему для импорта поля out используется ключевое слово static:
> import static java.lang.System.out;
