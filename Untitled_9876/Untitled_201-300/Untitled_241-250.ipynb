# Создание подкласса
Определение класса Employee приведено в листинге 8.1. Мы можем использовать то, что есть в листинге 8.1, и, расширив определение, создать новые, более специализированные классы. Определение одного из таких классов (FullTimeEmployee) приведено в листинге 8.3. 

    Листинг 8.3. Класс FullTimeEmployee, представляющий штатного сотрудника
    
    public class FullTimeEmployee extends Employee 
    {
        private double weeklySalary;
        private double benefitDeduction;
        public void setWeeklySalary(double weeklySalaryIn) { weeklySalary = weeklySalaryIn; }
        public double getWeeklySalary(){ return weeklySalary; }
        public void setBenefitDeduction(double benefitDedIn) { benefitDeduction = benefitDedIn; }
        public double getBenefitDeduction() { return benefitDeduction; }
        public double findPaymentAmount () { return weeklySalary - benefitDeduction; }
    }

Каждый экземпляр класса FullTimeEmployee имеет два закрытых поля: weeklySalary (оплата за неделю) и benefit Deduction (пенсионные отчисления). Hо это не все поля, доступные в классе FullTimeEmployee. Первая строка листинга 8.3 говорит о том, что класс FullTimeEmployee pасширяет существующий класс Employee. Это означает, что, кроме полей weeklySalary и benefit Deduction, каждый экземпляр класса FullTimeEmployee имеет еще и поля name и jobTitle. Эти два поля определены в классe Employee (см. листинг 8.1). В листинге 8.3 имеется "волшебное" слово extends (расширяет). Если некоторый класс расширяет существующий класс, то расширяющий класс автоматически наследует функциональность, определенную в существующем класcе. Таким образом, класс FullTimeEmployee наследует поля name и jobTitle. Кроме того, он наследует все методы, определенные в базовом классе: setName, getName, setJobTitle, getJobTitle и cutCheck. Класс FullTimeEmployee является подклассом, или производным классом, класса Employee, а класс Employee - суперклассом по отношению к классу FullTimeEmployee. Рассмотрение отношений между такими классами возможно также в терминах кровного родства. Говорят, что класс FullTimeEmployee является дочерним по отношению к классу Employee, тогда как последний является родительским по отношению к первому. Производный класс можно представлять себе как класс, в определение которого перенесены все поля и методы базового класса (листинг 8.4), хотя эта аналогия и не совсем точна.

    Листинг 8.7. Еще один производный класс
    public class PartTimeWithOver extends PartTimeEmployee
    {
        @Override
        public double findPaymentAmount(int hours)
        {
            if(hours <= 40) { return getHourlyRate() * hours; }
            else { return getHourlyRate() * 40 + getHourlyRate() * 2 * (hours - 40); }
        }
    }

Отношения между классами показаны на рис. 8.8. Класс PartTimeEmployee, производный по отношению к Employee, является базовым по отношению к PartTimeWithOver. Hа практике в объектно-ориентированных программах такие цепочки - обычное дело. Часто они намного длиннее, чем данная цепочка из двух звеньев.
Класс PartTimeWithOver расширяет класс PartTimeEmployee, но не совсем так, как в предыдущих примерах. В данном случае класс PartTimeWithOver сам выбирает, что он хочет унаследовать от базового класса.
