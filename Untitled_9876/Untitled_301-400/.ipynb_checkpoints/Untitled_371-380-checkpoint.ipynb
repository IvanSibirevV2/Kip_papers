# Побочные эффекты лямбда-выражений
Вот интересный пример лямбда-выражения:

    (sale) -> System.out.println(sale.getPrice())
    
Это лямбда-выражение делает примерно то же, что и следующий код.

    void display(Sale sale) { System.out.println(sale.getPrice()); }
    display(sale);
    
Лямбда-выражение получает объекты из потока и вызывает для каждого из них метод, похожий на метод display. Слово void в заголовке метода display означает, что метод не возвращает никакого значения. Вызывая метод display (или используя эквивалентное ему лямбда-выражение), вы не ожидаете получить какое-либо значение. Вместо этого вы рассчитываете на то, что код должен что-либо сделать в ответ на вызов метода (например, вывести текст на экран). Чтобы провести четкую границу между формулировками "вернуть значение" и "что-либо сделать", в функциональном программировании существует специальный термин, заменяющий формулировку "что-либо сделать без возвращения значения": побочный эффект. В функциональном программировании побочные эффекты считаются "гражданами" второго сорта, т.е. вынужденной мерой, тактикой, к которой следует прибегать, когда вы не можете просто вернуть результат. К сожалению, отображение информации на экране (а это встречается в компьютерных программах сплошь и рядом) относится к разряду побочных эффектов. Любая программа, которая отображает свой вывод (на экране, на бумаге или на чайных листиках, плавающих в стакане), не является чисто функциональной программой.

# Таксономия лямбда-выражений
B Java лямбда-выражения разбиты на 45 категорий. Краткий их список приведен в табл. 12.2.

    Таблица 12.2. Некоторые виды лямбда-выражений
    Название     Описание     Пример
    Function     Имеет один параметр; выдает результат любого типа     (sale) -> sale.price
    Predicate     Имеет один параметр; выдает результат, оцениваемый как булево значение     (sale) -> sale.item. equals ("DVD")
    BinaryOperator     Имеетдва параметра; выдает результат того же типа     (price1, price2) -> pricel + price2
    Consumer     Имеет один параметр; не выдает никакого результата     (sale) -> System.out. println(sale.getPrice())
    
- запомни - Категории, приведенные в табл. 12.2, не являются взаимоисключающими. Например, каждое лямбда-выражение категории Predicate одновременно относится к категории Function. (Каждое выражение категории Predicate имеет один параметр и возвращает результат. В данном случае результат имеет тип boolean.)

    import java.text.NumberFormat;
    import java.util. ArrayList;
    public class TallySales
    {
        public static void main(String[] args)
        {
            ArrayList<Sale> sales = new ArrayList<Sale>();
            NumberFormat currency = NumberFormat.getCurrencyInstance ();
            fillTheList (sales);
            double total = sales.stream()
                .filter((sale) -> sale.getItem().equals("DVD")) 
                .map((sale) -> sale.getPrice())
                .reduce(0.0, (pricel, price2) -> pricel + price2)
            ;
            System.out.println(currency.format (total));
        }
        static void fillTheList (ArrayList<Sale> sales)
        {
            sales.add(new Sale("DVD", 15.00));
            sales.add(new Sale("Книга", 12.00));
            sales.add(new Sale("DVD", 21.00));
            sales.add(new Sale("CD", 5.25));
        }
    }

# К чему все эти холопы/хлопоты?

Цепочка вызовов методов в листинге 12.5 делает то же самое, что и цикл в листинге 12.4. Однако в первом случае используются концепции функционального программирования. Возникает вопрос: а что это нам дает? Стоит ли использовать код, приведенный в листинге 12.5, вместо кода, приведенного в листинге 12.4? Конечно, стоит. На протяжении нескольких последних лет широкое распространение получили многоядерные процессоры. Благодаря нескольким ядрам процессор может выполнять несколько инструкций одновременно, что сокращает время выполнения программ в два, четыре, восемь, а то и больше раз. Когда работа распределяется между несколькими ядрами, программа выполняется намного быстрее. Но как это сделать? Можно было бы модифицировать императивный код, приведенный в листинге 12.4. Например, используя ряд трюков, можно организовать выполнение различных итераций цикла на разных ядрах. Но в таком случае результирующий код получается очень запутанным. Чтобы обеспечить нормальную работу кода, приходится брать на себя детальное управление работой цикла, выполняя всевозможные проверки, гарантирующие корректность окончательного значения total.
В противоположность этому модифицировать функциональный код не составляет большого труда. Чтобы воспользоваться преимуществами многоядер- ных процессоров, в листинге 12.5 достаточно изменить всего одно слово!

# НЕТ ПЕРЕМЕННЫХ - НЕТ ПРОБЛЕМ!
Вернемся к задачам, примеры которых приводились в начале раздела "Функциональное программирование": несколько клиентов одновременно пытаются обновить информацию о котировках акций Oracle или два посетителя интернет-магазина пытаются купить один и тот же товар в одно и то же время. Источником проблем являются разделяемые (совместно используемые) данные. Сколько клиентов могут получить доступ к данным о стоимости акций Oracle? Сколько клиентов могут одновременно получить доступ к кнопке Купить на веб-странице? Сколько ядер процессора могут пытаться изменить значение одной и той же переменной? Если вы избавитесь от совместно используемых данных, проблемы многоядерной обработки улетучатся сами собой.
В императивном программировании переменная - это место, в котором инструкции совместно используют хранящееся там значение. Можете ли вы избежать использования переменных в своем коде?
Сравним цикл из листинга 12.4 с кодом в стиле функционального программирования, приведенным в листинге 12.5. В листинге 12.4 переменная total используется всеми итерациями цикла. Поскольку каждая итерация может изменить значение total, вам не удастся назначить выполнение каждой отдельной итерации разным ядрам процессора. Если бы вы это сделали, то возник бы риск, что два ядра одновременно пытаются изменить значение total. (При этом были бы велики шансы, что ни одному ядру не удастся корректно обновить значение переменной!) Однако в коде, написанном в стиле функционального программирования (см. листинг 12.5), переменная total отсутствует. В функциональной версии кода метод reduce применяет операцию суммирования к значениям, поступающим из входного потока. Входной поток для этого метода "выталкивается" из вызова предыдущего метода (а именно - из метода mар) и поэтому не обладает именем. Это очень хорошо. Для хранения потока значений вам даже не нужна переменная.
Функциональное программирование, в отличие от императивного, помогает избавиться от лишних переменных, уменьшая долю совместно используемых данных. Многие из трудностей, связанных с использованием многоядерных процессоров, растворяются в воздухе. Ваш код получает возможность задействовать одновременно несколько ядер. Это элегантное решение важной задачи программирования.
