Если бы метод main не являлся статическим, то обратиться к нему можно было бы только после создания соответствующего объекта. Но в момент запуска программы никаких объектов еще не существует. Воспоминание об этом сообщении так глубоко укоренилось в моем подсознании, что я помню о нем до сих пор, хотя теперь знаю, почему оно появлялось так часто. Сейчас могу даже вызвать его умышленно. Прежде чем я смогу объяснить вам, почему появляется это сообщение и как исправить ошибку, следует ознакомиться с кое-какой терминологией. Если метод не является статическим, значит, он не статический. (Ничего удивительного, не так ли?) Теперь вы знаете необходимые термины, и я могу приступить к объяснению ошибки. В любом из этих случаев вы получите вышеупомянутое сообщение. Вы берете нечто, принадлежащее какому-то объекту (нестатическую сущность), и помещаете его в контекст, в котором объекты невидимы.
