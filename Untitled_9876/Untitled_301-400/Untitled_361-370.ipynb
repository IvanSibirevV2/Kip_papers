# Другие классы коллекций Java
Класс ArrayList, который используется в этой главе во многих примерах, является лишь верхушкой айсберга коллекций Java. В библиотеке Java содержится много других классов коллекций, каждый из которых имеет свои преимущества. Их краткий список приведен в табл. 12.1.

Таблица 12.1. Некоторые классы коллекций Java
- Имя класса - Характеристика -
- ArrayList - Динамический массив, который может изменять свой размер во время выполнения программы
- LinkedList - Связный список значений, каждое из которых имеет поле, указывающее на следующее значение в списке
- Stack - Структура в виде стека, растущего снизу вверх. Эта структура оптимизирована для доступа к верхнему значению. Вы сможете легко добавить значение на вершину стека или удалить находящееся там значение
- Queue - Структура в виде очереди, растущей с одной стороны. Эта структура оптимизирована для добавления значений на одной стороне очереди (в конце) и удаления значений на другой стороне (в начале)
- PriorityQueue - Очередь с приоритетом - структура, подобная очереди, которая обеспечивает перемещение некоторых (высокоприоритетных) значений ближе к началу очереди
- HashSet - Неупорядоченная коллекция, исключающая появление дубликатов
- HashMap - Коллекция пар "ключ-значение"

Каждый класс коллекции имеет собственный набор методов, наследуемых от абстрактного класса AbstractCollection, который является предком всеx классов коллекций.

# Функциональное программирование
В 1953-1957 годах Джон Бэкус вместе с коллегами разработал язык программирования FORTRAN, послуживший основой для создания множества языков программирования ХХ столетия. Соответствующий стиль программирования в духе "сделай то, сделай это" получил название императивного программирования. Спустя несколько лет после возникновения языка FORTRAN Джон Маккарти создал язык под названием LISP. В отличие от FORTRAN, язык LISP представляет функциональное программирование. В чисто функциональном программировании избегают инструкций в стиле "сделай то, сделай это". Вместо этого используются инструкции в стиле "вот как следует выполнять такое-то преобразование, если это надо будет сделать". По целому ряду причин императивное программирование заняло доминирующее положение. Поэтому неудивительно, что язык Java в существенной степени ориентирован на императивный стиль программирования. Однако в последнее время функциональное программирование все чаще заявляет о себе как о мощном и плодотворном подходе к разработке кода. В развитие этой идеи в Java введено понятие потока. Потоки подобны итераторам, но отличаются от них тем, что не требуют использования метода next. После того как поток создан, он может автоматически выдавать значения, входящие в коллекцию. Чтобы получать значения из потока, не нужно вызывать метод next. В действительности типичный поток не имеет данного метода. Каким образом воплотить эти идеи в виде программного кода? Как создать поток, извлекающий значения? Как сообщить потоку, когда начать извлечение значений и куда они должны направляться? Ответы на эти и другие вопросы даются в нескольких разделах.
# Лямбда-выражения
В 1930-х годах математик Алонзо Черч использовал греческую букву "лямбда" (А) для представления некоторой математической конструкции, которая создается "на лету"1. На протяжении нескольких последующих десятилетий эта идея обсуждалась на страницах математических и компьютерных журналов. В наши дни в Java термин лямбда-выражение представляет небольшой фрагмент кода, который служит одновременно и объявлением, и вызовом метода, создаваемого "на лету".
Простейшее лямбда-выражение. Ниже приведен пример простейшего лямбда-выражения:

    (sale)-> sale.getItem().equals ("DVD")
    
Его смысл проиллюстрирован на рис. 12.7.

- (sale) - Выбрать очередной объект извходного потока. Временно присвоить переменной sale ссылкуна этотобъект.
- sale.getItem() - Получить значение поля Item данного объекта sale.
- sale.getItem().equals ("DVD") - Проверить, чему равен результат сравнения значения поля Item объекта sale и строки "DVD": true или false.
