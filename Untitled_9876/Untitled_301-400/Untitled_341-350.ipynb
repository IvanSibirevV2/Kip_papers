# Ограничения массивов
Массивы - очень полезное средство, но им все же присущ ряд ограничений. Предположим, например, что в массиве хранятся имена ста клиентов, упорядоченных некоторым образом. Приведенный ниже код создает массив, состоящий из ста элементов типа String.

    String name[] = new String[100]; for (int i = 0; i < 100; i++) name[i] = new String();

Все идет хорошо, пока в один прекрасный день не появляется 101-й клиент. Пока программа выполняется, вы пытаетесь ввести данные для нового клиента, отчаянно надеясь на то, что массив расширится в соответствии с вашими нуждами.
Однако эти надежды напрасны. Программа генерирует исключение ArrayIndexOutOfBoundsException и аварийно завершается.
"В будущем я буду создавать массивы длиной не менее 1000", - говорите вы себе. И когда это будущее наступает, вы пишете следующее.

    String name[] = new String [1000]; for (int i = 0; i < 1000; i++) { name[i] = new String(); }

Однако вскоре происходит экономический спад, и вместо прежних ста одного клиента у фирмы остаются только три. Поскольку вы зарезервировали память для тысячи клиентов, большой объем памяти тратится понапрасну. Какое расточительство!
А что если экономического спада не случится? Вы будете спокойно жить со своим массивом размером 1000, имея 825 клиентов. Элементы с индексами от 0 до 824 будут использоваться, а элементы с индексами от 825 до 999 -- дожидаться своего часа, когда они будут заполнены данными очередных клиентов. Наконец наступает день, когда у вас появляется новый клиент. Поскольку данные о клиентах упорядочены определенным образом (например, по фамилиям, номерам социальной страховки или как-то иначе), вы хотите втиснуть новые данные в соответствующее место массива. Проблема состоит в том, что этому клиенту должен соответствовать элемент массива с индексом 7. Как быть? Ваш вероятный порядок действий должен быть таким. Вы берете имя, хранящееся в элементе с номером 824, и перемещаете его в элемент 825. Затем вы берете имя, хранящееся в элементе 823, и перемещаете его в элемент 824.
