Ключ к устранению ошибки - имя исключения java.lang. NumberFormat Exception (неправильный формат числа), отображенное на консоли рабочей средой Eclipse. Оно означает, что класс исключения NumberFormatException находится в пакете java.lang. Вызов метода Integer.parseInt каким-то образом привел к появлению указанного исключения. Краткое описание метода Integer. parseInt можно найти в главе 11.

В язык Java встроен механизм обработки исключений. Используя этот механизм, программа на Java может обнаружить, что операцию нельзя выполнить, и отреагировать на это созданием некоего совершенно нового объекта. Согласно официальной терминологии этот процесс называется генерированием исключения. Упомянутый новый объект - объект исключения (экземпляр класса Exception) -- словно обжигающая руки картофелина, передается от одного участка кода к другому, пока где-то в коде не будет принято решение перехватить исключение. Когда исключение перехвачено, программа выполняет код восстановления, отменяет исключение и переходит к выполнению следующей нормальной инструкции, как будто ничего не случилось (рис. 13.2). Перехват и обработка исключений выполняются с помощью нескольких ключевых слов Java.
