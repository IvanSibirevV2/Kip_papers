# Блок catch с несколькими типами исключений
Начиная с версии Java 7 допускается перехват нескольких исключений в одном блоке catch. Предположим, в предыдущем примере вы не хотите проводить никаких различий между исключениями NumberFormatException и OutOfRangeException. В таком случае можно переписать часть листинга 13.6 следующим образом.

    try {
        int numBoxes = Integer.parseInt (numBoxesIn);
        if (numBoxes < 0){throw new OutOfRangeException();}
        if (numBoxes > 1000) { throw new NumberTooLargeException();}
        out.print("Общая стоимость равна: ");
        out.println(currency.format (numBoxes * boxPrice));
    }
    catch (NumberFormatException | OutOfRangeException e) 
    {
        out.print (numBoxesIn);
        out. println("? Это невозможно!");
    }
    catch (Exception e)
    {
        out.print("Что-то здесь не так, ");
        out.print ("но что именно, непонятно.");
        out.println("Тип исключения неизвестен.");
    }

Символ | сообщает Java о том, что блок должен перехватить исключение, принадлежащее к любому из указанных типов: NumberFormatException или OutOfRangeException. Если блок try crенерирует исключение любого из этих типов, компьютер отобразит сообщение Это невозможно!. Если же блок try сгенерирует исключение какого-то другого типа, будет выполнен следующий блок catch, потому что исключение Exception является базовым для исключений любого типа.

# Не будем чрезмерно осторожничать
Персонаж рассказа А. П. Чехова "Человек в футляре" всегда, даже в солнечный день, ходил с зонтиком, на случай, если вдруг пойдет дождь, и часто повторял фразу "Как бы чего не вышло". К тому же компилятор Java не позволяет нам впасть в паранойю и отказывается компилировать код, перехва- тывающий исключения, которые не могут произойти.

    // Неправильный код!

    Exception java.io.IOException is never thrown
