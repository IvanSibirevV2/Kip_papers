# ИНСТРУКЦИИ И ВЫРАЖЕНИЯ
Описать поведение префиксных и постфиксных операторов инкремента и де- кремента можно одним из двух способов: так, как это почти все понимают, и так, как это есть на самом деле. Приведенное выше объяснение (в терминах "сначала" и "затем") принадлежит к первой категории (так, как это почти все понимают). К сожалению, эта концепция не совсем правильная. Увидев в коде ++ или --, вы, возможно, будете рассуждать в терминах времени - до того и после того. Но профессиональные программисты иногда находят такие причудливые способы применения операторов ++ и --, что понятия "сначала" и "затем" становятся бессмысленными. Поэтому, если вы хотите стать профессиональным программистом, сосредоточьтесь и постарайтесь представить себе эти операторы в терминах инструкций и выражений.
В первую очередь, вспомните, что инструкция приказывает компьютеру сделать что-либо, а выражение возвращает значение (иными словами, выражение равно результирующему значению). Ккакой из этих двух категорий принадлежит предложение numberOfBunnies++? Как ни странно, к обеим! Это инструкция и выражение одновременно.
В предыдущем примере, перед тем как компьютер выполняет код out.
println (numberOfBunnies++), переменная numberOfBunnies равна 28.
- Pассматриваемое как инструкция предложение numberOfBunnies++ приказывает компьютеру увеличить значение numberOfBunnies на единицу.
- Рассматриваемое как выражение значение numberOfBunnies++ равно 28, а не 29.

Таким образом, несмотря на то что компьютер добавляетединицук переменной numberOfBunnies, код out.println(numberOfBunnies++) отображает 28.
Почти все, что вы прочитали о фразе numberOfBunnies++, справедливо и по отношению к фразе ++numberOfBunnies. Единственное отличие состоит в том, что ++numberOfBunnies ведет себя более "прилично", т.е. является "правильным выражением", а именно равно результату суммирования переменной и единицы, в то время как numberOfBunnies++ не равно ему.
- Рассматриваемое как инструкция предложение ++numberOfBunnies приказывает компьютеру увеличить значение numberOfBunnies на единицу.
- Рассматриваемое как выражение предложение ++numberOfBunnies равно результату добавления единицы, т.е. числу 29.
Следовательно, инструкция out.println (++numberOfBunnies) приказывает компьютеру прибавить к переменной единицу, и вызов данного метода ведет себя так же, как вызов out.println (29). Введите текст, выделенный полужирным, построчно в JShell и понаблюдайте за реакцией JShell.

# техничесике подробности
Эта инструкция выполняет определенные действия (часто называемые "сбором мусора") по отключению программы от компьютерной клавиатуры. (Объем работы, которую для этого необходимо выполнить, больше, чем вы, возможно, себе представляете!) Если опустить этуинструкцию в листинге 5.1, ничего страшного не произойдет. Как правило, виртуальная машина Java очень тщательно подчищает за собой весь "мусор" в конце работы. Но вызов метода close () для явного отключения от клавиатуры является общепринятой практикой, а некоторые IDE даже предупредят вас об отсутствии инструкции keyboard.close (), если вы забыли включить ее в программу. Поэтому во всех приводимых мною примерах переменные типа Scanner обязательно закрываются в конце работы программы. В главе 13 будет продемонстрирован более надежный способ включения вызовов keyboard.close ( ) в программы на Java. Когда в программе встречается вызов System.out.println (), осуществляется вывод данных на компьютерный экран. Тогда почему в таком случае после вызова всех методов System.out.println () мы не вызываем метод close () ? Ответ на этот вопрос довольно прост. В листинге 5.1 ваш код самостоятельно подключает клавиатуру к программе путем вызова метода new Scanner (in). Поэтому позднее код должен сам обеспечить "уборку мусора" с помощью метода close (). В случае же вызова System. out.println() ваш код никакого участия в подключении экрана к программе не принимает. (Переменная out ссылается на выходной поток PrintStream, но вам не приходится выполнять вызов new PrintStream ( ) для подготовки вызовов System.out.println().) Это делает вместо вас виртуальная машина Java. В ее коде (скрытом от вас) и содержится вызов метода new PrintStream (), выполняющий всю необходимую подготовительную работу. И когда это потребуется, виртуальная машина, как и подобает законопослушному коду, своевременно вызовет метод close () без каких-либо усилий с вашей стороны.

Но если поменять исходную позицию на противоположную, то ситуация обратится, и более вероятным станет выпадение "решки". Иными словами, подбрасывание монеты не гарантирует принятие "справедливого" решения.
