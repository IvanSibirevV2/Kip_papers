- Внимание
    - Если ранее вы писали программы на других языках, то должны быть знакомы с кодировкой ASCII, применяемой в большинстве языков программирования. Однако в Java вместо ASCII применяется другой стандарт кодирования символов: Unicode. В то время как в кодировке ASCII для представления каждого символа используется 8 бит, в кодировке Unicode возможны варианты представления символов с помощью 8, 16 или 32 битов. Если возможности ASCII ограничены буквами латинского алфавита (латиницы), то стандарт Unicode позволяет представлять знаки большинства письменных языков. Единственная проблема состоит в том, что некоторые библиотечные методы Java ориентированы на работу только с 16-битовыми символами Unicode. Иногда это может доставлять определенные неудобства. Если вы используете некий метод для вывода на экран слова Hello, а в результате отображается слово E L L O, обратитесь к описанию метода и выясните, как он обрабатывает символы Unicode.

Переменная allTenOK имеет тип boolean. Чтобы вычислить значение этой переменной, программа сравнивает значение numberOfPeople с числом 10. (Символы >= означают больше или равно.) Сейчас нам следует уточнить используемую терминологию. В Java любая часть кода, возвращающая значение, является выражением. Если записать WeightOfAPerson = 65; то в этом предложении 65 - выражение (значением которого является число 65). В предложении результат = 2 + 2; правая часть (2 + 2) также является выражением, поскольку возвращает значение 4. Точно так же выражением является правая часть следующего предложения (ee значение определяется значениями переменных ElevatorWeightLimit и WeightOfAPerson на момент выполнения вычисления): int numberOfPeople = ElevatorWeightLimit / WeightOfAPerson;
