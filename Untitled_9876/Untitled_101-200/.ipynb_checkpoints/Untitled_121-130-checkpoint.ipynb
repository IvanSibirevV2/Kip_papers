# ИНСТРУКЦИИ И ВЫРАЖЕНИЯ
Описать поведение префиксных и постфиксных операторов инкремента и де- кремента можно одним из двух способов: так, как это почти все понимают, и так, как это есть на самом деле. Приведенное выше объяснение (в терминах "сначала" и "затем") принадлежит к первой категории (так, как это почти все понимают). К сожалению, эта концепция не совсем правильная. Увидев в коде ++ или --, вы, возможно, будете рассуждать в терминах времени - до того и после того. Но профессиональные программисты иногда находят такие причудливые способы применения операторов ++ и --, что понятия "сначала" и "затем" становятся бессмысленными. Поэтому, если вы хотите стать профессиональным программистом, сосредоточьтесь и постарайтесь представить себе эти операторы в терминах инструкций и выражений.
В первую очередь, вспомните, что инструкция приказывает компьютеру сделать что-либо, а выражение возвращает значение (иными словами, выражение равно результирующему значению). Ккакой из этих двух категорий принадлежит предложение numberOfBunnies++? Как ни странно, к обеим! Это инструкция и выражение одновременно.
В предыдущем примере, перед тем как компьютер выполняет код out.
println (numberOfBunnies++), переменная numberOfBunnies равна 28.
- Pассматриваемое как инструкция предложение numberOfBunnies++ приказывает компьютеру увеличить значение numberOfBunnies на единицу.
- Рассматриваемое как выражение значение numberOfBunnies++ равно 28, а не 29.

Таким образом, несмотря на то что компьютер добавляетединицук переменной numberOfBunnies, код out.println(numberOfBunnies++) отображает 28.
Почти все, что вы прочитали о фразе numberOfBunnies++, справедливо и по отношению к фразе ++numberOfBunnies. Единственное отличие состоит в том, что ++numberOfBunnies ведет себя более "прилично", т.е. является "правильным выражением", а именно равно результату суммирования переменной и единицы, в то время как numberOfBunnies++ не равно ему.
- Рассматриваемое как инструкция предложение ++numberOfBunnies приказывает компьютеру увеличить значение numberOfBunnies на единицу.
- Рассматриваемое как выражение предложение ++numberOfBunnies равно результату добавления единицы, т.е. числу 29.
Следовательно, инструкция out.println (++numberOfBunnies) приказывает компьютеру прибавить к переменной единицу, и вызов данного метода ведет себя так же, как вызов out.println (29). Введите текст, выделенный полужирным, построчно в JShell и понаблюдайте за реакцией JShell.

# Операторы присваивания
Прочитав предыдущий раздел об операторах, которые добавляют или вычитают единицу, вы, наверное, подумали: "А можно ли использовать эти операторы для добавления 2, 5 или 1000?" Можно ли написать numberOfBunnies++++? Нет, нельзя. При наличии такой записи компилятор выдаст сообщение об ошибке в коде.

Как видите, операторов присваивания довольно много, и их можно использовать по-разному. С их помощью можно суммировать, вычитать, умножать и делить переменную на любое число. Обратите внимание на то, что выражение += 5 прибавляет 5, а выражение $*$= 2 умножает на 2. Можно даже использовать значение другого выражения (в листинге 4.8 это numberExtra) в качестве операнда. Последние две инструкции в листинге 4.8 демонстрируют расширенные свойства операторов присваивания: их можно использовать в других инструкциях. Предпоследняя инструкция вызова println() вычитает 7 из переменной numberOfBunnies. До этого переменная была равна 172, следовательно, инструкция выводит на консоль число 165.
Последняя инструкция в листинге 4.8 одновременно присваивает переменной значение 100 и выводит его на консоль. Следовательно, инструкцию присваивания можно применять в любой другой инструкции. Не забывайте об этой полезной особенности оператора присваивания.

- технические подробности
    - Каждый оператор присваивания играет в коде двойную роль: работает и как инструкция, и как выражение. Во всех случаях выражение равно присваиваемому значению. Например, перед выполнением инструкции out.println(numberOfBunnies -= 7) переменная numberOfBunnies была равна 172. В качестве инструкции запись numberOfBunnies = 7 приказывает компьютеру вычесть из переменной число 7 (в результате чего переменная становится равной 165). А в качестве выражения запись numberOfBunnies = 7 должна быть чему-то равна, и равна она присваиваемому значению. Это означает, что оператор out.println(numberOfBunnies = 7) выводит на консоль то же значение, что и out.println(165).
    
Если вы забыли, в чем состоит различие между инструкциями и выражениями, прочитайте еще раз врезку "Инструкции и выражения".

В листинге 5.1 приведена программа, иллюстрирующая использование инструкции if. На рис. 5.1 показан результат двукратного выполнения программы.

    //Листинг 5.1. Игра с угадыванием числа
    import static java.lang.System.out;
    import java.util.Scanner;
    import java.util.Random;

    public class GuessingGame
    {
        public static void main(String args[])
        {
            Scanner keyboard = new Scanner(System.in);
            out.print("Введите число от 1 до 10: ");
            int inputNumber = keyboard.next Int();
            int randomNumber = new Random().nextInt(10) + 1;
            if (inputNumber == randomNumber)
            {
                out.println("********");
                out.println("*Вы выиграли.*");
                out.println("**************");
            }

            else
            {
                out.println("Вы проиграли.");
                out.print ("Случайное число равно");
                out.println(randomNumber + ".");
            }
            out.println("Спасибо за игру. ");
            keyboard.close();
        }
    }
