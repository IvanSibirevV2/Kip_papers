При входе в цикл программа проверяет, истинно ли условие продолжения цикла. Если условие равно false, инструкции в теле цикла не выполняются, иногда даже ни разу. В действительности можно легко написать цикл while, инструкции которого никогда не выполняются (хотя я ума не приложу, для чего нужны такие циклы). Ниже приведен пример такого цикла.

    int twoPlusTwo = 2 + 2;

Как бы там ни было, цикл while -- наиболее часто используемая разновидность циклов Java. В частности, он наиболее пригоден в ситуациях, когда требуется предварительная проверка выполнения некоторых условий. Предположим, например, что программа перечисляет деньги с текущего счета на ряд других счетов в цикле. Прежде чем начать транзакцию, программа должна проверить, есть ли деньги на текущем счете, чтобы не начинать операцию, которая все равно не будет успешно завершена.
Но в определенных ситуациях необходимо выполнить тело цикла хотя бы один раз, независимо от каких-либо условий. Например, такое возможно, когда от пользователя ожидается ответная реакция на поставленный ему вопрос. Ответ пользователя может как удовлетворять формальным требованиям, так и не удовлетворять. В последнем случае вопрос необходимо повторить, поскольку пользователь мог либо не понять сути вопроса, либо случайно задеть соседнюю клавишу, в результате чего ответ оказался лишенным смысла. На рис. 6.5 показаны два сеанса программы, предлагающей удалить файл. В первую очередь, программа должна спросить пользователя, не возражает ли он. Пользователь должен ввести у или n, в результате чего программа удаляет файл или отменяет данную операцию. Но если пользователь введет любой другой символ, то программа должна спросить еще раз, и так до тех пор, пока пользователь не введет у или n.

Обратите внимание на то, что первый проход цикла должен быть выполнен в любом случае независимо от каких-либо условий. Пока пользователь не ввел что-либо, проверять нечего. Цикл начинает выполняться не с проверки условия, а с первой инструкции тела цикла. Сначала необходимо предоставить пользователю возможность осуществить ввод и только после этого проверять соблюдение условия.
Именно поэтому в данной программе следует использовать цикл do (который также называют циклом do…while). Программа сначала выполняет тело цикла и только после этого проверяет условие. Если условие равно true, программа опять выполняет тело цикла. Если же условие равно false, программа выходит из цикла и выполняет первую инструкцию, расположенную непосредственно за ним.

# Чтение одиночного символа с клавиатуры
В листинге 5.3 пользователь вводил с клавиатуры слово. Метод keyboard. next () читал введенное слово и помещал его в переменную password типа String. Все работало безукоризненно, потому что переменная типа String может хранить много символов одновременно, а метод next () считывает группу символов. Однако в листинге 6.4 не требуется чтение строки символов. Пользователь должен ввести одну букву: у или п. Поэтому программа создает не переменную типа String, в которой можно хранить много символов, а переменную типа char, хранящую всего один символ. В Java API нет метода nextChar (), который считывал бы с клавиатуры один символ, поэтому мне пришлось импровизировать. Ниже приведен результат импровизации:

    keyboard.findWithinHorizon(".",0).charAt(0);
    
Всякий раз, когда вам понадобится прочитать одиночный символ, можете использовать этот код в том виде, в каком он приведен.
- запомни - Переменная типа String может содержать произвольное количество символов, в том числе только один. Однако переменная типа String, содержащая один символ, и переменная типа char, которая также содержит один символ, - это совершенно разные вещи. Что бы ни находилось в переменной типа String, способы обработки переменных этого типа отличаются отспособов обработки переменных типа char.

Переменную reply нельзя упоминать за пределами фигурных скобок даже в операторе while. Если вы попытаетесь скомпилировать этот код, то получите три сообщения об ошибке -- два сообщения о наличии переменной reply в операторе while и одно сообщение о ее наличии в операторе if. Поэтому в листинге 6.4 ваши руки связаны. Впервые переменная reply pe- ально используется в цикле. Но для того чтобы ее можно было использовать и после выхода из цикла, она должна быть объявлена до цикла. Наилучший выход в этой ситуации состоит в том, чтобы объявить ее, не инициализируя.
