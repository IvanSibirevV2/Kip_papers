Рис. 5.4. Использование оператора == и метода equals ()

Вызов инструкции keyboard.next () приводит к захвату произвольного слова, которое вводит пользователь на клавиатуре компьютера. Затем код присваивает это слово переменной password. Потом инструкция if с помощью двух разных методик сравнивает значение переменной password со значением "swordfish". В листинге 5.3 применены оба способа сравнения двух строк. Правильный способ - тот, в котором используется метод equals (). Вызов метода equals () выглядит несколько необычно: мы записываем метод справа от одной строки, поместив между ними точку, тогда как вторая строка указывается в круглых скобках в качестве аргумента метода. Именно такой синтаксис и следует использовать. При вызове метода equals не имеет значения, в какой строке находится точка, а в какой - скобки. Например, в листинге 5.3 можно написать

    if ("swordfish".equals (password))
Этот метод также хорошо справляется со своей задачей.

- технические подробности - Может создаться впечатление, что при вызове метода equals () строки играют неравнозначные роли, но это не так. Кажущийся дисбаланс между строками, указываемыми после точки и в скобках, имеет свое объяснение. Дело в том, что в действительности вы имеете два равноправных объекта: объект password и объект "swordfish". Оба они имеют тип String (при этом password - переменная типа String, а "swordfish" -- литерал типа String). Записывая выражение password.equals ("swordfish"), вы вызываете метод equals() для объекта password с передачей объекта "swordfish" в качестве аргумента.

# Импортируем все за один раз
Конечно, "втиснуть" все библиотеки Java в одно объявление import не удастся, но писать это объявление для каждого класса все же необязательно. В листинге 5.3 продемонстрирован способ импорта "для ленивых": входной поток System.in и выходной поток System.out (а заодно и все члены класса System, но они нам пока что не нужны) импортируются одной инструкцией. Чтобы за один раз импортировать все содержимое класса System, нужно добавить в инструкцию импорта точку и групповой символ $*$, означающий "все, что там есть". Таким образом, инструкция импорта java.lang.System. $*$ эквивалентна приблизительно тридцати отдельным инструкциям import, включая System.in, System.out, System. err и т.п. Использование звездочки в объявлении import обычно считается плохой практикой, поэтому в примерах данной книги она встречается редко. Но в некоторых программах, когда в одном файле члены какого-либо класса встречаются десятки раз, звездочка может быть довольно удобной.

- технические подробности - Символ звездочки нельзя поставить в любом месте объявления импорта. Например, нельзя импортировать все библиотеки Java, написав import java. *. Звездочку можно поставить только вместо имени класса или статических членов класса. Использование звездочки в объявлении import более подробно рассматривается в главе 9. Статические члены класса будут рассмотрены в главе 10.
