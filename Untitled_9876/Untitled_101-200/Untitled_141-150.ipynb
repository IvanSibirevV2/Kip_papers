# Импортируем все за один раз
Конечно, "втиснуть" все библиотеки Java в одно объявление import не удастся, но писать это объявление для каждого класса все же необязательно. В листинге 5.3 продемонстрирован способ импорта "для ленивых": входной поток System.in и выходной поток System.out (а заодно и все члены класса System, но они нам пока что не нужны) импортируются одной инструкцией. Чтобы за один раз импортировать все содержимое класса System, нужно добавить в инструкцию импорта точку и групповой символ $*$, означающий "все, что там есть". Таким образом, инструкция импорта java.lang.System. $*$ эквивалентна приблизительно тридцати отдельным инструкциям import, включая System.in, System.out, System. err и т.п. Использование звездочки в объявлении import обычно считается плохой практикой, поэтому в примерах данной книги она встречается редко. Но в некоторых программах, когда в одном файле члены какого-либо класса встречаются десятки раз, звездочка может быть довольно удобной.

- технические подробности - Символ звездочки нельзя поставить в любом месте объявления импорта. Например, нельзя импортировать все библиотеки Java, написав import java. *. Звездочку можно поставить только вместо имени класса или статических членов класса. Использование звездочки в объявлении import более подробно рассматривается в главе 9. # Условия в скобках и скобки в условиях
Скобки часто помогают визуально выделить в коде некоторую сущность, но для этого нужно уметь ими пользоваться. Если в условии несколько логических операторов объединены таким образом, что скобки необязательны, то лучше все же их поставить, чтобы выражение было понятным с первого взгляда. Рассмотрим следующее выражение: 2 < 5 || 100 < 6 && 27 < 1

Может показаться, что оно равно false. Визуально оно читается как (некоторое выражение) && 27 < 1. Поскольку 27 всегда больше единицы, можно сделать вывод, что все выражение ложное. Однако это не так, потому что в Java оператор && имеет более высокий приоритет, чем | |. Следовательно, оператор && вычисляется в первую очередь. Это означает, что приведенное выше выражение нужно читать так: 2 < 5 | | (некоторое выражение). Поскольку два всегда меньше пяти, все выражение всегда равно true.
Чтобы изменить значение всего выражения с true на false, нужно заключить в скобки первые два оператора сравнения: (2 < 5 || 100 < 6) && 27 < 1

- совет - В Java оператор | | реализует включающую дизъюнкцию. Это означает, что результат равен true в следующих трех случаях: когда левый операнд равен true, правый операнд равен true или оба операнда равны true. Например, выражение 2 < 10 | | 20 < 30 равно true.
- внимание - В Java нельзя объединять операторы сравнения, как в естественном языке. Например, по-русски можно сказать: "За столом будут сидеть от 3 до 10 человек". Если же написать 3 <= n <= 10, то компилятор сгенерирует сообщение об ошибке. Чтобы выполнить указанную операцию, нужно написать 3 <= n && n <= 10.
