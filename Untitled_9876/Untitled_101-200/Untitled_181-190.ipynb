# Открытые классы
В листинге 7.1 первая строка кода имеет следующий вид: public class Account {

Класс называется открытым (public), если он объявлен с использованием ключевого слова public. Открытый класс может использоваться любым другим классом. Например, если вы пишете программу ATMController, работающую в отдаленном уголке киберпространства, она может содержать ссылку на класс Account. В ней можно написать инструкцию myAccount.balance = 24.02. (Kо- нечно, код программы ATMController должен знать, в какой точке киберпространства находится определение класса Account, но это уже другая история.) В листинге 7.2 имеется строка myAccount.balance = 24.02. Глядя на нее, вы могли бы сказать себе: "Класс Account должен быть открытым, поскольку другой класс (код в листинге 7.2) использует его". К сожалению, реалии использования открытых классов несколько сложнее, чем кажется на первый взгляд. В действительности при благоприятном расположении звезд на небе один класс может использовать другой и в тех случаях, когда этот другой класс объявлен без использования ключевого слова public. (О благоприятном расположении звезд мы поговорим в главе 14.)

Правда жизни такова, что объявление некоторых классов как public в примерах этой главы, просто позволяет мне чувствовать себя великодушным. Именно так: некоторые вещи программисты делают исключительно из великодушия. В листинге 7.1 мое великодушие проявляется в том, что классом Account могут воспользоваться многие программисты. Когда я создаю класс, который определяет некую полезную сущность с легко распознавае- мым именем, таким как Account, Engine, Customer, BrainWave, HeadAche или SevenLayerCake, я всегда объявляю его как public. Класс UseAccount в листинге 7.2 также объявлен как public. Если класс содержит метод main, Java-программисты объявляют его как public, не задумываясь особо о том, будет ли кто-нибудь его использовать. Поэтому, даже если никакой другой класс не будет задействовать мой метод main, я все равно объявляю класс UseAccount как открытый. Большинство классов в этой книге содержит метод main, и, соответственно, большинство из них являются классами public.

- внимание - Класс, объявляемый как public, должен находиться в файле, имя которого в точности совпадает с именем класса (но дополнительно снабжается расширением . java). Например, если вы объявляете класс public class MyImportantCode, код класса следует сохранить в файле MyImportantCode.java, причем буквы M, I и в имени файла должны быть прописными, а остальные - строчными. Это соглашение о правилах именования файлов имеет важное следствие: если ваш код объявляет два открытых класса, то он должен храниться по крайней мере в двух . java-файлах. Иными словами, объявление двух классов как public в одном и том же . java-файле не допускается.
