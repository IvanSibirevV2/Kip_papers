Стратегия, примененная в листинге 13.8, сама по себе правильная. Программа вызывает метод sleep класса Thread, входящего в Java API. Число, передаваемое методу, -- это длительность паузы в миллисекундах. Выражение Thread. sleep (5000) означает, что выполнение текущего потока будет приостановлено на пять секунд. Проблема состоит в том, что метод sleep может сгенерировать исключение InterruptedException (исключение прерывания). Поэтому при попытке скомпилировать этот код компилятор вернет сообщение, подобное следующему.

Unreported exception java.lang.InterruptedException; must be caught or declared to be thrown (Незарегистрированное исключение java. lang. InterruptedException; должно быть перехвачено или объявлено сгенерированным) Либо сообщение может иметь следующий вид: Unhandled exception type InterruptedException (Hеобработанный тип исключения InterruptedException)

Как бы там ни было, подобные сообщения нежелательны.

- технические подробности - Чтобы понять общий принцип работы исключений, вовсе не обязательно вникать в суть исключения InterruptedException. Все, что вам действительно нужно знать, - это то, что вызов Thread. sleep может вызвать одно из исключений типа InterruptedException. Подобное исключение генерируется, когда какой-то другой код прерывает паузу данного кода. В Java несколько потоков кода могут выполняться одновременно. Поэтому может случиться так, что один поток вызывает метод Thread.sleep, но в то же самое время другой поток вызывает метод interrupt. Второй поток хочет временно приостановить первый, чтобы выполнить какую-то срочную работу, но первый поток и без того уже в режиме паузы. На создавшуюся ситуацию метод Thread. sleep отреагирует генерированием исключения InterruptedException.
