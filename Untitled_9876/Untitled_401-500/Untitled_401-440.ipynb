В языке Java есть исключения двух видов: проверяемые и непроверяемые.
- Проверяемые исключения должны быть перехвачены и обработаны в коде, иначе программа будет аварийно завершена.
- Непроверяемые исключения можно не перехватывать и не обрабатывать. Но в случае необходимости такое исключение может быть обработано.

Исключение InterruptedException относится к категории проверяемых. Когда код вызывает метод, который может сгенерировать проверяемое исключение, нужно в этом же коде запрограммировать его перехват и обработку, иначе код не будет компилироваться.
Таким образом, если код метода может сгенерировать проверяемое исключение, необходимо запрограммировать реакцию на него одним из двух способов.

- Заключить место, в котором может произойти исключение, в блок try. Добавить блок catch (возможно, пустой), перехватывающий исключение данного типа.
- Добавить в заголовок метода директиву throws и тип исключения (можно добавить базовый тип Exception, обозначающий любое исключение).

Проиллюстрируем эти два пункта в следующих двух листингах. В листинге 13.9 метод sleep, который может сгенерировать исключение, находится в блоке try. Непосредственно после него расположен блок catch, в списке параметров которого указан тип InterruptedException.

Одно из решений (как вы дальше увидите - плохое) состоит в том, чтобы поместить еще один блок try в блок catch. Это предохранит программу от неожиданных инцидентов во время выполнения блока catch. Другое, лучшее решение проблемы состоит в создании блока finally. Как и блок catch, он располагается после блока try. Его существенное отличие от блоков try и catch состоит в том, что его инструкции всегда выполняются независимо от того, возникло ли исключение и было ли оно обработано. Оптимальное решение данной проблемы появилось в версии Java 7: ресурсы можно включать в заголовок блока try (листинг 13.13).

# Модификаторы доступа к членам классов
Если вы дошли до этого места в книге, значит, уже знаете, что в объектно- ориентированном программировании много внимания уделяется сокрытию одних данных от других. Программист, пишущий свою часть кода, не должен менять фрагменты кода, написанные другими программистами. Это обусловлено не требованиями безопасности или секретности, а принципом модульности. Скрывая подробности, вы фактически скрываете сложность своего кода и делаете его простым для использования. Ваш код выглядит как черный ящик с простыми входами и выходами. Вы упрощаете задачу людям. Вы облегчаете их жизнь и предохраняете от ошибок, экономите их деньги, бережете их время и способствуете продлению их жизни. Впрочем, вы и себе облегчаете этим жизнь, потомучто уменьшаете вероятность конфликтов имен в своем коде.В предыдущих главах на каждом шагу встречались примеры использования закрытых полей. Если поле объявлено с модификатором private, оно невидимо за пределами класса и недоступно для изменения. Сокрытие полей и методов улучшает модульность и упрощает код.
Но не все поля в примерах книги закрытые. Есть также немало открытых полей, т.е. объявленных с модификатором public. Как мировая знаменитость, личная жизнь которой выставлена напоказ, открытое поле доступно всем и каждому, т.е. к нему можно обращаться в любом месте программы. Многие знают, какую зубную пасту использовал Элвис Пресли. Аналогично этому любой программист может сослаться в программе на открытую переменную. Разница лишь в том, что открытую переменную можно как читать, так и изменять, а название зубной пасты Элвиса любой его поклонник может только узнать, изменить его он не может. Ключевые слова public и private называются модификаторами доступа. В примерах книги вам часто встречались поля и методы, объявленные без модификаторов доступа. # ПОИСК ФАЙЛА ВО ВСЕХ ВОЗМОЖНЫХ МЕСТАХ
Вы пытаетесь компилировать программу, представленную в листинге 14.1. Компилятор просматривает код и отмечает отсутствующие части программы. В первую очередь он видит, что нужно найти класс ArtFrame. Для решения этой проблемы в виртуальной машине Java (JVM) определена переменная среды CLASSPATH. В этой переменной хранится список мест, в которых компилятор и JVM должны искать нужный код. Как бы то ни было, JVM нуждается в списке мест, в которых она может искать код. Не располагая никакой информацией, ни компилятор, ни виртуальная машина никакого поиска выполнять не будут. В таком случае они не смогут найти классы Art Frame или Drawing, а вы получите либо сообщение cannot find symbol, либо ошибку NoClassDefFoundError и будете чувствовать себя очень несчастным.

# Защищенный доступ
Когда я познакомился с Java, я думал, что ключевое слово protected (защищенный) означает "усиленный", "безопасный" или что-то подобное. В Java защищенный член класса менее скрытый и менее безопасный, чем имеющий доступ, применяемый по умолчанию. Поэтому пусть вас не вводит в заблуждение слово "защищенный".
