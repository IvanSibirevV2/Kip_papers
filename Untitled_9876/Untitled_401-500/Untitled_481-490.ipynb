# Подключение и отключение базы данных
Кроме вызова метода executeUpdate, код листинга 17.1 выполняет ряд вспомогательных операций. Ниже приведено описание каждого фрагмента кода.

- DriverManager.getConnection. Подключение к конкретной базе данных. Метод getConnection принадлежит классу DriverManager. В листинге 17.1 этот метод создает экземпляр базы данных Account Database и устанавливает соединение с ней. Может случиться, что при вызове метода getConnection база данных Account Database уже существует. Выражение create=true в строке CONNECTION означает, что в таком случае вызов метода getConnection ни на что не влияет. В листинге 17.1 обратите внимание на двоеточия в строке CONNECTION. Код не только присваивает базе данных имя Account Database, но и сообщает классу DriverManager, какой протокол необходимо использовать для подключения к базе данных. Запись jdbc: derby: играет ту же роль, что и префикс http: в веб-адресе: он указывает компьютеру использовать протокол jdbc для коммуникации по протоколу derby, который обеспечивает подключение к базе данных Account Database.

- conn.createStatement. Создание инструкции SQL. B JDBC объект statement, представляющий инструкцию SQL, создается один раз. После этого его можно многократно использовать с различными SQL-строками для выполнения множества различных операций с базой данных с помощью метода statement. executeUpdate. Объект statement создается путем вызова метода conn.createStatement.
- Блоки try с ресурсами. Освобождение ресурсов при возникновении исключений. Как говорит моя супруга, нельзя предъявлять претензии к другим, если сам оставляешь после себя беспорядок. Каждое соединение с базой данных блокирует ряд ресурсов компьютера. Когда вы закончите пользоваться этими ресурсами, необходимо освободить их. Для решения этой проблемы применяется инструкция try c ресурсами, которая автоматически закрывает и освобождает ресурсы, когда необходимость в них отпадает. Кроме того, эта инструкция заботится о разрешении всех проблем, связанных с неудачными попытками корректного перехвата исключений.
