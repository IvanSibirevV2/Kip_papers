# Правильное использование регистра букв

Не ошибается лишь тот, кто ничего не делает". Это сказал один из моих профессоров в колледже. Я, правда, не помню его имени, поэтому не могу воздать ему должное в полной мере. Полагаю, это моя ошибка.

Язык Java чувствителен к регистру букв, поэтому при выборе имен переменных, классов и членов классов нужно соблюдать стиль именования, привычный для других программистов. Помните о следующих нюансах при написании программ.

- Ключевые слова Java всегда пишутся только в нижнем регистре. Например, if нельзя написать как If или IF.
- Если в коде используются имена из Java API (Application Programming Interface - интерфейс прикладного программирования), регистры букв в именах должны совпадать с теми, которые используются в API.
- Проследите за тем, чтобы каждое созданное вами имя использовалось во всех частях программы в одном и том же написании. Например, если переменная объявлена как myAccount, нельзя обратиться к ней в другом месте как к myaccount. Если регистр хотя бы одной буквы изменится, компилятор будет считать их двумя разными переменными.

# Проход через блок switch
Если выполнение блока case не завершается инструкцией break, компьютер будет выполнять все последующие наборы инструкций. Например, если значение переменной verse равно 3, следующий код напечатает все три строки: потому что он все равно уже мокрый, потому что он очень расстроен и потому что дождь идет очень долго.

    switch (verse) 
    {
        case 3:
            out.print("потому что он все равно уже мокрый, ");
            out.println("потому что он все равно уже мокрый,");
        case 2:
            out.print("потому что он очень pасстроен, ");
            out.println("потому что он очень расстроен,");
        case 1:
            out.print("потому что дождь идет очень долго, ");
            out.println("потому что дождь идет очень долго,");
    }

# Сравнение двух значений
Не забывайте, что для сравнения двух значений нужно использовать двойной знак равенства, а не одинарный. Например, эта строка записана правильно: $if (inputNumber == randomNumber)$

А следующая строка - неправильно: $if (inputNumber = randomNumber)$

# Определение обязательных конструкторов
В класс можно добавить конструктор с параметрами:

    public Temperature (double number)
Однако не следует забывать, что в таком случае компилятор не создаст конструктор без параметров, определенный по умолчанию. Следовательно, создать объект с помощью конструктора без параметров, как показано ниже, вы не сможете:

    Temperature roomTemp = new Temperature();
Это можно будет сделать, только если вы определите собственный конструктор без параметров.

# Соблюдение границ массива
Не забывайте, что при объявлении массива с десятью элементами они нумеруются от 0 до 9. Предположим, массив guests создается следующим образом:

    int guests[] = new int[10];
Тогда на элементы массива можно ссылаться так: guests $[$0$]$, guests $[$1$]$… guests $[$9$]$. На этом массив заканчивается. Если написать guests $[$10$]$, произойдет ошибка, потому что элемента номер 10 не существует.
