# Реализация интерфейсов
Листинг 15.3 реализует интерфейсы Displayable и Summarizable, а также предоставляет тела для методов display и summarize.

    Листинг 15.3. Реализация двух интерфейсов
    public class ColumnOfNumbers implements Displayable, Summarizable 
    {
        double numbers[];
        public ColumnOfNumbers (double[] numbers) { this.numbers = numbers; }
        @Override
        public void display() { for (double d : numbers) { System.out.println (d);}}
        @Override
        public String summarize() {
            double total = 0.0;
            for (double d : numbers){ total += d; }
            return Double.toString(total) ;
        }
    }

- запомни - Реализуя интерфейс, вы предоставляете тела для его абстрактных методов.

Компилятор Java серьезно относится к использованию ключевого слова implements. Если вы удалите одно из двух объявлений метода из листинга 15.3, оставив оператор implements, в окне вашего редактора IDE появятся пугающие сообщения об ошибках. Java ожидает, что вы будете соблюдать контракт, как подразумевает директива implements. Если вы не выполняете условия контракта, Java отказывается компилировать ваш код.
- совет - Вы можете воспользоваться сообщениями об ошибках Java в своих интересах. Начните с ввода некоторого кода, содержащего инструкцию implements Displayable, Summarizable. Встретив ключевое implements, редактор кода отобразит метку ошибки и перечислит названия методов, которые вы должны были объявить, но не сделали этого. В примере из данного раздела используются названия методов display и summarize. После нескольких дополнительных щелчков мышью IDE сгенерирует для вас простые методы display и summarize.

        public Table(String fileName)

            while (diskFile.hasNextLine()) { lines.add(diskFile.nextLine());}
        }
        @Override
        public void display() { for (String line : lines) { System.out.println(line); } }
        
        @Override
        public String summarize() { return lines.get (0);}
