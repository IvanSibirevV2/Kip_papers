В языке Java есть исключения двух видов: проверяемые и непроверяемые.
- Проверяемые исключения должны быть перехвачены и обработаны в коде, иначе программа будет аварийно завершена.
- Непроверяемые исключения можно не перехватывать и не обрабатывать. Но в случае необходимости такое исключение может быть обработано.

Исключение InterruptedException относится к категории проверяемых. Когда код вызывает метод, который может сгенерировать проверяемое исключение, нужно в этом же коде запрограммировать его перехват и обработку, иначе код не будет компилироваться.
Таким образом, если код метода может сгенерировать проверяемое исключение, необходимо запрограммировать реакцию на него одним из двух способов.

- Заключить место, в котором может произойти исключение, в блок try. Добавить блок catch (возможно, пустой), перехватывающий исключение данного типа.
- Добавить в заголовок метода директиву throws и тип исключения (можно добавить базовый тип Exception, обозначающий любое исключение).

Проиллюстрируем эти два пункта в следующих двух листингах. В листинге 13.9 метод sleep, который может сгенерировать исключение, находится в блоке try. Непосредственно после него расположен блок catch, в списке параметров которого указан тип InterruptedException.

Этот инцидент поможет нам понять, как обрабатывается исключение. Что произойдет, если во время обработки одного исключения возникнет другое? Ведь инструкции в блоке catch также не застрахованы от исключений.
Одно из решений (как вы дальше увидите - плохое) состоит в том, чтобы поместить еще один блок try в блок catch. Это предохранит программу от неожиданных инцидентов во время выполнения блока catch. Однако для дополнительного блока try нужен свой блок catch. Таким образом, проблема не решается, а отодвигается, причем уровня вложенности, на котором она будет решена окончательно, не существует.
Другое, лучшее решение проблемы состоит в создании блока finally. Как и блок catch, он располагается после блока try. Его существенное отличие от блоков try и catch состоит в том, что его инструкции всегда выполняются независимо от того, возникло ли исключение и было ли оно обработано. Данная концепция иллюстрируется в листинге 13.11.

# Использование инструкции try при работе с ресурсами
Предположим, есть программа, которая получает входные данные из двух различных файлов либо из потока Scanner и файла на диске. Чтобы гарантировать своевременное освобождение ресурсов, необходимо поместить вызов метода close в блок finally (листинг 13.12).

    Листинг 13.12. Использование двух файлов
    import java.io.File;
    import java.io.IOException;
    import java.utiil.Scanner;
    public class Main
    {
        public static void main(String args[])
        {
            Scanner scanl = null;
            Scanner scan2 = null;
            try {
                scanl = new Scanner(new File("Filel.txt"));
                scan2 = new Scanner(new File("File2.txt")); // Выполнить что-то полезное
            } catch (IOException e) { // Ой!
            } finally {
            scanl.close();
            scan2.close ();
            System.out.println("Готово!");
        }
    }

Tеоретически компьютер должен выполнить вызовы scanl.close() и scan2.close () в любом случае, независимо от того, пойдет ли что-либо не так при выполнении инструкций в блоке try. Однако это в теории. На практике другой программист (не вы, конечно) мог изменить код, поместив операцию освобождения ресурса scan1 в блок try.

    {

        scan1 = null;
    } catch (IOException e) {
    // Ой!
    } finally {
        scan1.close ();
        scan2.close();
        System.out.println("Готово!");

Вот теперь у вас сложилась действительно затруднительная ситуация. В блоке finally значение scanl равно null. Вызов scanl.close () не может быть выполнен, поэтому программа генерирует исключение NullPointerException (нулевой указатель), и ее выполнение прекращается еще до того, как выпол- нится вызов scan2.close (). При наихудшем стечении обстоятельств ресурс scan2 не освобождается, и файл File2.txt блокируется вашей программой, не давая другим программам использовать его.
Когда в программе используется множество ресурсов (файлы, базы данных, сетевые соединения и т.п.), структура блоков try и директив throws может стать слишком сложной, потому что придется также включать блоки try в блоки catch и применять другие запутанные комбинации. Оптимальное решение данной проблемы появилось в версии Java 7: ресурсы можно включать в заголовок блока try (листинг 13.13).

# Модификаторы доступа к членам классов
Если вы дошли до этого места в книге, значит, уже знаете, что в объектно- ориентированном программировании много внимания уделяется сокрытию одних данных от других. Программист, пишущий свою часть кода, не должен менять фрагменты кода, написанные другими программистами. Это обусловлено не требованиями безопасности или секретности, а принципом модульности. Скрывая подробности, вы фактически скрываете сложность своего кода и делаете его простым для использования. Ваш код выглядит как черный ящик с простыми входами и выходами. Вы упрощаете задачу людям. Вы облегчаете их жизнь и предохраняете от ошибок, экономите их деньги, бережете их время и способствуете продлению их жизни. Впрочем, вы и себе облегчаете этим жизнь, потомучто уменьшаете вероятность конфликтов имен в своем коде.В предыдущих главах на каждом шагу встречались примеры использования закрытых полей. Если поле объявлено с модификатором private, оно невидимо за пределами класса и недоступно для изменения. Сокрытие полей и методов улучшает модульность и упрощает код.
Но не все поля в примерах книги закрытые. Есть также немало открытых полей, т.е. объявленных с модификатором public. Как мировая знаменитость, личная жизнь которой выставлена напоказ, открытое поле доступно всем и каждому, т.е. к нему можно обращаться в любом месте программы. Многие знают, какую зубную пасту использовал Элвис Пресли. Аналогично этому любой программист может сослаться в программе на открытую переменную. Разница лишь в том, что открытую переменную можно как читать, так и изменять, а название зубной пасты Элвиса любой его поклонник может только узнать, изменить его он не может. Ключевые слова public и private называются модификаторами доступа. В примерах книги вам часто встречались поля и методы, объявленные без модификаторов доступа. Такие поля и методы имеют доступ, установленный по умолчанию. В большинстве глав не имеет значения, что такое доступ, установленный по умолчанию, но только не в данной главе. Здесь мы рассмотрим этот вопрос подробнее. В главе рассматривается также еще один модификатор доступа, который не использовался в предыдущих главах. Это модификатор protected, определяющий защищенный доступ.

# Правила доступа к членам класса
Этот раздел посвящен исключительно членам класса. Но это не значит, что классы в нем игнорируются. В Java все делается только в классах. Важный вопрос, которому посвящена вся данная глава: можно ли использовать имя некоторого члена класса в другом месте кода? Для ответа на него в первую очередь посмотрите, где находится это место: внутри или вне класса, в котором определен данный член. Если член класса объявлен как закрытый (с модификатором private),то обратиться к нему можно только в данном классе.

    class SomeClass { private int myField = 10; }
    class SomeOtherClass
    {
        public static void main(String args[])
        {
            SomeClass someObject = new SomeClass();
            // Этот код не работает
            System.out.println(someObject.myField);
        }
    }

Если член класса объявлен как открытый (с модификатором public), то к нему можно обратиться по имени в любой точке программы.

    class SomeClass
    {
        public int myField = 10;
        class SomeOtherClass
        {
            public static void main(String args[])
            {
                SomeClass someObject = new SomeClass(); // Этот код работает
                System.out.println (someObject.myField);
            }
        }
    }

# ПОИСК ФАЙЛА ВО ВСЕХ ВОЗМОЖНЫХ МЕСТАХ
Вы пытаетесь компилировать программу, представленную в листинге 14.1. Компилятор просматривает код и отмечает отсутствующие части программы. В первую очередь он видит, что нужно найти класс ArtFrame. Откуда компилятор получает информацию о местоположении классов ArtFrame и Drawing?
Должен ли компилятор просмотреть весь ваш жесткий диск в поисках этих файлов? Диск может иметь объем в сотни гигабайтов и даже терабайтов. Нужно ли искать файлы на сетевых дисках? Потенциально пространство поиска может быть неограниченным.
Для решения этой проблемы в виртуальной машине Java (JVM) определена переменная среды CLASSPATH. В этой переменной хранится список мест, в которых компилятор и JVM должны искать нужный код. Существует несколько способов установки значения переменной CLASSPATH. Одни программисты создают новую переменную CLASSPATH всякий раз, когда выполняют програм- муна Java. Другие создают общесистемную переменную CLASSPATн. (Те из вас, кто знаком с переменной РАТн, которая используется на компьютерах с операционной системой Windows или Linux, должны знать, как все это работает.) Как бы то ни было, JVM нуждается в списке мест, в которых она может искать код. Не располагая никакой информацией, ни компилятор, ни виртуальная машина никакого поиска выполнять не будут. В таком случае они не смогут найти классы Art Frame или Drawing, а вы получите либо сообщение cannot find symbol, либо ошибку NoClassDefFoundError и будете чувствовать себя очень несчастным.

Не забывайте, что при импорте родительского пакета вложенные пакеты не импортируются. # Защищенный доступ
Когда я познакомился с Java, я думал, что ключевое слово protected (защищенный) означает "усиленный", "безопасный" или что-то подобное. Раз некоторая сущность является защищенной, значит, к ней тяжело подобраться и ее тяжело повредить. Но оказалось, что это не совсем так. В Java защищенный член класса менее скрытый и менее безопасный, чем имеющий доступ, применяемый по умолчанию. Поэтому пусть вас не вводит в заблуждение слово "защищенный".

# Реализация интерфейсов
Листинг 15.3 реализует интерфейсы Displayable и Summarizable, а также предоставляет тела для методов display и summarize.

    Листинг 15.3. Реализация двух интерфейсов
    public class ColumnOfNumbers implements Displayable, Summarizable 
    {
        double numbers[];
        public ColumnOfNumbers (double[] numbers) { this.numbers = numbers; }
        @Override
        public void display() { for (double d : numbers) { System.out.println (d);}}
        @Override
        public String summarize() {
            double total = 0.0;
            for (double d : numbers){ total += d; }
            return Double.toString(total) ;
        }
    }

- запомни - Реализуя интерфейс, вы предоставляете тела для его абстрактных методов.

Компилятор Java серьезно относится к использованию ключевого слова implements. Если вы удалите одно из двух объявлений метода из листинга 15.3, оставив оператор implements, в окне вашего редактора IDE появятся пугающие сообщения об ошибках. Java ожидает, что вы будете соблюдать контракт, как подразумевает директива implements. Если вы не выполняете условия контракта, Java отказывается компилировать ваш код.

Встретив ключевое implements, редактор кода отобразит метку ошибки и перечислит названия методов, которые вы должны были объявить, но не сделали этого. В примере из данного раздела используются названия методов display и summarize. После нескольких дополнительных щелчков мышью IDE сгенерирует для вас простые методы display и summarize.

# ДВЕ РАЗНОВИДНОСТИ МЕТОДОВ
Любой метод, не имеющий тела и находящийся в объявлении интерфейса, называется абстрактным. Начиная с версии Java 8 в область объявления интерфейса можно также включать методы, имеющие тела. Метод, обладающий телом, называется методом по умолчанию. В коде интерфейса объявление каждого метода по умолчанию начинается ключевым словом default.

    public interface MyInterface
    {
        void methodl ();
        default void method2() { System.out.print1n("Привет!"); }
    }
В объявлении MyInterface методmethod1 является абстрактным, a method2 - методом по умолчанию. Предположим, вы создаете класс, который реализует интерфейс MyInterface, например:
class MyClass implements MyInterface
В только что объявленном классе MyClass нужно объявить его собственный метод methodl и обеспечить тело для него. При желании ваш класс MyClass может также объявить собственный метод method2. Если этого не сделано, класс наследует тело метода method2 из интерфейса MyInterface.

В листинге 16.1 главная новинка для вас -- обработка щелчка на кнопке. - Объявление implements ActionListener в заголовке класса Game Frame означает, что класс реализует интерфейс ActionListener, который обеспечивает прослушивание события. # Потоки выполнения
Программа на Java может быть многопоточной. Это означает, что она может одновременно делать много разных вещей. Конечно, компьютер выполняет не только написанный вами код программы, но и другой код, который вы чаще всего даже не видите и никогда не увидите. Операционная система автоматически организует одновременное выполнение множества программ. Но чтобы в одной программе одновременно решалось несколько задач, нужно запустить их в разных потоках. Когда компьютер выполняет метод main, он может проверить, нет ли еще потоков, которые можно запустить, чтобы не терять времени, пока выполняется метод main. Для большинства простых программ Java другие потоки являются частью Java Virtual Machine (JVM). В частности, виртуальная машина Java автоматически создает поток обработки событий. Когда выполняется код программы, одновременно (в фоновом режиме) выполняется код прослушивания событий, который распознает щелчки мышью и запускает обработчик события (рис. 16.3).

Поток обработки событий видит, что произошел щелчок, и думает: "Что я должен сделать в связи с этим?" Конечно, в коде он быстро находит ответ: "Вызвать метод actionPerformed". Поток обработки событий выполняет код, похожий на приведенный ниже.

    if(выполнен щелчок на кнопке)
    {
        object1.actionPerformed (информация о щелчке);
        object2.actionPerformed(информация о щелчке);
        object3.actionPerformed (информация о щелчке)
    }

Поток обработки событий должен найти ответ еще на один вопрос: "Где находится нужный метод actionPerformed?" Кроме того, поток должен знать, на какие события следует реагировать, а какие должны быть проигнорированы. Что если вы не хотите, чтобы поток, обрабатывающий события, вызывал некоторые из методов actionPerformed, притаившихся в вашем коде? Вопрос решается просто: поток отреагирует только на те события, прослушивание которых включено в коде программы. Иными словами, слушатель событий должен быть явно зарегистрирован с помощью метода addActionListener. В листинге 16.1 подключение слушателя событий осуществляется с помощью следующей инструкции:

    button.addActionListener(this);
Эта инструкция приказывает потоку обработки событий добавить метод actionPerformed в список методов, вызываемых после щелчка на кнопке. В данном случае получателем событий служит экземпляр фрейма, обозначенный ключевым словом this. В классе фрейма определен метод action Performed, который становится обработчиком события. Чтобы компьютер вызвал метод actionPerformed, его нужно зарегистрировать в потоке обработки событий. Регистрация выполняется путем вызова метода addActionListener, принадлежащего объекту, на котором ожидается щелчок. В листинге 16.1 щелчок ожидается на кнопке button, поэтому метод addActionListener принадлежит объекту кнопки.

# Ключевое слово this
В главах 9 и 10 ключевое слово this обеспечивало доступ к переменным экземпляра класса внутри метода. Что же означает это ключевое слово? Чтобы лучше понять его смысл, рассмотрим пример с фразой ваше имя:
Я, (ваше имя), торжественно клянусь соблюдать конституцию США
Фраза ваше имя является заполнителем, вместо которого каждый вписывает собственное имя:
Я, Боб, торжественно клянусь…
Я, Фред, торжественно клянусь…

А теперь представьте, что вместо фразы ("Я… торжественно клянусь…") подставляется фрагмент кода класса Java. В этом фрагменте кода используется фраза- заполнитель ваше имя. Всякий раз, когда экземпляр класса (человек) выполняет код (дает клятву), экземпляр класса подставляет вместо заполнителя свое имя.
В Java ключевое слово this обозначает текущий экземпляр класса, в котором оно находится. В данном примере (см. листинг 16.1) слово this находится в классе GameFrame, следовательно, оно означает фрейм.

Вызов метода addActionListener(this) находится в конструкторе класса GameFrame, поэтому при создании экземпляра фрейма происходит подключение приемника к событию. В данном случае приемником служит текущий фрейм:

    button.addActionListener(данный экземпляр фрейма);    
Вызвав метод button.addActionListener(this), мы приказываем потоку обработки событий добавить метод actionPerformed в список методов, вызываемых после щелчка на кнопке button. Класс GameFrame должен содержать метод actionPerformed, потому что он объявлен в интерфейсе ActionListener, который реализуется классом GameFrame.

# Тело метода actionPerformed
В методе actionPerformed используется несколько трюков из арсенала Java API. Эти трюки вкратце описаны ниже.
- Каждый экземпляр класса JTextField (как и JLabel) имеет собственный набор методов доступа, включая getText и setText. Вызов getText возвращает строку символов, отображаемую данным элементом интерфейса. Вызов setText записывает заданную строку символов в элемент интерфейса. В листинге 16.1 метод getText извлекает из текстового поля число, а метод setText записывает в текстовое поле либо пустую строку (""), либо сообщение о том, что число угадано.
- Укаждого класса (в том числе JTextField и JButton) пакета javax. swing есть метод setEnabled. Вызов setEnabled ( false) отключа- етэлемент интерфейса, в результате чего он окрашивается в серый цвет и перестает реагировать на щелчки мышью и нажатия клавиш.
- У каждого класса, определенного в пакете javax. swing, есть метод request Focus. При его вызове элемент интерфейса получает фокус ввода. Это означает, что теперь именно этот элемент будет реагировать на нажатие клавиш. В листинге 16.1 инструкция textField. requestFocus передает фокус текстовому полю. После каждого щелчка на кнопке фокус переходит к ней, однако данная инструкция каждый раз возвращает фокус текстовому полю. Метод getSource объекта события возвращает имя объекта, сгенери- ровавшего событие. С его помощью можно проверить, действительно ли button -- элемент интерфейса, на котором щелкнул пользователь. Для этого добавьте в тело метода actionPerformed инструкцию if (e.getSource () == button). Если во фрейме две кнопки (например, button1 и button2), таким способом можно выяснить, на какой кнопке сделан щелчок. Для этого напишите if (e.getSource() == button1) или if (e.getSource() == button2).

# Идентификатор версии
В главе 9 рассматривалось применение аннотации @SuppressWarnings для отмены предупреждения компилятора о том, что не задан идентификатор версии serialVersionUID. Этот идентификатор помогает Java избежать конфликтов версий при передаче объекта из одного места в другое. Например, можно отправить состояние объекта JFrame на экран другого компьютера. Принимающий компьютер может проверить номер версии фрейма во избежание нежелательных эффектов. В главе 9 я обошел проблему идентификации версий с помощью аннотации @SuppressWarnings, которая приказывает компилятору игнорировать отсутствие номера версии. В листинге 16.1 применен более смелый подход: объекту JFrame присвоен номер serialVersionUID. Данный фрейм -- наша первая версия класса GameFrame, поэтому я присвоил версии номер 1. Но поскольку номер версии имеет тип long, литерал имеет вид 1L (см. главу 4). Нужно ли изменять номер версии при каждом изменении кода? Очевидно, нет. В общем случае рекомендуется изменять его, когда в результате изменения кода новая версия становится несовместимой со старой версией других компонентов приложения. Несовместимость означает, что принимающий компьютер или другие классы программы не смогут работать с измененным объектом.

Класс MyActionListener в листинге 16.5 - это внутренний класс. Почти во всем он аналогичен обычному классу. Различие между ними лишь в том, что код внутреннего класса может ссылаться на поля внешнего класса. # Правильное использование регистра букв

Не ошибается лишь тот, кто ничего не делает". Это сказал один из моих профессоров в колледже. Я, правда, не помню его имени, поэтому не могу воздать ему должное в полной мере. Полагаю, это моя ошибка.

Язык Java чувствителен к регистру букв, поэтому при выборе имен переменных, классов и членов классов нужно соблюдать стиль именования, привычный для других программистов. Помните о следующих нюансах при написании программ.

- Ключевые слова Java всегда пишутся только в нижнем регистре. Например, if нельзя написать как If или IF.
- Если в коде используются имена из Java API (Application Programming Interface - интерфейс прикладного программирования), регистры букв в именах должны совпадать с теми, которые используются в API.
- Проследите за тем, чтобы каждое созданное вами имя использовалось во всех частях программы в одном и том же написании. Например, если переменная объявлена как myAccount, нельзя обратиться к ней в другом месте как к myaccount. Если регистр хотя бы одной буквы изменится, компилятор будет считать их двумя разными переменными.

# Указатели на null
Примеры, приведенные в книге, не вызывают исключений типа NullPointer Exception (пустой указатель), поскольку они специально сделаны простыми и неприхотливыми. Однако в реальной жизни исключения этого типа - кошмар любого программиста. Чаще всего исключение NullPointerException возникает в тех случаях, когда метод должен был вернуть объект, но ничего не вернул. Рассмотрим простой пример.

    import static java.lang.System.out;
    import java.io.File;
    class ListMyFiles
    {
        public static void main(String args[])
        {
            File myFile = new File("/Users");
            String dir[] = myFile.list();
            for (String fileName : dir) { out.println (fileName); }
        }
    }
Стр. 497

----------

Предполагается, что программа отобразит на консоли список всех файлов, находящихся в папке Users.
Однако что произойдет, если программа не найдет папку /Users или если в строке приведено неверное имя папки?
File myFile = new File("6*8$!!"); В этом случае оператор new File вернет null (специальное слово Java, которое означает "ничто"). Соответственно, переменной myFile будет присвоено значение null. В следующей инструкции переменная dir не будет создана, и цикл станет невыполнимым. Программа сгенерирует исключение NullPointerException и (если код не перехватывает это исключение) будет аварийно завершена. Чтобы избежать подобных ситуаций, внимательно ознакомьтесь с документацией по используемым библиотечным классам. Если код вызывает метод, который может вернуть null, добавьте блок обработки исключений.

Ошибка NoClassDefFoundError (не найдено определение класса) может возникнуть по многим причинам, но чаще всего она возникает, когда Java не может найти нужный файл с определением класса. Чтобы избежать этой ошибки, придерживайтесь следующих правил:
- папка проекта должна содержать все файлы Java с классами, используемыми в коде;
- если используются именованные пакеты, папка проекта должна содержать соответствующую иерархию вложенных папок;
- правильно задайте переменную среды CLASSPATH в настройках операционной системы.
