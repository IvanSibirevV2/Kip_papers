# ПОИСК ФАЙЛА ВО ВСЕХ ВОЗМОЖНЫХ МЕСТАХ
Вы пытаетесь компилировать программу, представленную в листинге 14.1. Компилятор просматривает код и отмечает отсутствующие части программы. В первую очередь он видит, что нужно найти класс ArtFrame. Когда он найдет его, то увидит, что нужно найти класс Drawing. В листинге 14.1 есть только определение класса ShowFrame, а определений классов ArtFrame и Drawing в нем нет. Откуда компилятор получает информацию о местоположении классов ArtFrame и Drawing?
Должен ли компилятор просмотреть весь ваш жесткий диск в поисках этих файлов? Диск может иметь объем в сотни гигабайтов и даже терабайтов. Нужно ли искать файлы на сетевых дисках? Потенциально пространство поиска может быть неограниченным.
Для решения этой проблемы в виртуальной машине Java (JVM) определена переменная среды CLASSPATH. В этой переменной хранится список мест, в которых компилятор и JVM должны искать нужный код. Существует несколько способов установки значения переменной CLASSPATH. Одни программисты создают новую переменную CLASSPATH всякий раз, когда выполняют програм- муна Java. Другие создают общесистемную переменную CLASSPATн. (Те из вас, кто знаком с переменной РАТн, которая используется на компьютерах с операционной системой Windows или Linux, должны знать, как все это работает.) Как бы то ни было, JVM нуждается в списке мест, в которых она может искать код. Не располагая никакой информацией, ни компилятор, ни виртуальная машина никакого поиска выполнять не будут. В таком случае они не смогут найти классы Art Frame или Drawing, а вы получите либо сообщение cannot find symbol, либо ошибку NoClassDefFoundError и будете чувствовать себя очень несчастным.

# Создание фрейма
В первых трех листингах этой главы представлен пример, состоящий из нескольких частей. В данном разделе рассматривается последняя, третья часть примера. Она не очень важна для демонстрации модификаторов доступа, которые являются главной темой главы. Поэтому, если вас интересуют только модификаторы доступа, пропустите данный раздел и листинг 14.3. Но, с другой стороны, вам будет полезно ознакомиться с классами пакета Swing, предназначенными для создания графических пользовательских интерфейсов.

    Листинг 14.3. Класс ArtFrame
    package com.burdbrain.frames;
    import java.awt.Graphics;
    import javax.swing.JFrame;
    import com. burdbrain.drawings.Drawing;
    public class ArtFrame extends JFrame
    {
        private static final long serialVersionUID = 1L;
        Drawing drawing;
        public ArtFrame (Drawing drawing)
        {
            this.drawing = drawing;
            setTitle ("Произведение абстрактного искусства");
            setDefaultCloseOperation(EXIT ON CLOSE);
        }
        public void paint (Graphics g){ drawing.paint (g) ;}
    }

В листинге 14.3 имеется все необходимое для помещения рисунка во фрейм Java. В коде используется ряд имен, определенных в Java API. Большинство из них рассмотрено в главах 9 и 10.
Единственное новое имя - имя метода paint. В листинге 14.3 этот метод всего лишь вызывает еще один метод paint, принадлежащий объекту типа Drawing. Объект типа ArtFrame создает на экране компьютера фрейм - плавающее окно со строкой заголовка и областью рисования. Объект типа Drawing, передаваемый фрейму через список параметров конструктора, определяет, что именно будет нарисовано во фрейме.
В листингах 14.1-14.3 можно заметить одну особенность: они не содержат ни одного вызова метода paint. На самом деле он все же вызывается, но не явно, кодом программы, а виртуальной машиной Java. Аналогичное поведение свойственно многим компонентам Java, связанным с созданием окон: вам достаточно лишь объявить метод paint, а все остальное будет сделано автоматически в процессе выполнения программы.
Именно это и происходит с объектами javax. swing.JFrame. В листинге 14.3 метод paint фрейма незаметно вызывается "за кулисами" событий. В свою очередь, метод paint фрейма вызывает метод paint объекта Drawing, который и рисует во фрейме окружность. Именно так и была нарисована окружность, которую вы видите на рис. 14.4.

Код не компилируется, потому что поле с доступом, установленным по умолчанию, за пределами своего пакета недоступно даже для производного класса. То же самое справедливо и для метода с доступом, установленным по умолчанию.
Поля и методы класса называются членами класса. Все правила доступа применяются к членам класса единообразно, независимо от того, поля это или методы.
- запомни - Рассматриваемые в данном разделе правила доступа не применяются к локальным переменным, т.е. к переменным, объявленным в методах. К локальной переменной можно обращаться только в методе, в котором она объявлена.
- перекрестная ссылка -Локальные переменные рассматривались в главе 10.

В Java члены класса, имеющие доступ, установленный по умолчанию, доступны только в текущем пакете. Иными словами, если член класса объявлен без модификатора доступа public, private или protected, к нему можно обращаться только в пакете, в котором определен данный класс. Это правило проиллюстрировано на рис. 14.7 и 14.8.
- внимание - Точки в именах пакетов могут ввести в заблуждение. Они как бы намекают на то, что пакеты вложены друг в друга. На самом деле это не совсем так. Вложены только папки в структуре файлов. Не забывайте, что при импорте родительского пакета вложенные пакеты не импортируются. Например, при написании программы, в которой предусмотрено реагирование на щелчок мышью на кнопке, нужны классы, находящиеся в двух разных пакетах. Поэтому нужно записать два объявления импорта: import java.awt.* и import java. awt.event.*. Это необходимо потому, что при импорте пакета java. awt паkет java.awt.event не импортируется.
