# Защищенный доступ
Когда я познакомился с Java, я думал, что ключевое слово protected (защищенный) означает "усиленный", "безопасный" или что-то подобное. Раз некоторая сущность является защищенной, значит, к ней тяжело подобраться и ее тяжело повредить. Но оказалось, что это не совсем так. В Java защищенный член класса менее скрытый и менее безопасный, чем имеющий доступ, применяемый по умолчанию. Поэтому пусть вас не вводит в заблуждение слово "защищенный".

Возможно, прочитав о модификаторах доступа к членам классов, вы почувствовали легкое головокружение, несмотря на относительно подробное рассмотрение этой темы в предыдущих разделах. Класс может быть объявлен либо как открытый, либо вообще без использования какого-либо модификатора доступа. Открытый класс объявляется следующим образом:

    class ShowFrame

# Открытые классы
Если класс открытый, на него можно ссылаться в любом месте программы. Конечно, это правило применимо с учетом ограничений, обусловленных расположением класса (читайте выше о структуре папок проекта). Необходимо также правильно сослаться на пакет, в котором определен класс. Например, в листинге 14.1 можно написать так.

    import com. burdbrain.drawings. Drawing;
    import com.burdbrain.frames.ArtFrame;
    //...
    ArtFrame artFrame = new ArtFrame(new Drawing());
Можно обойтись и без объявлений import и написать так.

    com. burdbrain.frames.ArtFrame artFrame= new com. burdbrain.frames. ArtFrame(new com. burdbrain.drawings.Drawing());
    
В любом случае компилятор должен откуда-то узнать, что классы ArtFrame и Drawing находятся в именованных пакетах.
