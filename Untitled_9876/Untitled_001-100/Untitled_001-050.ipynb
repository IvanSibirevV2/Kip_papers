Как хорошо было бы, если бы с компьютером можно было общаться на естественном языке. К сожалению, пока что мы этой возможности не имеем. Кто-то должен хорошенько подумать и решить, каким образом можно сформулировать для компьютера, что именно требуется сделать. После этого кто-то другой должен написать набор инструкций, которые компьютер обязан выполнить.
В научно-фантастических фильмах и романах встречается множество сюжетов, в которых выполнение роботами команд, отдаваемых на человеческом языке, оборачивалось катастрофическими последствиями. Естественные языки не подходят для общения с компьютером в силу следующих причин.

- Предложение на естественном языке может быть неправильно истолковано. Слушающий человек правильно понимает вкладыва- емый смысл на основе текущей ситуации, своего жизненного опыта и других факторов, но компьютер лишен жизненного опыта человека и не осознает текущую ситуацию. Например, когда раздается звонок и человек говорит "Открой дверь!", другой человек понимает, что нужно открыть входную дверь, а не дверь шкафа или кухни, хотя в предложении "Открой дверь!" нет информации о том, какую именно дверь нужно открыть.
- Очень сложные команды нелегко выразить на естественном языке. Сформулировать на естественном языке сложную команду в виде длинной цепочки сложносочиненных и сложноподчиненных предложений весьма затруднительно.
- Предложение на естественном языке обычно перегружено побочными смыслами и дополнительной, ненужной информацией. Даже человек иногда не может отделить в предложении "зерна от плевел".
- Трудность интерпретации команд на естественном языке. Интерпретация команд, сформулированных на естественном языке, представляла бы не меньшие, если не большие, трудности, чем подготовка самой формулировки такой команды.

# Немного истории
Ниже дан краткий экскурс в историю компьютерного программирования.

- 1954-1957 годы: разработка языка FORTRAN. FORTRAN был первым высокоуровневым языком программирования (термин "высокоуровневый" означает, что язык состоит не из машинных команд, а из инструкций, понятных человеку и близких к естественному языку). Долгое время FORTRAN был единственным языком, применяемым для научных вычислений. Его долговечность удивительна: он применялся еще долгое время после того, как ушли со сцены многие более поздние языки. Даже в наши дни существуют трансляторы FORTRAN для всех современных операционных систем. Объясняется это не только уважением к почтенному прошлому, но и тем, что в исследовательских учреждениях до сих пор используется большое количество научных программ на языке FORTRAN.

- 1959 год: Грейс Хоппер, работавшая в компании Remington Rand, разработала язык COBOL. Буква "В" означает "business" (бизнес), и неспроста, потому что это главное, для чего был создан COBOL. Основной задачей языка была последовательная обработка записей. В течение нескольких лет после выпуска транслятора COBOL этот язык начал широко использоваться для обработки коммерческих данных.
- 1972 год: Деннис Ритчи, работая в AT&T Bell Labs, создал язык С. Примеры кода Java, которые вы будете встречать в книге, весьма похожи на примеры кода С. В коде на языке С используются фигурные скобки, операторы if и for, а также другие подобные конструкции. С точки зрения функциональных возможностей на языке С можно решать все задачи, выполняемые с помощью FORTRAN, Java и других современных языков программирования. Различие между ними не в возможностях, а в удобстве и простоте работы с большими программами. В этом отношении Java значительно превосходит С.

- 1986 год: Бьярне Страуструп, опять же, в AT&T Bell Labs, разработал язык С++. В отличие от своего непосредственного предка С, язык С++ поддерживает модель объектно-ориентированного программирования (об этом мы поговорим далее). Это был огромный шаг вперед.

Это делает Java языком, идеальным для распространения кода через Интернет. Кроме того, Java оказался прекрасным языком общего назначения. С его помощью можно создавать приложения с графическим интерфейсом пользователя, управлять базами данных, писать приложения для мобильных устройств и т.п. # Объектно-ориентированные языки
Java - объектно-ориентированный язык. Что это значит? В отличие от таких процедурных языков, как FORTRAN, позволяющих программисту определять каждый шаг в процессе решения задачи, в центре внимания объектно- ориентированных языков находится структура данных. Разумеется, объектно- ориентированные программы по-прежнему приказывают компьютеру, что он должен сделать, но при создании прогаммы разработчик начинает с организации данных, а команды появляются позже.
Концепция объектов позволяет усовершенствовать структуру данных и тем самым расширить возможности их эффективной обработки. При изменении способов обработки данных можно воспользоваться существующими объектами, дополняя их новыми методами, а не начинать разработку "с нуля". Несмотря на то что программисты, как правило, схватывают все "на лету", им также требуется некоторое время, чтобы привыкнуть к новому подходу. # ИЗВИЛИСТАЯ ДОРОГА ОТ FORTRAN K JAVA
В середине 1950-х годов команда разработчиков создала язык программирования FORTRAN. Он был основан на том, что компьютеру отдавались четкие приказы, записанные в коде программы. Приказы, называемые командами или инструкциями, были, с одной стороны, похожи на слова английского языка (IF, FOR, BEGIN и т.п.), а с другой - намного более четкими и однозначными, не допускающими двоякого толкования. На протяжении двадцати последующих лет разработчики создали множество новых компьютерных языков. Многие из них были основаны на той же модели, что и FORTRAN, т.е. программа на таком языке представляла собой последовательность четких инструкций компьютеру. Такие языки называются процедурными. Наиболее популярным в этом ряду был язык с необычным названием, состоящим всего из одной буквы: С. Разрабатывались также языки, отклоняю- щиеся от командного принципа. Например, в языках SIMULA и Smalltalk команды были отодвинуты на второй план, а главное внимание было сосредоточено на описании данных. В этих языках не принято писать "Найти подозрительные счета и распечатать их". Вместо этого нужно определить, что такое счет, какие именно счета нужны и какого рода операция над ними должна быть выполнена. 

Такие языки в большей степени не командные, а описательные (их принято называть декларативными). Приблизительно в это же время возникла и другая ветвь языков, со временем победившая в конкурентной борьбе: объектно-ориентированные языки. Как и в процедурных языках, программа на объектно-ориентированном языке представляет собой последовательность команд, однако на этом их сходство заканчивается. В объектно-ориентированной программе все вращается вокруг данных. Объекты - это наборы данных и методов работы с ними. Программа состоит из объектов, а команды, хотя и выполняются, играют вспомогательную роль. Со временем опыт показал, что объектно-ориентированные языки - прекрасный инструмент для создания сложных программ большими коллективами разработчиков. Это объясняется следующими причинами.

- Программист должен больше внимания уделять не командам, а данным, потому что именно в них заключается суть любой задачи.
- Объекты позволяют разделить работу над большой программой. Каждый программист может работать над своим объектом, не мешая своим коллегам. Принципы объектно-ориентированного программирования позволяют многократно использовать одни и те же объекты, часто даже в разных программах. Любые же попытки научить созданную ранее программу на FORTRAN новым трюкам показывают, насколько старые программы хрупкие. Они просто перестают работать.

В 1970-е годы декларативные языки, такие как SIMULA и Smalltalk, сошли со сцены, и осталисьлишь в компьютерных журналах, обсуждаемые своими упрямыми сторонниками. Между тем процедурные языки типа FORTRAN размножались, как кролики.
Но все изменилось в 1986 году, когда Бьярне Страуструп создал С++, который стал первым мощным объектно-ориентированным языком. Этот язык быстро завоевал популярность благодаря тому, что в нем использовались синтаксис и терминология распространенного в то время языка С. К старому доброму С была лишь добавлена объектно-ориентированная структура. Многие компании немедленно отвернулись от традиционных FORTRAN и С, приняв в качестве стандарта программирования язык С++.

Стр. 34

-----

Однако С++ обладаетрядом существенных недостатков. Используя С++, можно легко обойти стороной все объектно-ориентированные принципы и написать программу в стиле FORTRAN или С. Начиная писать на С+ + программу для решения упомянутой выше задачи выбора заказов, можно пойти по одному из двух путей.
- Начать непосредственно с команды вывода на экран подозрительных счетов, а затем создать команду распечатки найденных заказов.
- Начать с определения объекта "счет". Это и будет объектно-ориентированный подход. Вы должны определить, какие данные характеризуют счет (принадлежность, дата создания, выполняемые транзакции и т.п.) и как обрабатываются эти данные.

Одни программисты считают, что С++ предоставляет лучшие средства обоих подходов, другие же (и я отношусь к ним) возражают, что современные программы не должны создаваться в стиле FORTRAN или С, поэтому нужно исключить возможность написания программ в этом стиле. Простую программу легче и быстрее написать в процедурном стиле, и многие программисты так и поступали бы. Но такая программа плохо вписывается в современную объектно-ориентированную среду.
В 1995 году Джеймс Гослинг, работая в то время в компании Sun Microsystems, создал язык Java. Он заимствовал из С++ синтаксис и терминологию, но безжалостно выбросил все командные средства. Затем он добавил в Java средства, облегчающие применение принципов объектно-ориентированного программирования. В результате получился чистый и строгий язык. При программировании на Java у вас нет выбора: язык вынуждает вас работать только с объектами, как принято в современном объектно-ориентированном мире компьютерных программ.

-------

# Объекты и классы
В объектно-ориентированном языке для организации данных используются объекты и классы.
Предположим, вам нужно написать программу, осуществляющую информационную поддержку строительства коттеджей для нового жилого комплекса. Коттеджи немного отличаются один от другого. Разными могут быть такие параметры, как цвет фасада, внутренняя отделка, стиль кухни и т.п. В вашей объектно- ориентированной компьютерной программе каждый коттедж -- это объект. Однако объекты -- это еще не все. Несмотря на неизбежные различия между коттеджами, мы можем использовать для их описания некие общие характеристики, например такие, как цвет фасада. Каждому дому присущи и другие характеристики, например стиль кухни. В объектно-ориентированной программе должен быть главный список, включающий все характеристики коттеджей, которые мы собираемся учитывать. Такой список называется классом. Возможно, лучше было бы назвать такой стиль не объектно-ориентированным программированием, а программированием на основе классов и объектов.

Слово "классы" я поместил на первое место не случайно, а потому, что понятие класса важнее понятия объекта. Представьте себе строительную площадку. В ее углу посреди большой лужи стоит трейлер, служащий конторой прораба, в которой хранится чертеж коттеджа. Чертеж -- это аналог класса, поскольку он содержит исчерпывающую информацию о каждом коттедже и список изменяемых параметров. Например, на чертеже может быть просто написано "фасад", тогда как фактический объект коттеджа будет иметь серый фасад. На чертеже может быть указано "кухня". Ну а реальный объект коттеджа может быть оборудован кухней в стиле хай-тек.
Аналогия между чертежом и классом не ограничивается наличием списка параметров. Подобие между ними более близкое. Чертеж создается один раз. На его основе подрядчик может построить хоть десяток коттеджей. То же самое справедливо и для классов. Сначала программист один раз создает код класса, а затем компьютер по мере выполнения программы многократно создает объекты, относящиеся к данному классу.
В этом и заключена суть отношений между классами и объектами: программист описывает класс, а затем на основе определения класса компьютер создает индивидуальные объекты. Часто объекты называются экземплярами класса.

# Преимущества объектно-ориентированного подхода
В продолжение рассмотренного в предыдущем разделе примера строительства коттеджей предположим, что была написана компьютерная программа, предназначенная для отслеживания инструкций по строительству новых домов. И ваш начальник распорядился, чтобы половина новых коттеджей имела три спальни, а оставшаяся половина - четыре.
Если вы программируете с применением устаревшего стиля FORTRAN/C, то, скорее всего, воспользуетесь следующими инструкциями.

    Выройте котлован для подвала.
    Залейте бетоном периметр котлована.
    Установите каркас для подвала размером два на четыре.
    ...

При использовании подобного подхода вместо краткого и понятного проекта в распоряжении архитектора окажется длинный список инструкций. Если возникнет необходимость в изменении проекта, придется просматривать длинный список инструкций, чтобы найти, например, инструкции по строительству спален. И что еще хуже, эти инструкции могут быть разбросаны по разным страницам, например находиться на страницах 234, 394-410 , 739, 10 и 2. Все еще более усложнится, если строителям, к примеру, придется расшифровывать инструкции, написанные другими людьми.

В то же время классы (как и при использовании проектов) позволяют применить принцип наследования, заключающийся в следующем. Вы создаете базовый класс - проекты первого и второго этажей, одинаковые для трех- и четырехкомнатных коттеджей. Этот базовый класс не включает перегородок между комнатами. Затем вы создаете два производных (подчиненных) класса путем расширения базового класса за счет включения в него всего того, чего в нем недостает, т.е. проекта второго этажа. Производный класс наследует все характеристики базового класса, поэтому оба производных класса будут содержать все проекты, необходимые для построения коттеджа с тремя и с четырьмя спальнями. Назовем два производных класса (проекта) ТрехкомнатныйКоттедж и Четырехкомнатный Коттедж.

Ваши коллеги-строители будут просто поражены силой вашей логики и степенью организации, но все равно зададут определенные вопросы. И один из вопросов будет звучать так: "Вы назвали один из проектов трехкомнатным коттеджем. Но этот проект содержит лишь план второго этажа, а не всего коттеджа. Как это понимать?"

В ответ на это вы снисходительно улыбнетесь и скажете следующее: "В проекте трехкомнатного коттеджа сказано: для получения информации о нижних этажах обратитесь к исходному проекту дома. Фактически проект трехкомнат- ного коттеджа описывает весь коттедж. То же самое можно сказать и о проекте четырехкомнатного коттеджа. Благодаря этому можно воспользоваться всеми ранее созданными проектами, чтобы разработать оригинальный проект коттеджа и сэкономить кругленькую сумму".

Класс Коттедж является суперклассом по отношению к классам Трехком- натныйКоттедж и ЧетырехкомнатныйКоттедж. Если в будущем кто-то захочет спроектировать пятикомнат- ный коттедж, он сможет сэкономить 90% рабочего времени, создав проект на основе класса Коттедж".

Стр. 38

----

Объектно-ориентированные языки появились не по чьей-то прихоти. Их появление было обусловлено тем, что в индустрии программирования наступил глубокий, затяжной кризис. Сначала программисты долго писали код, а затем, когда обнаруживалось, что нужно что-то изменить, им приходилось выбрасывать написанный код и начинать все "с нуля". Оказалось, что, когда в написанный код необходимо внести малейшее изменение, его невозможно использовать повторно, а легче написать код с самого начала. Это была не единственная причина кризиса в программировании. Другая, не менее важная причина состояла в том, что при создании большой программы люди неумышленно пере- определяли данные друг друга, причем выяснить, какая инструкция ошибочно изменила некоторое значение, было очень тяжело, потому что это могло произойти в любом месте огромной программы. Для создания большой программы жизненно необходимы средства изоляции ее частей другот друга. Все эти проблемы были решены в объектно-ориентированных языках.

# Наглядное представление классов и объектов
Создавая программу на Java, вы постоянно будете работать с классами и объектами. Оба этих понятия весьма важны, поэтому в данной главе я подвожу вас к ним с разных сторон с помощью ряда аналогий.
Закройте на минутку глаза и подумайте о том, что мы подразумеваем, когда говорим, что некоторый предмет является стулом.
У стула есть сиденье, спинка и ножки. Каждое сиденье имеет форму, цвет, степень мягкости и т.п. Это свойства, которыми обладает стул. Слово "стул", как и класс, - это не сам стул, а мысленная модель, или понятие стула. В терминах объектно-ориентированного программирования класс Стул -- это не предмет, а описание или представление класса предметов.
Теперь оторвите взгляд от книги и посмотрите вокруг себя. Вы видите в комнате несколько стульев, каждый из которых является конкретным объектом. Каждый объект -- это реализация идеи стула, а говоря языком объектно- ориентированного программирования -- экземпляр класса Стул.

- запомни
    - Класс - это не набор предметов, а абстрактная идея, отражающая конкретные предметы. Когда я говорю о классе стульев, расположенных в вашей комнате, я имею в виду, что в комнате есть предметы со спинками, сиденьями и ножками, имеющие характеристики, определенные в классе Стул. Разные стулья обладают разным цветом, но в классе Стул определен не цвет, а параметр цвета, который у разных экземпляров может иметь разные значения, т.е. конкретные цвета (зеленый, синий и т.п.).

В большинстве случаев имеет смысл представлять себе объект как конкретный экземпляр класса. Фактически официальная терминология совпадает с этим представлением. Когда вы пишете на Java программу, в которой используется класс Стул, каждый конкретный объект стула является экземпляром класса Стул. Рассмотрим еще одно наглядное представление понятия класса. В соответствии с этой таблицей каждый счет характеризуется номером и типом счета, а также балансом находящихся на нем денежных средств. # Что должно быть установлено на компьютере
Однажды я встретил слесаря, и он показал мне инструменты, с помощью которых создаются лекала для изготовления других инструментов. Я был в восторге от этой встречи и уже тогда решил для себя, что когда-нибудь обязательно проведу аналогию между компьютерными программистами и изготовителями лекал.

Программист использует существующие программы в качестве инструментов для создания новых программ. Существующие и вновь создаваемые программы решают совершенно разные задачи. Например, программа на Java (создаваемая вами) может быть предназначена для учета заказов. Для ее разработки вы используете (вернее, будете использовать в ближайшем будущем) уже существующую программу, которая помогает вам вводить код, находить ошибки, запускать создаваемую программу на выполнение и т.п. С ее помощью нельзя решить поставленную задачу, но можно создать программу, от- слеживающую заказы. Что нужно, чтобы начать создавать программы на Java? Как новичку вам понадобятся всего лишь три инструмента.

- Компилятор - Это программа, которая преобразует код на языке Java в формат, пригодный для выполнения на компьютере (байт-код).
Люди не могут легко создавать или декодировать инструкции в формате байт-кода. Для генерирования и интерпретации байт-кода используется определенное программное обеспечение, устанавливаемое на компьютере.
- Виртуальная машина Java - Это программа, которая выполняет на компьютере созданный вами код (а также Java-код, написанный другими людьми).
- Интегрированная среда разработки - Интегрированная среда разработки (IDE) упрощает работу с кодом и предоставляет удобные средства для его написания, компиляции и выполнения. По правде говоря, устанавливать интегрированную среду разработки вовсе не обязательно. Некоторые разработчики принципиально используют самые обычные текстовые редакторы, такие как Блокнот (Windows), TextEdit (Macintosh) или vim (Unix). Однако начинающим программистам все же лучше пользоваться специальной рабочей средой, в которой все будет получаться намного легче и быстрее.

В Интернете предлагается множество бесплатных версий каждого из перечисленных выше инструментов.

Инструкции в этом листинге представляют собой так называемый байткод Java. Создавая программу на языке Java, вы пишете инструкции исходного кода (как в листинге 2.1), после чего обрабатываете этот код с помощью специальной программы, называемой компилятором, которая транслирует (преобразует) исходный код в байт-код, выполняемый компьютером. Иными словами, компилятор получает код, подготовленный на понятном для вас языке (см. листинг 2.1), и переводит его на язык, понятный компьютеру (см. листинг 2.2).

- Технические подробности
    - Предположим, вы сохранили исходный код, приведенный в листин- не 2.1, в файле Hotel.java. Тогда компилятор, вероятнее всего, поместит байт-код в файл Hotel.class. Если вы любознательны и захотите просмотреть содержимое файла Hotel.class с помощью обычного текстового редактора, то вместо текста, приведенного в листинге 2.2, увидите странные символы: квадратики, точки, стрелочки и т.п. Дело в том, что компилятор не только преобразует исходный код в байт-код, но и специальным образом его кодирует. Чтобы увидеть байт-код в том виде, в каком он показан в листинге 2.2, откройте файл Hotel.class с помощью программы Java Bytecode Editor, доступной по следующему адресу: http://www.cs.ioc.ee/~ando/jbe
- Запомни
    - Ни один разработчик программ на Java не занимается написанием байт-кода, который создается компилятором. Единственным побудительным мотивом к просмотру содержимого файлов .class может быть лишь ваша любознательность.

# Что такое виртуальная машина Java
Как отмечалось в предыдущем разделе, компьютер следует инструкциям наподобие тех, которые представлены в листинге 2.2. В последней фразе слово "наподобие" употреблено не случайно. Дело в том, что, несмотря на сходство инструкций в листинге 2.2 с инструкциями, выполняемыми компьютерами, в каждом компьютере вместо байт-кода Java используется собственный набор выполняемых инструкций, зависящий от типа установленного процессора, причем между наборами инструкций одного и того же процессора, предназначенными для разных операционных систем, имеются незначительные различия.
