# ПЛАВАЮЩАЯ ДЕСЯТИЧНАЯ ТОЧКА
B Java есть два числовых типа с плавающей точкой: double и float. Чем они отличаются друг от друга? Объявляя переменную типа double, вы приказываете компьютеру выделить для нее 64 бита памяти, тогда как для переменной float требуется 32 бита.
Вы могли бы изменить листинг 4.2, объявив для переменной amount InAccount тип float: float amountInAccount; Казалось бы, для хранения числа 50. 22 тридцати двух битов более чем достаточно. Однако это не совсем так. Число 50 без проблем представляется 32 битами. Более того, при этом можно было бы обойтись всего лишь б битами. Однако главным фактором здесь является не величина числа, а достигаемая точность.

В 64-битовой переменной типа double большинство битов используются для хранения той части числа, которая записывается после десятичной точки. Для сохранения дробной части числа 50.22, т.е. 0.22, тех 32 битов, которые предоставляет тип float, оказывается уже недостаточно. Не верите тому, что только что прочитали, а именно тому, что для хранения числа 0.22, имеющего десятичную дробную часть, 32 битов недостаточно? Тогда я попытаюсь убедить вас в этом, внеся в листинг 4.2 некоторые изменения и объявив переменную amount InAccount как переменную типа float. В этом случае первые строки кода будут выглядеть так.

В 64-битовой переменной типа double большинство битов используются для хранения той части числа, которая записывается после десятичной точки. Для сохранения дробной части числа 50.22, т.е. 0.22, тех 32 битов, которые предоставляет тип float, оказывается уже недостаточно. Не верите тому, что только что прочитали, а именно тому, что для хранения числа 0.22, имеющего десятичную дробную часть, 32 битов недостаточно? Тогда я попытаюсь убедить вас в этом, внеся в листинг 4.2 некоторые изменения и объявив переменную amount InAccount как переменную типа float. В этом случае первые строки кода будут выглядеть так.

    float amountInAccount;
    amountInAccount = 50.22F;
    amountInAccount = amountInAccount + 1000000.00F;
(Чтобы узнать, что означает буква F в записях 50.22F и 1000000.00F, обратитесь к табл. 4.1.) Запустив программу, вы получите следующий результат:

Стоило мне использовать вместо типа double тип float, как на счету Чарльза оказалось на целых три цента больше, чем ему причитается. Почему так получилось? Потому что переход к 32-битовому типу float привел к потере точности значения, cохраняемого в переменной amountInAccount, уже во втором разряде после десятичной точки, а это недопустимо.
Сиспользованием типа float связана еще одна проблема чисто"косметичес- кого" характера. Согласно синтаксическим правилам Java, литералам 50.22 и 1000000.00 соответствует тип double. Объявляя amount InAccount как переменную типа float, вы сталкиваетесь с проблемой, заключающейся в том, что 64-битовые литералы не умещаются в 32-битовой переменной. Поэтому литералы типа double также должны быть приведены к типу float, что достигается дописыванием буквы F в конце литерала, хотя при этом числа приобретают довольно забавный вид.

# ЧЕТЫРЕ СПОСОБА ХРАНЕНИЯ ЦЕЛЫХ ЧИСЕЛ
B Java определены четыре типа переменных, в которых можно хранить целые числа: byte, short, int и long. В отличие от той ситуации, которую мы разбирали, сравнивая между собой типы float и double, здесь все выглядит гораздо проще. В данном случае единственным решающим фактором при выборе целочисленного типа является величина числа, которое вы хотите сохранить в переменной. Тип byte нельзя использовать для хранения чисел больше 127, а тип short - для хранения чисел больше 32767.
Чаще всего вы будете использовать тип int. Но если вам потребуется сохранить число, превышающее 2147483647, то следует использовать тип long. (Для типа long максимально допустимым значением является 9223372036854775807.) Более подробно целочисленные типы описаны в табл. 4.1.

# Инициализация при объявлении
Вернемся к листингу 4.3. Для объявления трех переменных типа int в нем использованы три отдельные инструкции. То же самое можно было сделать с помощью одной инструкции: int weightOfAPerson, elevatorWeightLimit, numberOfPeople;

- внимание
    - Если переменные имеют разные типы, то объявить их в одной инструкции невозможно. Например, для создания переменной WeightOfFred типа int и переменной AmountInFredsAccount типа double необходимо использовать два отдельных объявления.
    
Одновременно с объявлением переменной можно указывать ее начальное значение. Например, как показано в листинге 4.3, одно объявление может заменить несколько строк метода main (за исключением вызовов print и println):

    int weightOfAPerson = 68, elevatorWeightLimit = 635, numberOfPeople = elevatorWeightLimit/ weightOfAPerson;

Описанные выше действия не называются присваиванием значений переменных. Фрагменты объявлений, включающие знаки равенства, именуются не операторами присваивания, а инициализацией переменных. Различия между присваиванием и инициализацией существенны, и их следует учитывать при разработке кода.

Как и многое другое, инициализация переменной имеет свои преимущества и недостатки.
- Объединение шести строк листинга 4.3 в одну позволяет сократить объем кода. В одних случаях это упрощает чтение кода, в других - затрудняет. Решение о том, как лучше поступить, программист принимает, исходя из специфики ситуации.
- Инициализация переменных позволяет автоматически избежать некоторых ошибок. Подробнее об этом - в главе 7.
- В некоторых ситуациях увас не будетвыбора: сам характер кода будет диктовать вам, инициализировать ли ту или иную переменную. Пример подобной ситуации рассматривается в главе б.

    //Скрин так и приведен

Необходимость подобного введения в программе Java объясняется тем, что:
    - в Java вся программа является классом;

В любом случае повторный ввод шаблонного кода в окне редактора может раздражать, особенно в том случае, когда нужно проверить результат выполнения всего нескольких инструкций. Чтобы устранить эту проблему, разработчики в версии Java 9 предложили новое средство: JShell.
- в интернете

При использовании JShell вам вряд ли когда-либо придется вводить код всей программы. Достаточно ввести инструкцию Java, на которую JShell отреагирует соответствующим образом. После этого введите вторую инструкцию, на которую JShell снова отреагирует должным образом, третью инструкцию и т.д. Как видите, единственной инструкции достаточно для получения отклика от JShell. JShell -- это пример языка, включающего среду REPL (Read Evaluate Print Loop - цикл "чтение-вычисление печать"). Многие языки программирования включают эту среду; к их числу относится и Java 9. Как показано на рис. 4.7, с помощью JShell определяется, каким образом Java реагирует на инструкции присваивания, находящиеся в листингах 4.2 и 4.3. При использовании JShell диалоговое окно принимает следующий вид.

    jshell> Вы ввели инструкцию JShell отвечает
    jshell> Вы ввели другую инструкцию JShell отвечает
    
Например, как показано на рис. 4.7, была введена инструкция double amountInAccount и нажата клавиша <Enter>. В ответ на это JShell выводит следующую строку: amountInAccount ==> 0.0

    - B JShell точки с запятой обычно необязательны.Как показано на рис. 4.7, точка с запятой содержится в конце лишь двух из девяти строк. В главе 5 рассматривается использование точек с запятой в JShell.
    - Среда JShell немедленно реагирует на ввод пользователем каждой строки кода.
После объявления переменной amount InAccount как double JShell сообщает о том, что переменная amountInAccount имеет значение 0.0. После ввода инструкции amountInAccount = amountIn Account + 1000000.00 JShell сообщает о том, что новое значение переменной amount InAccount будет 1000050.22.

- Можно использовать инструкции, находящиеся в разных программахJava. Как показано на рис. 4.7, применяются инструкции, находящиеся в листингах 4.2 и 4.3. Для JShell это не имеет никакого значения.
- В среде JShell можно запросить значение выражения. Вовсе не обязательно присваивать значение выражения переменной. Например, как показано на рис. 4.7, можно ввести следующую строку: elevatorWeightLimit / weightOfAPerson ; В ответ на этот запрос JShell выдает сообщение о том, что значение выражения elevatorWeightLimit / weightOfAPerson равно 9. Для этого значения JShell создает временное имя. Как показано на рис. 4.7, в качестве этого имени используется $\$$8. Поэтому в следующей строке, показанной на рис. 4.7, в ответ на запрос значения для $\$$8 + 1 JShell отображает 10.
- Можно получать ответы от JShell даже без использования переменных. В последней строке кода, показанной на рис. 4.7, запрашивается
значение выражения 42 + 7, и JShell отображает ответ 49.

- совет
    - В среде JShell вам не придется повторно вводить ранее указанные команды. Достаточно лишь один раз нажать клавишу <1>, как JShell отобразит список недавно введенных команд. Если нажать эту же клавишу дважды, то JShell отобразит предпоследнюю введенную команду. Как только JShell отобразит команду, можно воспользоваться клавишами <<-> и <->>, чтобы перейти к любому символу команды. Также можно изменять символы, образующие команду. И наконец, как только вы нажимаете клавишу $<$Enter$>$, JShell вызывает на выполнение обновленную команду.
    
Чтобы завершить выполнение среды JShell, введите команду /exit (начните с символа косой черты). Учтите, что /exit является лишь одной из множества команд, которые поддерживаются в JShell. Чтобы получить сведения о других командах JShell, введите /help. В среде JShell можно тестировать инструкции, прежде чем использовать их в программе на Java. Благодаря подобным возможностям JShell превращается в очень полезный инструмент отладки кода.

# А как же визуальные эффекты?
Программы, код которых приведен в листингах 4.2 и 4.3, основаны на использовании текста. У такой программы отсутствуют диалоговые окна и другие подобные объекты. Все, что вы видите на экране, -- простой неформа- тированный текст. Пользователь вводит какой-либо текст, в ответ на который отображается результат в виде обычного текста.
Полную противоположность текстовым программам представляют программы с графическим интерфейсом пользователя (graphical user interface - GUI). Такая программа имеет окна, текстовые поля, кнопки и другие визуальные компоненты.
