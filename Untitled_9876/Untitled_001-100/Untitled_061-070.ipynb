# Слова в программе на Java
Любой профессионал скажет вам, что в исходном коде Java используются слова двух категорий: идентификаторы и ключевые слова. Это действительно так. Но если не дополнить это утверждение необходимыми пояснениями, то в некоторых случаях оно может ввести в заблуждение. Поэтому я несколько расширю формулировку и скажу, что в Java используются слова трех категорий: ключевые слова, идентификаторы API и пользовательские идентификаторы.Различие между этими тремя категориями приблизительно такое же, как и между словами естественного языка. В предложении "Барри - человек" слово человек - аналог ключевого слова Java. Независимо от того, кто его использует, оно всегда означает примерно одно и то же.

Что касается слова Барри, то оно играет ту же роль, что и идентификаторы в Java, поскольку означает имя конкретной личности. Такие слова, как Барри, Джон или Алекс, не наделены никаким изначальным смыслом. Они применяются в отношении различных людей в зависимости от обстоятельств и являются просто именами, которыми родители нарекают своих детей.
Теперь рассмотрим предложение "Юлий Цезарь - человек". Тот, кто произносит эту фразу, вероятнее всего, имеет в виду человека, который правил Римом до мартовских ид. Несмотря на то что за именем Юлий Цезарь в языке не закреплен предопределенный смысл, почти каждый сразу же поймет, о ком идет речь. Продолжая нашу аналогию между языком программирования Java и естественным языком, имя "Юлий Цезарь" можно сравнить с идентификатором API. Итак, резюмируем вышесказанное: в исходном коде Java используются слова следующих категорий..

Стр. 62

-------

- Ключевые слова. Это слова, имеющие специальное значение в языке программирования Java, смысл которых остается одним и тем же в любой программе. Примеры ключевых слов: if, else, do. Полный набор ключевых слов Java определяется спецификацией, которая утверждается комитетом JCР. Никакой придуманный вами идентификатор не должен совпасть ни с одним ключевым словом Java.
- Идентификаторы. Это просто имя, присвоенное какой-либо сущности. Один и тот же идентификатор может иметь разный смысл в разных программах, но для одних идентификаторов это более характерно, чем для других.
- Пользовательские идентификаторы. Создавая программу на Java, вы придумываете новые имена для классов и других сущностей, встречающихся в исходном коде. Предположим, у вас в программе используется имя Prime, а программист из соседнего отдела случайно использовал аналогичное имя в своей программе. Грозит ли это какими-либо неприятностями? Абсолютно никакими, поскольку слово Prime не имеет в Java заранее установленного смысла. Например, в вашей программе оно может означать начальный уровень обучения, тогда как в программе вашего коллеги - первоклассный уровень обслуживания. В данном случае никакого конфликта имен не возникает, потому что они используются в двух разных программах.
- Идентификаторы API. Участники проекта JCP создали классы Java API, в которых используется более 40 тысяч имен. Классы API поставляются в составе каждого дистрибутива JDK, так что все эти имена доступны каждому, кто пишет программы на Java. Примеры идентификаторов API: String, Integer, JWindow, JButton, JTextField, File.

# Методы в Java
Предположим, вы работаете автомехаником в гараже. Ваш начальник, который всегда спешит, имеет привычку отдавать приказы скороговоркой: "Нужно починитьГенератор в том старомСеромФорде". Мысленно вы составляете список задач: "Загнать машину на яму, поднять капот, взять ключ, ослабить ремень генератора и т.п." Во всем этом можно выделить три компонента.

- Имя задания, которое вы должны выполнить. В данном случае имя задания - починитьГенератор.
- Мысленный список задач, ассоциируемых с именем задания починитьГенератор. Задачи решаются строго в той последовательности, в какой они указаны в списке. Например, нельзя сначала ослабить ремень генератора, а затем поднять капот.
- Сердитый начальник, который дал вам задание. Ваш начальник заставляет вас приступить к работе, произнеся слова починитьГене- ратор. Иначе говоря, начальник называет имя задания, которое вы должны выполнить.

В этом сценарии вполне уместно использовать термин метод. Вы знаете, как починить генератор. На языке программирования это означает, что у вас есть метод починки генератора. Ваш начальник посредством словесного приказа приводит этот метод в действие, а вы выполняете все операции, связанные с данным методом.
Если вы согласны с таким пониманием термина "метод", значит, готовы к восприятию того, что такое метод в языках программирования. В Java метод - это список инструкций, подлежащих выполнению. У каждого метода есть имя, и вы сообщаете компьютеру список того, что он должен сделать, указывая имя метода в конструкции языка, которая называется вызовом метода. Я никогда не писал программу для робота, ремонтирующего генератор, но если бы мне пришлось это сделать, я назвал бы метод починитьГенератор.
Список инструкций в теле метода починитьГенератор выглядел бы приблизительно так, как показано в листинге 3.2.

- Объявление метода в листинге 3.2 состоит из двух частей. Первая строка объявления (исключая открывающую фигурную скобку) - это заголовок метода. Остальная часть листинга 3.2 (фигурные скобки и все, что между ними) - тело метода.
- Термин объявление метода отличает список инструкций в листинге 3.2 от инструкции в листинге 3.3 - вызова метода.
<br/><br/>
- Запомни
    - Объявление метода предоставляет компьютеру информацию о том, какие действия должны быть выполнены в случае вызова метода, тогда как вызов метода (отдельная часть кода) -- это фактический приказ компьютеру выполнить действия, заданные в объявлении. Как правило, объявление и вызов метода - это две разные части программы Java.
