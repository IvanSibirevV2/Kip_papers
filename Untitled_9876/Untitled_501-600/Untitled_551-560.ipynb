    MyThing myThing, myThing2;
    myThing = new MySubThing();
    myThing2 = new MyOtherThing();

Даже несмотря на то что переменные myThing и myThing2 имеют тип MyThing, Java не вызывает метод display, объявленный в классе MyThing.
