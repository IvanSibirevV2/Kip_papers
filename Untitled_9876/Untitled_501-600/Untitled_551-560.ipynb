        public int getNumberOfBedrooms() { return numberOfBedrooms; }
        public void setNumberOfBedrooms (int numberOfBedrooms)
        { this.numberOfBedrooms = numberOfBedrooms; }
        public double getArea(){ return area;}
        public void setArea(double area) { this.area = area; }
    }

    public class House extends PlaceToLive
    {
        private double mortgage;
        private double propertyTax;
        public double getMortgage(){ return mortgage; }
        public void setMortgage (double mortgage) { this.mortgage = mortgage; }
        public double getPropertyTax() { return propertyTax; }
        public void setPropertyTax(double propertyTax) { this.propertyTax = propertyTax; }
        public double getTotalMonthlyPayment () { return mortgage + propertyTax / 12; }
    }
    public class Apartment extends PlaceToLive
    {
        private double rent;
        public double getRent() { return rent; }
        public void setRent (double rent) { this.rent = rent; }
