    package com. burdbrain.drawings;
    import java.awt.Color;
    import java.awt.Graphics;
    public class Drawing 
    {
        public int x = 40, y = 40, width = 40, height = 40;
        enum State { gReeN, YelloW, REDЪ

        State state = State.GREEN;
        public void paint (Graphics g) 
        {
            switch (state)
            {
                case GREEN:
                    showGreen (g, false);
                    showYellow(g, true);
                    state = State.YELLOW;
                    break;
                case YELLOW:
                    showYellow(g, false);
                    showRed(g, true) ;
                    state = State.RED;
                    break;
                case RED:
                    showRed(g, false);
                    showGreen (g, true);
                    state = State. GREEN; break;
            }
        }

        void showGreen (Graphics g, boolean t)
        {
            g.setColor(t ? Color.GREEN : Color.WHITE);
            g.fill0val(x, y + 100, width, height);
        }
        void showYellow(Graphics g, boolean t)
        {
            g.setColor(t ? Color.YELLOW : Color.WHITE);
            g.fill0val (x, y + 50, width, height);
        }
        void showRed(Graphics g, boolean t)
        {
            g.setColor(t ? Color.RED : Color.WHITE);
            g.fill0val(x, y, width, height);
        }
    }
