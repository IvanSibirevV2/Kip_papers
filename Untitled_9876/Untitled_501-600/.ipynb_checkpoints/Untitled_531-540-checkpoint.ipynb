            int number = keyboard.nextInt();

            while (number > 0)
            {
                if (number > largest){largest = number;}
                System.out.print("Введите число (0 или отрицательное, чтобы выйти): ");
                number = keyboard.nextInt();

            System.out.println("Наибольшее число: " + largest); 
            keyboard.close ();
        }
    }

    Цена за 100 грамм, цена одной порции и общее количество калорий в продукте
    public class FoodProduct
    {
        String type;
        double weight;
        int servings;
        int calories;
        double cost;
        public void display() 
        {
            System.out.print (type);
            System.out.print(" весит ");
            System.out.print (weight);
            System.out.print(" грамм, pасфасован на ");
            System.out.print (servings);
            System.out.print (" порций и содержит ");
            System.out.print (calories);
            System.out.println(" калорий.");
            System.out.println(" Цена за 100 грамм: " + costPer100Grams() + " " ) ;
            System.out.println(" Цена порции: " + costPerServing() + ".");
            System.out.println(" Общее количество калорий: " totalCalories() + ".");
        }
        public double costPer100Grams(){ return cost / (weight / 100);}
        public double costPerServing() { return cost / servings;}
        public int totalCalories() { return calories * servings; }
    }
