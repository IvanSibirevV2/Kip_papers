# Подсчет количества игральных карт
    public class PlayingCardPlus extends PlayingCard
    {
        public static int count = 0;
        public PlayingCardPlus (Suit suit, int number)
        {
            super(suit, number);
            count++;
        }
        public class Main
        {
            public static void main(String[] args)
            {
                PlayingCardPlus cardl = new PlayingCardPlus (Suit.ПИКИ, 13);
                PlayingCardPlus card2 = new PlayingCardPlus (Suit.ЧЕРВИ, 1);
                PlayingCardPlus card3 = new PlayingCardPlus (Suit.БУБНЫ, 3);
                System.out.println(cardl.getString());
                System.out.println(card2.getString());
                System.out.println(card3.getString());
                System.out.println("Количество карт: "+ PlayingCardPlus.count);
        }
    }

Средняя величина значений в массиве

    public class Main
    {
        public static void main(String[] args)
        {
            double[] values = (3.4, 5.7, 9.8, 0.0, 10.1);
            double total = 0;
            for (double value : values) { total += value; }
            System.out.println(total / values.length) ;
        }
    }
