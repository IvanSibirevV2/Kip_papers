        private double area;
        public String getAddress() { return address; }

        public int getNumberOfBedrooms() { return numberOfBedrooms; }
        public void setNumberOfBedrooms (int numberOfBedrooms)
        { this.numberOfBedrooms = numberOfBedrooms; }
        public double getArea(){ return area;}
        public void setArea(double area) { this.area = area; }
    }

    public class House extends PlaceToLive
    {
        private double mortgage;
        private double propertyTax;
        public double getMortgage(){ return mortgage; }
        public void setMortgage (double mortgage) { this.mortgage = mortgage; }
        public double getPropertyTax() { return propertyTax; }
        public void setPropertyTax(double propertyTax) { this.propertyTax = propertyTax; }
        public double getTotalMonthlyPayment () { return mortgage + propertyTax / 12; }
    }
    public class Apartment extends PlaceToLive
    {
        private double rent;
        public double getRent() { return rent; }
        public void setRent (double rent) { this.rent = rent; }
    }

# Каков результат выполнения кода MyThing?
Результат выполнения кода:
    
    B MySUBThing значение равно 7
    B MyOTHERThing значение равно 44
В программе используются следующие объявления

    MyThing myThing, myThing2;
    myThing = new MySubThing();
    myThing2 = new MyOtherThing();
Кроме того, методы display объявлены в классаx MyThing, MySubThing и MyOtherThing. В таком случае Java вызывает методы display, которые объявлены в классах MySubThing и MyOtherThing. Даже несмотря на то что переменные myThing и myThing2 имеют тип MyThing, Java не вызывает метод display, объявленный в классе MyThing.
