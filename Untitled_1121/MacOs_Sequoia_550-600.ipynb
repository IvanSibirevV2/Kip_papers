# Автоматический вход в систему  
Если вам не нравится входить в систему при каждом запуске компьютера Mac, вы можете настроить автоматический вход в систему с определенной учетной записью. Выполните следующие действия:  
1. Выберите "Системные настройки" ⇒ "Пользователи и группы", чтобы открыть панель "Пользователи и группы" приложения "Системные настройки".

# Внимание 
Автоматический вход в систему - это кошмар безопасности, потому что любой, кто может включить ваш Mac, может начать использовать его под учетной записью, которую вы настроили для автоматического входа в систему. В целях безопасности вам — и всем другим пользователям вашего Mac - всегда следует входить в систему вручную.

# AirPlay и воспроизведение AirPlay на Mac

# Укажите Time Machine, какие данные создавать, а какие пропускать  
Если на вашем Mac подключены другие жесткие диски, вам нужно указать Time Machine, какие из них создавать резервные копии, а какие нет. # Лампочка 
Вы также можете указать Time Machine создать резервную копию в любой момент, когда захотите. Просто щелкните значок Time Machine в строке меню, а затем нажмите Создать резервную копию сейчас.  
Для чего создается резервная копия Time Machine? 

Time Machine создает резервную копию всего вашего жесткого диска при первом запуске (за исключением всего, что вы исключили). После этого создается резервная копия файлов и папок, которые были изменены с момента вашего последнего резервного копирования. Это стандартная операционная процедура для систем резервного копирования. Но Time Machine делает больше. Он также создает резервные копии таких объектов, как контакты, изображения в вашей библиотеке фотографий, события в ваших календарях и электронные письма в ваших учетных записях электронной почты, не говоря уже о поддержке версий и блокировке. Единственное, чего Time Machine не создает, так это содержимого домашних папок, отличных от ваших собственных. 

Эти функции делают Time Machine непохожей на любую другую систему резервного копирования.

# Как мне восстановить файл (или контакт, фотографию, событие и т.д.)?  
Чтобы восстановить файл или любую другую информацию, выполните следующие действия: 

1. Запустите приложение, содержащее информацию, которую вы хотите восстановить. 

Если вы хотите восстановить файл, этим приложением является Finder, которое (как вы знаете) всегда запущено; просто откройте окно Finder для папки, содержащей файл. 2. Запустив соответствующее приложение (или открыв соответствующее окно Finder), запустите приложение Time Machine (см. Рисунок 21- 2). 

Если вы выбрали "Показать в строке меню" во всплывающем меню Time Machine на панели "Центр управления" в системных настройках, вы можете щелкнуть значок Time Machine в строке меню, а затем нажать "Просмотреть резервные копии Time Machine" (см. Рисунок 21-1). В противном случае откройте Launchpad, а затем нажмите Time Machine. 

# Лампочка
Будет проще восстановить файл в Finder, если папка, в которой находится (или находилась) файл, является активной (то есть самой открытой) при запуске приложения Time Machine. В противном случае вам придется перейти к соответствующей папке, прежде чем вы сможете выполнить Шаг 3. 

3. Щелкните одну из полосок с датами в правом нижнем углу экрана или щелкните большую стрелку вперед или назад в правой части окна Finder на рисунке 21-2, чтобы выбрать резервную копию для восстановления. 
4. Выберите файл, папку, контакты, фотографию, сообщение электронной почты или событие календаря, которые вы хотите восстановить. 
5. Нажмите кнопку Восстановить под окном.

Ваше первое резервное копирование с помощью коммерческого программного обеспечения может занять от нескольких минут до нескольких часов. Последующие резервные копии — инкрементные резервные копии, на языке программного обеспечения для резервного копирования — должны выполняться намного быстрее. 

Одно из преимуществ хорошего программного обеспечения для резервного копирования заключается в том, что вы можете настроить его таким образом, чтобы автоматизировать резервное копирование и выполнять его, даже если забудете. # Зачем вам нужны две резервные копии

 Однажды вы выходите из офиса на обед. Когда вы возвращаетесь, вы обнаруживаете, что ваш офис был ограблен, поражен молнией, затоплен, сожжен дотла или погребен под обломками при землетрясении — выбирайте сами.

# Проблемы безопасности без резервного копирования  
Резервное копирование ваших файлов имеет решающее значение, если вы не боитесь однажды потерять все свои данные. И хотя резервное копирование, безусловно, является вашей самой важной задачей в области безопасности, вашим данным могут угрожать некоторые другие факторы, такие как вирусы или другие типы вредоносных программ, включая червей, шпионское ПО и атаки злоумышленников. Это плохая новость. Хорошей новостью является то, что все эти вещи с гораздо большей вероятностью затронут пользователей Windows, чем Mac. 

Несмотря на это, рассмотрите возможность принятия следующих мер предосторожности. 

# О вирусах и других вредоносных программах  
Компьютерный вирус - это маленький неприятный фрагмент компьютерного кода, который реплицируется и распространяется с диска на диск. Вирус может привести к неправильной работе вашего Mac; некоторые вирусы могут уничтожать файлы или стирать диски без предупреждения. 

Вредоносное ПО (сокращение от malicious software) - это программное обеспечение, которое является враждебным, навязчивым, раздражающим или разрушительным. Вредоносное ПО часто предназначено для получения несанкционированного доступа к вашему компьютеру или сбора личных данных (включая пароли) без вашего ведома, или и того, и другого. 

Разница между вирусом и другими типами вредоносных программ заключается в том, что вирус распространяется автоматически, в то время как другие формы вредоносных программ полагаются на обман, имитацию и социальную инженерию, чтобы побудить ничего не подозревающих пользователей открыть вредоносный файл или установить вредоносное приложение. Итак, вирус - это разновидность вредоносного ПО, но не все вредоносные программы являются вирусными. 

Почти все вирусы специфичны для конкретной операционной системы. Вирусы Mac не повлияют на пользователей Windows, вирусы Windows не повлияют на пользователей Mac и так далее. Подавляющее большинство известных вирусов поражают только Windows, в основном потому, что это гораздо более крупная, сочная и прибыльная мишень, чем macOS. 

Единственным реальным исключением здесь является “подарок” от чудесного мира пользователей Microsoft Office (например, Word и Excel): страшные макровирусы, которые распространяются с документами Word, Excel и PowerPoint, содержащими макросы, написанные на языке Microsoft Visual Basic для приложений (VBA). Корпорация Майкрософт сделала реализацию VBA несколько более безопасной, но по-прежнему разумно с осторожностью относиться к любому входящему файлу Microsoft Office.

# Запомни 
Если вы используете диски, которые когда-либо были установлены в компьютер, который вы не знаете и которому не доверяете, вам может понадобиться программное обеспечение для обнаружения вирусов. Если вы загружаете и используете файлы из Интернета, вам также поможет обнаружение вирусов. 

- Вам не о чем особо беспокоиться, если
    - Вы загружаете файлы только из коммерческих онлайн-сервисов, таких как CNET или MacUpdate, которые добросовестно относятся к вредоносным ПРОГРАММАМ.
    - Вы покупаете программное обеспечение только в App Store.
    - Вы используете только коммерческое программное обеспечение и никогда не загружаете файлы с веб-сайтов со странными названиями.  
- Вам определенно следует беспокоиться о вредоносном заражении, если
    - Вы посещаете веб-сайты, предлагающие пиратское или украденное программное обеспечение.
    - Вы регулярно обмениваетесь дисками или флэш-накопителями USB с друзьями.
    - Вы перемещаете диски или флэш-накопители USB туда и обратно на другие компьютеры Mac.
    - Вы используете свои диски или флэш-накопители USB на общедоступных компьютерах, фотопечати или других компьютерах, которые, вероятно, были подключены ко многим дискам.
    - Вы загружаете файлы из самых разных мест в Интернете.
    - Вы получаете электронное письмо с вложениями (и открываете их). Обратите внимание, что вы можете получать вредоносное программное обеспечение в сообщениях, которые выглядят так, будто они от людей, которых вы знаете и которым доверяете. Этот тип атаки называется подменой, и его легко осуществить, поэтому тщательно подумайте, прежде чем открывать вложение, и свяжитесь с отправителем, если у вас есть какие-либо сомнения в подлинности сообщения.


Если вы подвержены риску, сделайте себе одолжение: купите коммерческое антивирусное приложение. Или попробуйте бесплатный сканер вредоносных программ, такой как Malwarebytes, Bitdefender или Avira. Если вы считаете, что подвергаетесь риску, просканируйте свой диск с помощью одной из бесплатных утилит, прежде чем тратить наличные. Получите их, если они вам нужны; не делайте этого, если нет.

# Автоматическая установка обновлений системы безопасности или более того  
По умолчанию компьютер Mac раз в день обращается к серверам Apple для поиска любого нового или обновленного программного обеспечения. Чтобы обеспечить максимальную безопасность вашего Mac, настройте его на автоматическую установку любых новых обновлений безопасности без консультации с вами. Вы также можете автоматически устанавливать обновления macOS или приложения из App Store. 

Чтобы настроить проверку наличия обновлений, выберите "Системные настройки" ⇒ "Общие" ⇒ "Обновление программного обеспечения". При открытии панели "Обновление программного обеспечения" подобным образом macOS проверяет наличие любых новых обновлений, поэтому информация, которую вы видите, будет актуальной. Когда проверка будет завершена, нажмите кнопку "Информация" (значок "Я в круге") справа от кнопки "Автоматические обновления", чтобы открыть диалоговое окно "Автоматические обновления" (см. Рисунок 21-3). Затем вы можете настроить автоматическое обновление, установив следующие четыре переключателя:  
- Загружать новые обновления, когда они доступны: Установите этот переключатель в положение Вкл. (синий) для автоматической загрузки обновлений. # Интерпретация запрещающих знаков и сообщений  
Когда вы включаете свой Mac, он запускает аппаратные тесты, а затем проверяет наличие загрузочного диска, содержащего работоспособную копию macOS. # Шаг 2: Загрузитесь в безопасном режиме  
Загрузка вашего Mac в безопасном режиме может помочь вам решить проблему с запуском, не загружая несущественное программное обеспечение (и не macOS) во время загрузки. Опять же, способ вызова безопасного режима зависит от типа процессора вашего Mac:  
- Компьютеры Mac с процессорами Intel: Нажмите и удерживайте клавишу Shift во время запуска.  
- Компьютеры Mac с процессорами Apple M: Выключите компьютер Mac, подождите 10 секунд, а затем нажмите и удерживайте кнопку питания на компьютере Mac, пока не появится окно Параметров запуска. Выберите загрузочный диск вашего компьютера Mac. Затем нажмите и удерживайте клавишу Shift, нажмите Продолжить в безопасном режиме и отпустите клавишу Shift. 

Когда появится экран входа в систему, вы увидите безопасную загрузку красным цветом в правом верхнем углу.   Если ваш Mac настроен так, что вам не нужно входить в систему, продолжайте нажимать клавишу Shift во время запуска, пока Finder не загрузится полностью. Если вы все-таки войдете в систему на своем Mac, введите свой пароль как обычно, но прежде чем нажать кнопку Входа, снова нажмите клавишу Shift и удерживайте ее нажатой до полной загрузки Finder.

# Техническая информация
Загрузка в безопасном режиме выполняет три действия, помогающие вам в устранении неполадок:  
- Принудительно проверяет каталог начального (загрузочного) тома.  
- Загружает только необходимые расширения ядра (некоторые элементы в /System/Library/Extensions).  
- Он запускает только основные элементы запуска, установленные Apple (некоторые элементы в /Library/StartupItems and /System/Library/StartupItems). Элементы запуска в папках библиотеки отличаются от элементов входа в панель входа в системных настройках. Элементы запуска запускаются во время загрузки еще до появления окна входа в систему; элементы входа запускаются только после входа в свою учетную запись пользователя.  Взятые вместе, эти изменения часто устраняют проблемы, вызванные повреждением программного обеспечения или каталога на стартовом томе.

# Шаг 3: Отключите PRAM / NVRAM  
# Внимание 
Если на вашем компьютере Mac установлен процессор Apple Silicon, проигнорируйте этот шаг. Он не применим к вашему компьютеру Mac. 

Иногда параметры оперативной памяти вашего Mac (PRAM) или энергонезависимой оперативной памяти (NVRAM) скремблируются, и их необходимо сбросить. Оба этих небольших фрагмента памяти не стираются и не забываются при выключении компьютера Mac. Они отслеживают такие вещи, как следующее:  
- Настройка часового пояса
- Запуск-выбор громкости
- Громкость динамика
- Любая свежая информация о сбоях в ядре
- Настройка региона DVD  
Чтобы сбросить (процесс, часто называемый перезапуском) вашу PRAM / NVRAM, перезагрузите компьютер Mac и нажимайте ⌘ + Option + P + R, пока ваш Mac не перезагрузится сам. Это похоже на сбой. Вы можете видеть указатель вращающегося диска в течение минуты или двух, пока ваш Mac обдумывает это. Затем значок исчезнет, и ваш Mac снова зазвонит (если только у вашего Mac нет chimeless model) и перезагрузится. Некоторые опытные пользователи считают, что вам следует нажать несколько раз, позволив Mac нажать два, три или даже четыре раза, прежде чем отпускать клавиши и позволять процессу запуска продолжаться.
