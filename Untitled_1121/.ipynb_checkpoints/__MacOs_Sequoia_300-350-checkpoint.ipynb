# Сосредоточенность
В течение нескольких лет Apple совершенствовала функции macOS, чтобы помочь вам работать без помех. Функция "Не беспокоить", которая отключала все оповещения и уведомлении, была первой, но теперь была заменена функцией "Фокус", более широкой функцией, которая позволяет создавать различные типы настроек "оставь меня в покое". Теперь вы можете не только разрешить некоторым приложениям и контактам отвлекать ваше внимание, но и создавать фильтры фокусировки, которые управляют поведением ключевых приложений (таких как почта и сообщения) при включении фокуса.

Чтобы настроить фокусировку и управлять ею, выберите Системные настройки ⇒ Фокусировка, а затем работайте на панели Фокусировки. Здесь вы найдете несколько стандартных настроек, таких как "Не беспокоить", "Вождение" и "Сон", настроенных и готовых к работе. Вы можете активировать другие стандартные фокусы (такие как чтение, игры и осознанность), которые поначалу обычно застенчиво скрываются, и вы можете создавать свои собственные фокусы с нуля. Еще лучше то, что вы можете настроить каждый фокус таким образом, чтобы точно указать, какие приложения вы готовы признавать и от каких голливудских агентов вам было бы интересно принимать звонки во время медитации. 

Если на панели фокусировки не отображается нужный фокус, нажмите "Добавить фокус", чтобы открыть диалоговое окно "На чем вы хотите сфокусироваться?" . Если здесь отображается нужный вам фокус, щелкните его; macOS закроет диалоговое окно, добавит фокус в список на панели "Фокус" и отобразит экран настройки фокуса, где вы можете продолжить, как описано далее. Если в диалоговом окне не отображается нужный фокус, нажмите "Пользовательский". В открывшемся пользовательском диалоговом окне введите название фокуса, щелкните цветовой круг, выберите эмодзи или глиф, которые вы хотите использовать (точно так же, как при создании списка в "Напоминаниях"), а затем нажмите "ОК". Пользовательское диалоговое окно закрывается, новый фокус занимает свое место на панели фокусировки, и появляется экран настройки фокуса.

Если вы просто хотите настроить один из фокусов, который уже был на панели фокусировки, щелкните его там, и появится экран его настройки. Вы готовы настроить фокус. Используйте раздел Разрешить уведомления, чтобы указать, какие люди и приложения (если таковые имеются) могут уведомлять вас, пока включен фокус. 

Чтобы добавить человека или группу, нажмите кнопку "Разрешенные люди" и перейдите в диалоговое окно "Разрешенные люди". Во всплывающем меню вверху выберите пункт "Разрешить некоторым людям" или пункт "Заставить замолчать некоторых людей". Затем нажмите Добавить (+) и добавьте людей в список. Если вы выбрали "Разрешить некоторым пользователям", вы можете открыть "Разрешить звонки из всплывающего меню" и выбрать "Все", "Только разрешенные пользователи", "Избранное", "Только контакты", "Коллеги" или "Друзья", в зависимости от обстоятельств. Вы также можете установить переключатель Разрешить повторные вызовы в положение Вкл. (синий), чтобы позволить кому-либо прервать фокус, позвонив дважды в течение 3 минут (что может сигнализировать о чрезвычайной ситуации). Если вы выбрали "Заставить замолчать некоторых пользователей", установите для параметра "Разрешить звонки от пользователей с замолчанием" значение "Вкл." (синий) или "Выкл." (белый) соответственно. Нажмите "Готово", когда закончите работу в диалоговом окне "Разрешенные пользователи".  

Вернувшись к экрану настройки фокуса, нажмите кнопку "Добавить расписание" в разделе "Настройка расписания", если вы хотите запускать этот фокус по расписанию, а не включать и отключать его вручную. 

В разделе "Фильтры фокусировки" нажмите "Добавить фильтр", если вы хотите добавить фильтр фокусировки. В диалоговом окне "Фильтры приложений" выберите приложение, на которое вы хотите повлиять, например Safari. В следующем диалоговом окне, которое варьируется в зависимости от приложения, выберите подходящие параметры для фокусировки. Например, в Safari вы можете выделить определенную группу вкладок или профиль Safari в качестве основного объекта; для группы вкладок вы можете решить, разрешать ли открывать ссылки, ведущие за пределы этой группы вкладок. Нажмите кнопку Добавить, чтобы добавить фильтр фокусировки на экран настройки фокусировки.

# Отслеживание действий с помощью Screen Time  
Всем нам не помешает отдохнуть от наших цифровых устройств, и решение Apple - Screen Time. Хотя Screen Time включает в себя родительский контроль, который раньше отображался на панели системных настроек родительского контроля в более старых версиях macOS, Screen Time также позволяет вам контролировать свое собственное поведение. 

Если у вас iPhone или iPad, вы, вероятно, знакомы с программой Screen Time, которая дает представление о том, как вы проводите время на своем iPhone или iPad, в том числе о том, какими приложениями вы пользуетесь и веб-сайты посещаете и как долго. На Mac экранное время отображается в виде панели системных настроек, поэтому выберите Системные настройки ⇒ Экранное время, чтобы начать.  

Затем перейдите в поле Ограничить использование и нажмите кнопку "Активность приложения и веб-сайта". В диалоговом окне "Отслеживать время, проведенное за экраном?" нажмите кнопку "Включить активность в приложении и на веб-сайте". 

В нижней части панели "Экранное время" установите переключатель "Общий доступ на всех устройствах" в положение "Вкл." (синий), если вы хотите, чтобы "Экранное время" сообщало данные для каждого iPhone, iPad и Mac, который выполняет вход в эту учетную запись iCloud. Обычно это то, что вам нужно; нет смысла внимательно следить за использованием Mac вашим подростком, позволяя ему запускать riot на своем iPad.

- Активность приложений и веб-сайтов: отображает подробную информацию о приложениях, которыми пользовался пользователь, и о том, как долго они ими пользовались.  
- Уведомления: отображает количество уведомлений, полученных пользователем за этот день, и время их получения.  
- Пикапы: показывает, сколько раз пользователь брал в руки свои устройства.

- Ограничения приложений: устанавливает временные ограничения для приложений и таких категорий приложений, как социальные сети, игры и "Производительность и финансы".  
- Всегда разрешено: указывает контакты и приложения, доступные пользователю во время простоя.  
- Расстояние до экрана: Установите переключатель Расстояния до экрана в положение Вкл. (синий), чтобы функция расстояния до экрана предупреждала пользователя, когда он подносит устройство Face ID iPhone или iPad слишком близко к своему лицу в течение длительного времени.  
- Ограничения связи: указывает, кому разрешено связываться с пользователем по телефону, FaceTime или iMessage во время просмотра и простоя.  
- Безопасность общения: Установите переключатель безопасности общения в положение ВКЛ. (синий), чтобы в сообщениях предпринимались попытки обнаружить фотографии обнаженной натуры до того, как ваш ребенок отправит их или просмотрит те, которые он получил.  
- Содержимое и конфиденциальность: заменяет опции родительского контроля в более ранних версиях macOS. Вы можете включать или отключать содержимое по типу и применять ограничения к интернет-магазину Apple и определенным приложениям. Вы также можете разрешить или запретить изменение пароля, учетной записи и другие параметры, для которых вы можете захотеть ограничить изменения.

# Поиск вашего текущего местоположения


Если вам интересно, где вы находитесь, Карты помогут вам разобраться. Запустите Maps из Dock или панели запуска, а затем щелкните значок текущего местоположения, который представляет собой маленькую стрелку (показана на полях) на панели инструментов в верхней части окна. Синяя точка указывает на ваше местоположение.

# Запомни
Если вы нажмете или перетащите карту, ваш Mac продолжит обновлять ваше местоположение, но не изменит центр синей точки, поэтому синюю точку можно прокрутить (или увеличить) за пределы окна. Если это произойдет, щелкните значок текущего местоположения еще раз, чтобы снова отцентрировать карту в соответствии с вашим текущим местоположением.

Эта функция и многие другие функции карт зависят от активного подключения к Интернету и наличия включенных служб определения местоположения. При первом запуске Maps macOS отображает диалоговое окно “Карты”, которое вы хотели бы использовать в вашем текущем местоположении, чтобы предложить вам включить службы определения местоположения; нажмите кнопку "Разрешить", чтобы сделать это. После этого вы можете проверить информацию о службах определения местоположения следующим образом: нажмите "Системные настройки" в Dock, нажмите "Конфиденциальность и безопасность" на боковой панели, нажмите "Службы определения местоположения", чтобы отобразить экран "Службы определения местоположения", а затем установите переключатель "Службы определения местоположения" в положение Вкл. (синий). Также установите переключатель в строке Карт в положение Вкл. (синий). 

Если в "Картах" отображается диалоговое окно "Получать уведомления, когда друзья делятся своими ETA", нажмите "Включить уведомления", если вы хотите видеть уведомления об их расчетном времени прибытия и обновлениях фотографий и отчетов. В противном случае нажмите "Не сейчас". Вы можете настроить уведомления Карт позже на панели уведомлений в Системных настройках.

# Экономия времени с библиотекой, путеводителями и новостями  
Приложение "Карты" предлагает три инструмента на боковой панели, которые избавят вас от необходимости вводить одни и те же местоположения снова и снова.  
# Библиотека 
# Новое 
Раньше на картах были закладки для хранения местоположений, которые вы хотели легко посещать. Затем закладки сменились на избранное. Теперь в macOS Sequoia функция библиотеки заменяет избранное. 

Чтобы добавить местоположение в библиотеку, щелкните его значок на карте, а затем нажмите кнопку + (Добавить) в правом верхнем углу карточки. Значок + меняется на флажок, указывающий на то, что Maps добавила местоположение (и предотвращающий повторные попытки добавить его). 

Вы также можете начать добавлять местоположение с боковой панели. Наведите указатель мыши на заголовок Библиотеки на боковой панели, а затем щелкните появившийся знак +. В открывшемся диалоговом окне "Добавить pin-код" либо щелкните одно из местоположений в списке предложений Siri, либо введите поисковый запрос в текстовое поле "Карты поиска" и укажите местоположение вручную. 

После добавления местоположения в библиотеку вы можете вернуться к нему в любое время, щелкнув по нему в разделе Библиотека на боковой панели. Если боковая панель скрыта, нажмите кнопку Переключения боковой панели, чтобы отобразить ее; если раздел библиотеки свернут, разверните его, наведя на него указатель и нажав появившуюся кнопку >. 

# Лампочка 
Первые местоположения, которые вы должны добавить в библиотеку, - это ваш домашний и рабочий адреса. Вы будете постоянно использовать эти адреса для работы с картами, поэтому добавьте их сейчас, чтобы не вводить их снова и снова. Библиотека предоставляет местоположение дома и рабочее местоположение вверху списка, пока вы их не настроите; просто нажмите ссылку Добавить под местоположением дома и используйте диалоговое окно Настройка дома, чтобы указать местоположение, затем повторите для рабочего местоположения. 

Когда вам больше не нужно местоположение в библиотеке, щелкните его кнопкой мыши или правой кнопкой мыши, а затем выберите Удалить в контекстном меню.

# Путеводители 
Путеводители - это подборки мест, которые вы можете создавать и делиться ими с другими. В нескольких крупных городах, таких как Сан-Франциско, Нью-Йорк, Лондон и Лос-Анджелес, есть путеводители, созданные "брендами, которым вы доверяете” (по крайней мере, так утверждает Apple). Самое замечательное в руководствах то, что они обновляются автоматически, поэтому, если вы делитесь ими с другими, их руководства будут обновляться при добавлении или удалении местоположений.

# Запомни 
Кнопка "Местоположения" появляется прямо над заголовком "Последние" на боковой панели. На кнопке "Местоположения" не отображается слово "Местоположения" — вместо этого отображается количество мест, справочников и маршрутов, например 12 мест, 6 справочников, 3 маршрута. При нажатии на кнопку "Местоположения" открывается окно библиотеки.  
Чтобы создать руководство, выполните следующие действия:  
1. Нажмите кнопку "Местоположения" на боковой панели, чтобы отобразить окно библиотеки.  
2. Нажмите кнопку "Руководства", чтобы отобразить окно "Мои руководства".  
3. Нажмите кнопку + (Добавить) в правом нижнем углу, чтобы открыть диалоговое окно "Новое руководство". 
4. Введите описательное название руководства.  
5. Щелкните значок фотографии, а затем выберите фотографию в открывшемся окне. Переместите и масштабируйте фотографию в диалоговом окне "Выбор фотографии", а затем нажмите кнопку "Использовать".  
6. Вернитесь в диалоговое окно "Новое руководство" и нажмите кнопку "Создать". Maps создает руководство и добавляет его в окно "Мои руководства".  
7. Щелкните новое руководство, чтобы отобразить его окно.  
8. Нажмите "Добавить место", чтобы отобразить диалоговое окно "Сохранить в мои руководства".  
9. Добавьте места с помощью поиска или из списка недавно просмотренных.  
10. Нажмите кнопку "Готово", чтобы закрыть диалоговое окно "Сохранить в мои руководства".

# Создание пользовательских маршрутов  
# Новое 
Отличной новой функцией Maps в macOS Sequoia является создание пользовательских маршрутов, которые проходят именно туда, куда вы хотите. Чтобы создать пользовательский маршрут, выполните следующие действия:  
1. Перейдите на карте к местности для маршрута, например к национальному парку.  Вы можете выполнить поиск национального парка, тропы или начала маршрута по названию. Оказавшись в нужном месте, увеличьте или уменьшите масштаб по мере необходимости.
 
2. Щелкните местоположение, например, начало тропы, чтобы отобразить его карту.  Это местоположение будет началом создаваемого вами маршрута. 
3. Нажмите кнопку "Создать пользовательский маршрут", чтобы отобразить окно "Создать пользовательский маршрут". 
4. Щелкните поочередно каждую поворотную точку маршрута.

Продольный профиль в нижней части окна "Создание пользовательского маршрута" показывает высоты маршрута, а также уклоны и места их возникновения.

Чтобы удалить точку, щелкните ее, а затем нажмите кнопку Удалить точку. Это работает для начальной и конечной точек, а также для промежуточных точек.

# Добавление и удаление акций, фондов и индексов  
Ваши шансы стать владельцем группы акций, фондов и индексов по умолчанию невелики, поэтому вы, скорее всего, захотите добавить свои собственные акции, фонды или индексы и удалить любые или все акции по умолчанию. Вот как добавить акцию, фонд или индекс:

2. Выполните одно из следующих действий: 
- Щелкните значок "Добавить в список наблюдения", маленький знак "+" в круге слева от названия элемента.

3. Когда вы закончите добавлять элементы, нажмите кнопку Готово (X) справа от поля поиска, чтобы вернуться в свой список наблюдения.

В качестве альтернативы щелкните элемент кнопкой мыши или правой кнопкой мыши и выберите "Удалить из списка наблюдения" в контекстном меню. Или вы можете щелкнуть элемент, а затем нажать значок "Удалить из списка наблюдения" - флажок в правом верхнем углу окна "Акции".

# Подключение к Интернету 
Прежде чем вы сможете пользоваться Интернетом (или выходить в Интернет), вам необходимо подключиться к Интернету. Если вы обычный домашний пользователь, для выхода в Интернет вам понадобятся три вещи: 

- Подключение к Интернету, такое как кабельное подключение, подключение по цифровой абонентской линии (DSL), оптоволоконное соединение или услуга спутникового Интернета. Это соединение обычно проявляется в виде физического устройства, которое устанавливает подключение к Интернету, поддерживает его и предоставляет общий доступ к вашим устройствам через локальную сеть. С технической точки зрения, это физическое устройство обычно представляет собой маршрутизатор и коммутатор в сочетании, и часто также является точкой беспроводного доступа. В этом разделе термин интернет-маршрутизатор используется для обозначения этого типа устройства

Локальная сеть, через которую ваш интернет-маршрутизатор предоставляет общее подключение к Интернету вашему компьютеру Mac и другим устройствам вашей семьи, может быть проводной сетью, беспроводной сетью или и тем, и другим. Почти все проводные сети используют стандарты семейства Ethernet. Почти все беспроводные сети используют семейство стандартов Wi-Fi.

# Запомни

Настольные компьютеры Mac без мониторов — Mac mini, мощный Mac Studio и Mac Pro — оснащены встроенными портами Ethernet. Mac Pro не довольствуется одним портом Ethernet, он может похвастаться двумя.

# Выбор интернет-провайдера (ISP) 
Возможно, вам придется выбрать компанию, которая предоставит вам доступ в Интернет: интернет-провайдера, или сокращенно ISP. Цены и услуги, предлагаемые интернет-провайдерами, меняются, часто в зависимости от минуты. При выборе интернет-провайдера учитывайте следующее.: 

- Если ваше подключение осуществляется от кабельной или телефонной компании, вероятно, именно эта компания является вашим интернет-провайдером. По сути, выбор интернет-провайдера в значительной степени делается за вас, когда вы выбираете услугу кабельного, оптоволоконного или DSL-соединения. 

- Стоимость широкополосного доступа в Интернет начинается примерно с 40 долларов в месяц. Если ваш поставщик услуг запрашивает значительно больше, выясните, почему.

Если вы считаете, что платите слишком много за интернет-услуги или вам не нравится ваш текущий провайдер, сделайте домашнее задание и определите, какие другие варианты доступны в вашем районе. 

Вообще говоря, интернет-провайдеры становятся все лучше в настройке интернет-соединений, чтобы они работали надежно и не требовали от вас постоянного прослушивания отделов технической поддержки интернет-провайдеров. Но прежде чем ваше интернет-соединение перестанет работать, убедитесь, что вы знаете, как связаться с отделом технической поддержки вашего провайдера по телефону или другим способом, не связанным с вашим интернет-подключением.
 
# Лампочка
Как вы знаете, решение компьютерных проблем номер один — это отключение питания - выключение этой проклятой штуковины, непринужденная ругань, а затем включение снова. Ну, угадайте, что: Решением номер один проблем с интернет-маршрутизатором также является отключение питания. (Вы даже можете использовать те же самые ругательства.)

# Подключение компьютера Mac к сети через Wi-Fi  
После того, как ваша сеть будет запущена, вы можете подключить к ней компьютер Mac через Wi-Fi либо через Центр управления, либо через панель Wi-Fi в системных настройках. Давайте начнем с Центра управления, потому что это метод, который вы, вероятно, захотите использовать, когда будете выходить из дома. 

1. Щелкните значок Центра управления (показан на полях) в строке меню, чтобы открыть Центр управления (показан слева на рисунке 13-1). 
2. Нажмите кнопку Wi-Fi, чтобы отобразить панель Wi-Fi (показана справа на рисунке 13-1).  Убедитесь, что переключатель Wi-Fi вверху установлен в положение On (синий).  
3. Щелкните название сети в списке Другие сети.  Центр управления закроется. Откроется диалоговое окно с запросом пароля сети Wi-Fi. 
4. Введите пароль в поле Пароль.  Если вы хотите видеть вводимые символы, установите флажок Показывать пароль. Отображение символов может стать спасением при вводе убийственно сложного пароля.  
5. Нажмите кнопку Присоединиться.
