# Введение 
Похоже, вы сделали три правильных выбора: у вас есть Mac, macOS Sequoia (она же macOS версии 15) и эта книга. Если вы новичок в Mac, вы готовы начать получать удовольствие от работы с самой совершенной операционной системой на планете. Если вы уже давно пользуетесь своим Mac и macOS, вы готовы начать получать от их использования еще больше удовольствия. Конечно, эта книга посвящена компьютерам, но она не из тех скучных книг, которые лучше всего подходят для того, чтобы служить подставкой для дверей; с ней знакомство со всеми тонкостями macOS Sequoia становится легким и даже увлекательным занятием!

# Об этой книге
“macOS Sequoia для чайников” - это последняя редакция бестселлера легендарного специалиста по Mac Боба "доктора Мака" Левитуса, технологического обозревателя the Houston Chronicle. Книга была полностью обновлена для macOS Sequoia, чтобы охватить все новейшие функции, такие как Apple Intelligence, инструменты для письма и зеркальное отображение iPhone, а также все остальное, что изменилось. Это издание сочетает в себе все старые, знакомые функции десятков предыдущих изданий с самой последней информацией о компьютерах Mac и macOS Sequoia.

Why write a For Dummies book about macOS Sequoia? Well, Sequoia
is a big, somewhat complicated, personal-computer operating system. So
macOS Sequoia For Dummies, a not-so-big, not-too-complicated book,
shows you what Sequoia is all about without boring you to tears or
poking you with sharp objects.



This book is chock-full of information and advice, explaining everything
you need to know about macOS Sequoia in easy-to-understand language
— along with time-saving tips, tricks, techniques, and step-by-step
instructions. I feel confident that you’ll find what you need to know
about using macOS Sequoia in this book.


Still, a book this size can’t explain everything you might want to know
about an operating system the size of macOS Sequoia and the apps that
run on it. If you’re looking for information on topics such as Microsoft
Office, Apple’s lifestyle and productivity apps (such as iMovie,
GarageBand, Numbers, and Pages), or programming, I recommend you
look at other books published by Wiley. The Wiley website
(https://www.wiley.com/) is the best place to start.


Within this book, you may note that some web addresses break across
two lines of text. If you’re reading this book in print and want to visit
one of these webpages, simply key in the web address exactly as it’s
noted in the text, as though the line break doesn’t exist. If you’re reading
it as an ebook, you’ve got it easy: Just click or tap the web address to be
taken directly to the webpage.

# Foolish Assumptions
Although I know what happens when you assume, I’ve made a few
assumptions anyway. I assume that you, gentle reader, know nothing
about using macOS — beyond knowing what a Mac is, that you want to
use macOS, that you want to understand macOS without having to digest
an incomprehensible technical manual, and that you made the right
choice by selecting this particular book. So I do my best to explain each
new concept fully.
