# Создание PDF-файлов без запуска приложения
Вы уже видели три быстрых действия: поворот, разметка и обрезка. В этом разделе показано еще одно быстрое действие "Создать PDF", которое создает PDF-файлы. В большинстве приложений вы можете создать PDF-файл из диалогового окна Печати (см. Главу 19). Но разве не было бы удобно, если бы вы могли создавать PDF-файлы, не запуская приложение и не выбирая Печать? В macOS Sequoia это можно сделать, но только для файлов изображений (таких как файлы JPEG, TIFF и PNG), а не для файлов, таких как документы Microsoft Word или рабочие книги Microsoft Excel. Просто воспользуйтесь быстрым действием "Создать PDF", которое вы найдете в окнах быстрого просмотра, на панелях предварительного просмотра окон Finder и в контекстном меню (щелчок кнопкой мыши или правой кнопкой мыши). Вы можете определить, можно ли преобразовать файл, поскольку команда "Создать PDF" или значок появляются только тогда, когда выбраны подходящие файлы, как показано на рис. 5-10.

# Съемка фотографий и фильмов на экране
Если вы давно пользуетесь компьютером Mac, вы, вероятно, знаете, что можно быстро сделать снимок того, что находится на вашем экране, с помощью утилиты захвата экрана:
- Захватите весь экран. Нажмите ⌘ + Shift + 3.
- Захватите часть экрана. Нажмите ⌘ + Shift + 4. Указатель изменится на значок перекрестия. Переместите перекрестие в то место, где вы хотите начать съемку, щелкните, а затем перетащите по диагонали в противоположный угол. Когда вы отпускаете кнопку мыши, функция захвата экрана захватывает выбранную область.
- Захват окна. Нажмите ⌘ + Shift + 4. Указатель изменится на значок перекрестия. Нажмите пробел, и перекрестие исчезнет, сменившись значком камеры. Переместите этот значок над окном, которое вы хотите запечатлеть, так, чтобы macOS выделила окно, а затем щелкните. Этот метод работает и для меню; просто откройте меню, прежде чем нажимать ⌘ + Shift + 4.

Эти сочетания клавиш по-прежнему работают в Sequoia. Когда вы делаете снимок экрана или его части с их помощью, в правом нижнем углу экрана появляется миниатюра снимка экрана. Если вы ничего не предпримете, миниатюра исчезнет примерно через 5 секунд, а снимок экрана будет сохранен. Чтобы просмотреть дополнительные параметры, щелкните правой кнопкой мыши миниатюру и выберите опцию в контекстном меню, как показано на рис. 5-11. Вы можете нажать "Разметка" в контекстном меню, например, чтобы открыть снимок экрана и начать украшать его разметкой.

# Знакомство с системными настройками
Системные настройки - это ваш основной инструмент для настройки macOS и настройки ее работы так, как вы предпочитаете. Начните с открытия окна системных настроек и освоения навигации по его экранам. Выполните следующие действия:
1. Откройте окно системных настроек, показанное на рисунке 6-1. Вы можете открыть системные настройки по крайней мере четырьмя способами:
- Выберите ⇒ Системные настройки.

- Щелкните значок системных настроек на вашей док-станции. Если "Системные настройки" не запущены, вы можете сразу перейти к категории "Настройки": щелкните правой кнопкой мыши значок "Системные настройки", чтобы отобразить контекстное меню со списком категорий "Настройки", а затем выберите нужную категорию.
- Откройте панель запуска, а затем щелкните значок Системных настроек.

Как обычно, боковая панель в окне системных настроек позволяет вам перемещаться по многочисленным категориям настроек приложения. На правой панели отображаются настройки в категории, которую вы щелкаете на боковой панели. Так, например, когда вы щелкаете значок "Общие" на боковой панели, на правой панели отображается категория "Общие", которую (для простоты) мы будем называть панелью "Общие".
Когда вы закончите работу в системных настройках, закройте приложение, выбрав "Системные настройки" ⇒ "Выйти из системных настроек" (или нажав ⌘ + Q) или нажав красную кнопку "Закрыть".

# Настройка экранной заставки или просто выключение дисплея
Если хотите, вы можете настроить запуск экранной заставки на вашем Mac, когда вы некоторое время не используете ее. Экранная заставка может быть интересной визуально, но не чувствуйте себя обязанным ее использовать. Простой перевод дисплея в спящий режим экономит немного энергии, а перевод компьютера Mac в спящий режим экономит еще больше. Кроме того, проводите ли вы время за просмотром компьютера Mac, когда им не пользуетесь?
# Предупреждаю
macOS включает в себя множество экранных заставок, а App Store предлагает различные приложения для создания заставок. Если хотите, вы можете дополнить их заставками сторонних производителей, но будьте осторожны при загрузке бесплатных заставок, потому что злоумышленники иногда используют их в качестве переносчика вредоносного ПО. 

Чтобы настроить экранную заставку, выполните следующие действия:

1. Откройте "Системные настройки", а затем щелкните пункт "Экранная заставка" в боковая панель для отображения панели экранной заставки (показано на рис. 6-4).
2. Нажмите на экранную заставку, которую вы хотите попробовать.
3. На панели вверху установите переключатель "Показывать на всех пространствах" в положение "Вкл." (синий), если на вашем Mac установлено несколько дисплеев и вы хотите отобразить экранную заставку на всех из них.
4. Если вы выбрали экранную заставку из ландшафта, выберите "Городской пейзаж".,
В категории "Подводные, земные или перемещаемые антенны" установите переключатель "Показывать в качестве обоев" в положение "Вкл." (синий), если вы хотите использовать экранную заставку также в качестве обоев.
5. Если вы выбрали экранную заставку из категории "Другое", посмотрите, появляется ли кнопка "Параметры" в верхней части панели "Экранная заставка".

Если появится кнопка "Параметры", щелкните ее, чтобы открыть диалоговое окно "Параметры", выберите "Параметры экранной заставки", а затем нажмите кнопку "ОК". Например, если вы выберете экранную заставку "Фотографии", диалоговое окно параметров позволит вам выбрать источник фотографий. Если вы выберете экранную заставку Flurry, диалоговое окно Параметров позволит вам выбрать цветовой цикл, указать количество потоков, а также установить их толщину и скорость.

6. Щелкните элемент "Экран блокировки" на боковой панели приложения "Системные настройки". Появится панель "Экран блокировки" (см. Рисунок 6-5).
7. Щелкните всплывающее меню "Пуск экранной заставки при неактивном", а затем выберите "Задержка".
Вы можете выбирать от 1 минуты до 3 часов или никогда. Вы бы выбрали "Никогда", если бы хотели запретить запуск экранной заставки.

# Лампочка
На MacBook вы можете выбрать отдельные настройки экрана блокировки для работы от аккумулятора и для работы от адаптера питания.

8. Щелкните всплывающее меню "Отключить отображение при неактивном режиме", а затем укажите задержку.
Ваши варианты те же, что и на предыдущем шаге. Вы бы выбрали "Никогда", если бы хотели, чтобы экранная заставка запускалась вечно. На MacBook настройте как параметр "Выключать дисплей при работе от батареи при неактивном режиме", так и параметр "Выключать дисплей при работе от адаптера питания при неактивном режиме".

9. Нажмите кнопку "Запросить пароль" После запуска экранной заставки или
Отображение отключенного всплывающего меню, а затем выберите нужную задержку. В целях безопасности выберите параметр "Немедленно". Если вы обнаружите, что экранная заставка срабатывает во время использования компьютера Mac, но она неактивна, выберите параметр "Через 5 секунд". Это позволит вам отменить экранную заставку без необходимости вводить пароль. Вы также можете выбрать более длительные настройки в диапазоне от "Через 1 минуту" до "Через 8 часов", но эти настройки могут позволить кому-либо другому получить доступ к вашему Mac без необходимости аутентификации. Параметр "Никогда" редко является хорошим выбором.
10. Когда вы закончите настройку экранной заставки, закройте Системные настройки.

- Предпочитайте вкладки при открытии документов: ваш выбор в этом всплывающем меню определяет, использует ли macOS окна с вкладками. Попробуйте все настройки — Никогда, В полноэкранном режиме или всегда — и выберите ту, которая вам больше нравится.
- Попросите сохранить изменения при закрытии документов: macOS может сохранять версии ваших документов автоматически и без каких-либо действий с вашей стороны. # Where Did That File Go?


Given the number of files most of us create, it’s all too easy to lose track
of particular files if you don’t keep your filesystem organized.
macOS provides two powerful tools that can help you track down files:

- Spotlight: This search feature maintains an index of files in your
Mac’s filesystem, enabling you to locate files by their names or
contents. Such searching is handy unless you have files with the
same name stored in different folders — in which case you’ll need to
work out which file is the one you want.


- Recents folder: This smart folder displays aliases (shortcuts) to all
the files you’ve used recently. To see Recents, open a Finder window
and click the Recents item in the Favorites section of the sidebar. Recents is a good way to find files you’ve used recently, unless the
sheer number of files in the Recents list overwhelms you.
