# Taking Actions from Quick Look


The Quick Look command displays the contents of the selected file in a
floating window, letting you see what’s in a file without opening it and
without launching an app. Quick Look is great for when you want to
take a quick peek at the contents of a file.
To use Quick Look, select a file or folder and then do any of the
following:


- Press the spacebar.


- Press ⌘ +Y.


- Choose File ⇒ Quick Look.


- Control-click or right-click the file’s icon, and choose Quick Look
from its contextual menu.


- Choose Quick Look from the Action menu on the toolbar.


Quick Look is especially useful for a folder full of images, such as the
one shown in Figure 9-3.


# Лампочка

While the Quick Look window is open, you can select different
icons in the Finder window — use the arrow keys to select the next
or previous icon — and see their contents in the Quick Look
window.


Although Quick Look works with many types of files — Microsoft
Office, Apple iWork, plain text, PDF, TIFF, GIF, JPEG, PNG, and most
types of audio and video files — it doesn’t work with all files. You’ll
know it doesn’t work if Quick Look shows you a big document, folder,
or app icon instead of the contents of that file.


If you select multiple items before you invoke Quick Look, as in Figure
9-3, the Previous icon (<) and the Next icon (>) appear at the top of the
Quick Look window near the left side, enabling you to navigate to the
previous item or the next item; you can also press the left-arrow key to
display the previous item or the right-arrow key to display the next item.
To the right of the current item’s name is the Index Sheet icon, which
shows four small squares. You can click Index Sheet to view all selected
items as an index sheet of thumbnails.

# Share and share alike with the Share menu

macOS makes sharing items easy. Clicking the Share icon (shown in
the margin) opens the Share menu, which gives you seven or more
options (depending on the type of file you selected):


- Mail: Launches the Mail app and attaches the selected file to a blank
message, ready for you to address and send.


- Messages: Launches the Messages app and puts the selected file in
an outgoing message, ready for you to address and send. You get to
know the Messages app in Chapter 15.


- AirDrop: Sends the selected file to users of iPhones, iPads, or other
Macs via Bluetooth and Wi-Fi Direct.



- Notes: Sends the selected file to the Notes app, where you can add it
to an existing note or create a note for it. (You meet the Notes app in
Chapter 10.)


- Add to Photos: Adds the selected item to the All Photos album in
the Photos app.


- Freeform: Opens the selected item in the Freeform whiteboarding
and collaboration app.


- Reminders: Creates a new reminder in the Reminders app (see
Chapter 10).


- More: Lets you add other services (such as Vimeo or LinkedIn) and
apps (such as Photos or Aperture) to your Share menu. To manage
these extensions, choose Edit Extensions from the Share menu.
Alternatively, choose System Settings ⇒ General, click the Login
Items & Extensions button, and then click the i-in-a-circle icon to the
right of Sharing in the Extensions pane. In the Select Extensions for
Sharing with Others dialog that opens, select the switch for each
extension you want to enable to On (blue), and then click Done.


If the file or folder you select is on your iCloud Drive, you’ll see an
additional option that lets you share the file or folder with others. You
specify who can see it, who can modify it, and who can share it with
others before sending them an invitation via Mail, Messages, Link, or
AirDrop.

# Finding Files and Folders Fast with Spotlight

When you lose track of a file or folder, turn to Spotlight, the powerful
search technology built into macOS. Spotlight indexes your files and
their contents while your Mac is idle, enabling you to search for:


- Files
- Folders
- Text inside documents
- Files and folders by their metadata (creation date, modification date, kind, size, and so on)


Spotlight quickly finds what you’re looking for and then organizes its
results logically.


Spotlight is both a technology and a feature. The technology is pervasive
throughout macOS and is the underlying power behind the Search boxes
in many Apple apps and utilities, such as Mail, Contacts, System
Preferences, and Finder. You can also use it by clicking the Spotlight
menu — the little magnifying glass on your menu bar. Finally, you can
reuse Spotlight searches in the future by turning them into smart folders
(see Chapter 8).


The following subsections explain how to search using the Search box
on the Finder toolbar and using the main Spotlight menu.

# Blasting Off with Mission Control


If you use your Mac extensively, chances are that you’ll use lots of apps
at the same time and that some of those apps will have multiple windows
open. Some apps may be full-screen. You may have multiple desktops as
well.


To navigate among your open windows, apps, and desktops, you can use
Mission Control, discussed here. (You can also use Stage Manager,
discussed in the upcoming section, “All the World’s a Stage … and
You’re the Manager.”)


Mission Control’s eponymous feature shows you big thumbnails of your
open windows, apps, full-screen apps, and desktops so you can organize
them or navigate swiftly to the one you want. Mission Control can also
display all the open windows for the active app, which is great for
navigating inside that app; this feature is called App Exposé. Less
excitingly but still usefully, Mission Control can display the desktop,
enabling you to get at the items you’ve parked on it; this feature is called
simply Show Desktop.


Before you start using Mission Control, you might want to configure it.
Open the System Settings app, click the Desktop & Dock item in the
sidebar, and then scroll down to the Mission Control section at the
bottom of the Desktop & Dock pane.


# Setting Mission Control parameters


The Mission Control section of the Desktop & Dock pane contains four
switches: Automatically Rearrange Spaces Based on Most Recent Use;
When Switching to an Application, Switch to a Space with Open
Windows for the Application; Group Windows by Application; and
Displays Have Separate Spaces. These long-winded switch names will
all make sense when you start using Mission Control. Experiment with
the settings, turning them on and off, to see which way you prefer them.
Below these four switches are two buttons: Shortcuts (discussed here)
and Hot Corners (discussed in the next subsection, “Setting actions for
hot corners”).

# Setting actions for hot corners

In the bottom-right corner of the Mission Control section of the Desktop
& Dock pane in System Settings is the Hot Corners button. Click this
button to open a dialog that lets you designate any or all of the corners of
your screen as hotspots to trigger Mission Control, App Exposé, Show
Desktop, Notification Center, Launchpad, Quick Note, Screen Saver,
Display Sleep, or Lock Screen. Click the menu for a corner, and then
select the feature you want to associate with that corner. Then, whenever
you move your pointer to that corner and leave it there for a second or
two, the feature executes.


# Предупреждение

 Hot corners can be great, but if you tend to park the mouse
pointer in a particular corner of the screen as you work with the
keyboard, you may find you trigger the hot corner unintentionally.
If this happens, either don’t set a hot corner for that corner or add
one or more modifier keys to the hot corner, as explained in the
next tip.


Mission Control is enabled by default, but you can disable any or all of
its features by turning off its trigger: Just choose the minus sign from a
pop-up menu instead of a keyboard shortcut.

# Удаление календаря  
Чтобы удалить календарь, выберите его в списке и нажмите ⌘ + Удалить. Откроется диалоговое окно подтверждения, в котором вас спросят, уверены ли вы.

Если вы выйдете из Calendar, не отменив удаление календаря, все содержимое этого календаря (или календарей) будет удалено навсегда (если только у вас нет Time Machine или другой резервной копии, описанной в главе 21).

# Оставаться организованным с помощью напоминаний

# Обмен списками и назначение напоминаний
Вы можете назначить напоминание только тому, с кем вы поделились списком. Чтобы поделиться списком, щелкните список кнопкой мыши или правой кнопкой мыши и выберите "Поделиться списком" в контекстном меню или щелкните список, чтобы выбрать его, а затем выберите "Файл ⇒ Поделиться списком". Появится всплывающее окно с такими параметрами, как Почта, Сообщения, заметки и ссылка для приглашения. Вверху находится индикатор общего доступа, в котором говорится, что либо приглашенные вами люди могут добавлять других, либо доступ имеют только приглашенные вами люди. Если текущая настройка вас устраивает, оставьте ее, но если вы хотите ее изменить, нажмите кнопку >, чтобы установить флажок Разрешить приглашать других пользователей, а затем установите ее или снимите по мере необходимости. Выберите подходящий способ обмена, например, по почте, и отправьте сообщение. Получатель получает ссылку, которая добавляет список к его напоминаниям, синхронизируя изменения (почти) в режиме реального времени. После того как вы поделились списком, вы можете назначить его напоминания всем, с кем вы поделились этим списком. Чтобы назначить напоминание, щелкните напоминание кнопкой мыши или правой кнопкой мыши и выберите "Назначить ⇒ имя пользователя" в контекстном меню или щелкните напоминание, чтобы выбрать его, а затем выберите "Редактировать ⇒ Назначить ⇒ имя пользователя".

# Все, что вам нужно знать о Центре уведомлений
Центр уведомлений (см. Рис. 10-6) управляет оповещениями из любого приложения, поддерживающего протокол уведомлений Apple, и отображает их. Вы можете отобразить или скрыть Центр уведомлений, щелкнув значок часов в строке меню или проведя двумя пальцами влево от правого края трекпада.

# Настройка настроек уведомлений
Вы можете управлять тем, какие приложения отображают уведомления, с помощью панели уведомлений системных настроек. Выберите Системные настройки ⇒ Уведомления, чтобы отобразить главную панель уведомлений, а затем при необходимости перейдите к ее подпанелям.

Начните с настройки следующих четырех параметров в разделе "Центр уведомлений" в верхней части главной панели "Уведомления":
- Показывать предварительные просмотры: Откройте это всплывающее меню и выберите "Всегда", когда разблокировано, или "Никогда", по мере необходимости. "Всегда" нескромно и никогда не замалчивается, поэтому "Разблокировано" обычно является лучшим выбором. Это настройка по умолчанию для предварительного просмотра уведомлений. Вы можете выбрать другую настройку для предварительного просмотра любого приложения или функции.
- Разрешить уведомления, когда дисплей находится в режиме ожидания: установите этот переключатель в положение Вкл. (синий) или Выкл. (белый), по мере необходимости.
- Разрешить уведомления при блокировке экрана: установите этот переключатель в положение Вкл. (синий) или Выкл. (белый), как вам удобнее
- Разрешить уведомления при зеркальном отображении или совместном использовании дисплея: установите этот переключатель в положение Вкл. (синий) или Выкл. (белый). Обычно лучше выбрать "Выкл.", если вы не демонстрируете, как работать с уведомлениями.

Затем перейдите к разделу Уведомлений приложений на главной панели уведомлений, который содержит кнопку для каждого приложения или функции, для которых вы можете настроить уведомления. Нажмите кнопку, чтобы отобразить панель приложения или функции, например отобразить панель календаря, чтобы настроить уведомления для приложения "Календарь". Затем вы можете выбрать настройки следующим образом:
- Разрешить уведомления: Установите этот переключатель в положение Вкл. (синий), чтобы разрешить приложению отправлять уведомления. Установите переключатель в положение Выкл. (белый), чтобы отключить уведомления.
- Стиль оповещения: в поле выберите "Нет", "Баннеры" или "Оповещения". Оповещения остаются на экране до тех пор, пока вы их не отключите, поэтому они подходят для уведомлений, которые вы не можете позволить себе пропустить; баннеры появляются в правом верхнем углу на несколько секунд, прежде чем исчезнуть.
