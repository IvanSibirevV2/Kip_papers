# Технические подробности 
Большинство домашних интернет-маршрутизаторов настроены на использование технологии под названием Dynamic Host Configuration Protocol (DHCP), которая автоматически предоставляет информацию о подключении компьютерам в сети. Эта информация о подключении включает адрес интернет-протокола (IP-адрес), номер, который идентифицирует компьютер; маску подсети, номер, используемый для организации сети; и IP-адрес или имя интернет-маршрутизатора, который сообщает компьютеру, где найти выход в Интернет. Чтобы подключить компьютер Mac к сети Wi-Fi через панель Wi-Fi в приложении "Системные настройки", выполните следующие действия:

2. Щелкните Wi-Fi на боковой панели, чтобы отобразить панель Wi-Fi.   Убедитесь, что переключатель Wi-Fi вверху установлен в положение On (синий).

# Использование и настройка панели инструментов  
Панель инструментов Safari, расположенная в верхней части каждого окна Safari, состоит из узкого ряда значков и поля поиска или ввода названия веб-сайта. Значки в значительной степени соответствуют своим названиям. Слева направо они изображены следующим образом 

- Показать / скрыть боковую панель: Щелкните этот значок (показан на полях), чтобы увидеть избранное или список чтения на боковой панели; щелкните его еще раз, чтобы скрыть боковую панель. 

- Новая группа вкладок: Щелкните этот значок (показан на полях), чтобы создать новую пустую группу вкладок. И если вы еще не знаете, что такое группа вкладок, вы скоро узнаете. Этот значок появляется при отображении боковой панели. 

- Средство выбора группы вкладок: Щелкните этот значок (показан на полях), чтобы отобразить всплывающее меню, содержащее список групп вкладок, а также команду Создать пустую группу вкладок и Новую группу вкладок с этой командой. Этот значок появляется, когда боковая панель скрыта.  
- Назад / вперед: когда вы открываете страницу и переходите на вторую страницу (или третью, или четвертую), кнопка назад возвращает вас к ранее посещенным страницам. Помните, что вам нужно вернуться назад, прежде чем заработает значок "Вперед".  
- Найдите или введите название веб-сайта (также известное как поле адреса): в этом поле справа от кнопки "Показать боковую панель" вы вводите веб-адреса или URL-адреса (единые локаторы ресурсов), которые вы хотите посетить. Просто введите один — или вставьте, если хотите, — и нажмите "Вернуться", чтобы перейти на этот сайт.  
Справа от поля поиска или ввода названия веб-сайта находятся еще три значка:

# Использование паролей

# Импорт контактов из других приложений

Если ваши контакты находятся на другом компьютере Mac, iPhone или iPad или сохранены Google или Microsoft, импортировать контакты не потребуется. Просто добавьте соответствующую учетную запись на панели "Учетные записи Интернета" системных настроек (если вы ее еще не добавили), а затем включите синхронизацию контактов.  Если у вас есть контакты в другом приложении (например, FileMaker Pro или ACT), вы можете экспортировать контактные данные в файл, а затем импортировать их в Contacts. Контакты могут импортировать данные в пяти форматах файлов:  
- vCard: файлы vCard представляют собой виртуальные адресные карточки и используют расширение файла .vcf. vCard может содержать данные об одном контакте или нескольких контактах. Если приложение, содержащее ваши контактные данные, может экспортировать их в vCard, этот формат - ваш лучший выбор.  
- Архив: в ранних версиях OS X использовались архивные файлы с расширением .abbu. С тех пор OS X стала macOS, а Адресная книга стала Контактами, но формат архива сохраняется. Если у вас есть файл в этом формате, вы можете импортировать его в Contacts.  
- Формат обмена данными LDAP (LDIF): LDAP - это сокращение от Lightweight Directory Access Protocol. Возможно, у вас есть файл LDIF (с расширением файла .ldif), экспортированный из корпоративной адресной книги. 
- Значения, разделенные запятыми (CSV): в этом формате текстового файла поля разделены запятыми. Файлы CSV обычно имеют расширение .csv. Файлы CSV - это хороший способ экспортировать контактные данные из листа Microsoft Excel для импорта в приложение "Контакты". Вам также может потребоваться создать CSV-файл при экспорте из адресной книги, которая не поддерживает формат vCard.  
- Значения, разделенные табуляцией: в этом формате текстового файла поля разделены табуляциями. Файлы используют расширение файла .tsv или .txt. Используйте значения, разделенные табуляцией, для экспорта контактных данных из приложения, которое не поддерживает ни формат vCard, ни формат CSV. После экспорта контактных данных вернитесь в приложение "Контакты". Выберите "Файл" ⇒ "Импорт", выберите файл экспортированных данных в диалоговом окне "Открыть файл", а затем нажмите кнопку "Открыть". Появятся импортированные контакты, и вы можете начать с ними работать.

 Почта имеет встроенную фильтрацию нежелательной почты, которую можно включить или отключить в настройках почты. (Выберите Почта ⇒ Настройки и щелкните значок нежелательной почты на панели вкладок.) Если вы получаете нежелательную почту, выберите ее и щелкните значок нежелательной почты, чтобы настроить фильтр нежелательной почты. Если выбранное сообщение было помечено как нежелательная почта, значок изменится на "Не нежелательная почта". 
- Ответить: создает ответ только отправителю.  
- Ответить всем: создает ответ отправителю и всем, кому было отправлено исходное сообщение, за исключением получателей Bcc.  
- Пересылка: создает копию этого сообщения, которое вы можете отправить кому-либо, кроме отправителя или других получателей.  
- Отметить / Снять флажок: устанавливает флажок для выбранного сообщения или сообщений. Флаг имеет цвет, указанный на кнопке панели инструментов. Откройте всплывающее меню, чтобы выбрать другой цвет флага. Вы можете указать флаги в поиске и смарт-почтовых ящиках.  
- Отключение звука: отключает уведомления о новых сообщениях. Щелкните этот значок (или выберите Сообщение ⇒ Отключение звука, или нажмите Control + Shift + M), чтобы отключить уведомления о новых сообщениях в выбранной теме сообщений. 
- Переместить: перемещает выбранные сообщения в выбранную вами папку.  
- Поле поиска: находит слово или фразу в любом сообщении, хранящемся в почте. Когда вы начинаете вводить текст, появляется всплывающее меню, позволяющее сузить поиск до людей или тем, соответствующих вашей поисковой фразе. Вы можете нажимать кнопки на панели Избранного, чтобы ограничить поиск определенным почтовым ящиком.

# Лампочка
Если вы не видите панель Избранного, выберите Вид ⇒ Показать панель избранного или нажмите ⌘ + Option + Shift + H. Если вы не видите текстовых меток на кнопках панели инструментов, щелкните правой кнопкой мыши панель инструментов, а затем выберите Значок и текст в контекстном меню. 

# Лампочка 
Цифры рядом с кнопками почтового ящика на панели Избранного или панели почтового ящика указывают количество непрочитанных сообщений в этом почтовом ящике. Почта помечает сообщение как прочитанное после того, как вы показываете его в течение секунды или двух. Вы также можете пометить сообщение как прочитанное, щелкнув по нему кнопкой мыши или правой кнопкой мыши, а затем выбрав "Пометить как прочитанное" в контекстном меню. 

Поиск в почте работает так же, как и в Finder. Так, например, если вы хотите сохранить результаты поиска в виде смарт-почтового ящика (почтовая версия смарт-папки Finder), щелкните значок + под полем поиска, чтобы добавить критерии. 

Почта заполняет панель Избранного почтовыми ящиками, которые, как ожидается, вы будете часто использовать: Входящие, Важные персоны, Помеченные, Отправить позже, Черновики и Отправленные. Вы можете добавить свои собственные избранные почтовые ящики, перетащив их с панели "Почтовый ящик" на панель "Избранное" или в раздел "Избранное" панели "Почтовый ящик".
