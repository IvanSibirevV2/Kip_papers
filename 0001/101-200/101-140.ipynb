В определенном смысле можно сказать, что информационная энтропия указывает на долю рассеяния энергии при работе с информацией. В обобщенном смысле «производимая» организацией энтропия - это мера излишней работы при достижении поставленной цели, доля бесполезных побочных процессов или явлений, сопровождающих какую-либо деятельность. __Для любой организации__ как открытой системы должен выполняться следующий принцип: структурное состояние системы должно стремиться к такому, при котором производство энтропии минимально. Системы, где этот принцип заметно нарушается, являются неоптимальными, они проигрывают в конкурентной борьбе с другими организационными системами и распадаются. В условиях организационных изменений реализация данного принципа более затруднительна по сравнению с состоянием устойчивого равновесия организационной системы.

# 3.3. Учет ограничений, налагаемых на управляемость организационной системы

Выражение для оптимальной интенсивности адаптивных преобраюваний (3.13) получено при условии неограниченно высокой потенциальной скорости внутриорганизационных изменений $V_R=\frac{dR}{dt}$ ляющей подразделениям эффективно осваивать весь объем ресурсов, выделяемых на каждом этапе преобразований. Такая модель является условной и требует уточнения путем введения ограничения на величину $V_R$. __На практике__ ограничение значения скорости $V^{\lim}_R$ преобразований может быть обусловлено рядом организационных факторов, наиболее значимыми из которых являются (Акофф, 2002; Ансофф, 1989; Андреева, 2004; Дафт, 2006; Мунин и др., 2008; Широкова, 2008): неадекватность организационной структуры целям и задачам организации; неадекватность системы управления организации ее производственной структуре и технологиям; морально-психологический климат в коллективе организации, не обеспечивающий понимания работниками целей и задач преобразований; отсутствие адекватных
мотивационных программ; чрезмерная централизация управления при многоуровневой вертикальной интеграции; неадекватный масштаб управляемости (излишне большая сфера контроля руководителей всех звеньев); не соответствующий характеру работ масштаб допустимых отклонений, принятый в системе контроля, и т.п.

Рассмотрим отношение $\chi=\frac{V^{\lim}_R}{V_max} $ ограниченной скорости организационных преобразований $V^{\lim}_R$ к максимально возможной скорости изменений внешней среды $V_max$ . __Расчеты__, проведенные с использованием имитационной математической модели, построенной на основе (2.4), (2.5), (3.1), (3.2) - (3.7), (3.9) - (3.13), продемонстрировали, что увеличение значения $\chi$, соответствующее улучшению алаптивного потенциала организации, приводит к снижению величин статистических характеристик рассогласования и увеличению среднего значения (По числу реализаций расчетов моделей.) $a^*$ оптимальной интенсивности изменений - среднего значения эффективно осваиваемых ресурсов (рис. 3.5). __Однако при__ определенных значениях влияющих факторов эти процессы выходят на «насышение», что позволяет говорить о некотором «пороговом» значении скорости преобразований $V^{\lim}_R$ VIm , после которого дальнейшее улучшение алаптивных свойств организации (т.е. повышенне ее организационной, производственной и рыночной гибкости) становится неоправданным.

# 3.4. Анализ зон адаптации
При соблюдении условия (3.13) организация реализует оптимальную интенсивность преобразований, где под 100%-ной интенсивностью понимается максимально возможная интенсивность преобразований, позволяющая эффективно осваивать выделяемые ресурсы с учетом реальной управляемости (потенциала алаптации) организации. Состояние внешней среды с учетом скорости изменения (подвижности) ее ситуационных переменных и уровня информационной неопределенности будем называть напряженностью внешней среды. На рис. 3,6 показано направление роста напряженности внешней среды. Анализ выражения (3.13) позволяет сделать вывод, что в зависимости от конкретных значений т, и 4, организация может находиться в одной из пяти зон напряженности (см. рис. 3.6). Как видно, крайне высокая и высокая интенсивности преобразований (зоны 1, 2 и 3) обоснованы лишь при низком уровне неопределенности внешней среды.

Рис. 3.6. Зоны адаптации

Функционирование организации в условиях средней и высокой неопределенности внешней среды (в зонах напряженности 4 и 5) заставляет менеджмент искать пути целенаправленного снижения интенсивности преобразований, что позволяет достигать минимума статистических характеристик рассогласования за счет «осторожных» (инкрементальных) изменений. Однако абсолютное значение характеристик рассогласования в этом случае, безусловно, окажется более высоким по сравнению с адаптацией в благоприятных в информационном смысле условиях. Тем не менее необоснованное увеличение интенсивности преобразований (затрат ресурсов) в зонах 4 и 5, противоречащее условию (3.13), неизбежно приведет к еще большему ухудшению статистичсских характеристик рассогласования. __Процесс адаптации__, характеризующийся зонами 4 и 5, можно назвать «процессом осторожного преследования», в то время как зоны 1,2 и 3 иллюстрируют процесс "энергичного управления изменениями".

Ансоффу, характеризуются: а) событиями неожиданными, но имеющими аналогию в прошлом (4) и неожиданными и совершенно новыми (5); б) темпами изменений, сравнимыми с реакцией фирмы (4) и превышающими реакцию фирмы (5); в) частичной предсказуемостью (4) и полной непредсказуемостью (5) будушего (Ансофф, 1989). __Очевидно__, что такой подход к адаптации в условиях высокой напряженности (зоны 4 и 5) (см. рис. 3.6) не в полной мере отвечает целям и задачам современных органических структур с высоким потенциалом изменений, поскольку не позволяет им в полной мере использовать их адаптивные возможности. Это противоречие указывает на необходимость перехода таких организаций к проактивному (упреждающему) управлению изменениями, основанному на превентивных мерах. При этом большие уровни неопределенности предъявляют высокие требования к процессам обработки информации в системах управления такими организациями. __Таким образом__, высокий динамический потенциал организационной системы является условием, обеспечивающим ее устойчивость и способность адаптироваться к изменяющимся условиям среды. Рост сопределенности внешней среды влечет за собой либо необходимость снижения интенсивности преобразований, либо переход к стратегии проактивной (упреждающей) адаптации.

Стабильная простая среда: механистические структуры, ориентация на текущие процессы, низкая реакция на изменения. Преобладани «осторожного преследования». __Во вторых__. Нестабильная простая среда: органические структуры, команды, децентрализация, ориентация на планирование, взаимодействие, быстрая реакция. «Энергичное управление изменениями». __В третьих__. Нестабильная сложная среда: органические структуры, децентрализация, интенсивное планирование, интеграторы, очень быстрая реакция. «Осторожное преследование».

Рис. 3.7. Стратегии адаптации и зависимости от типа внешней среды

Энергичное управление изменениями возможно лишь в условиях относительно простой, нестабильной внешней среды. В остальных случаях оптимальной (по рассмотренному критерию) является стратегия «осторожного преследования», заключающаяся в обеспечении процесса преобразования «малыми порциями» ресурсов для обеспечения минимума рассогласований с внешней средой.

# Глава 4. Динамическое моделирование процессов организационного развития в классе систем со случайной скачкообразной структурой
## 4.1. Основные понятия и определения теории систем со случайной скачкообразной структурой 
Как было отмечено выше, широкое применение системного подхода к исследованию хозяйственных организаций предопределило использование в рамках теории организации, менеджмента и других социальных и экономических дисциплин разнообразных моделей социально-экономических систем. Общие принципы построения таких моделей базируются на представлении организаций в виде сложных открытых систем, подверженных воздействию со стороны внешней среды и оказывающих на нее собственное воздействие. __Разнообразие подходов__ к моделированию и анализу организационных систем определяется, с одной стороны, многообразием целей и задач исследования, а с другой - стремлением исследователей к адекватному отражению в моделях как можно большего количества свойств организационных систем. При этом попытки создания универсальной методологии математического моделирования социально-экономических систем неизбежно сталкиваются с проблемой описания сложных процессов, обладающих комплексами экономических, социально-психологических, технических и иных аспектов, плохо поддающихся формализации. __Успехи математической__ теории динамических систем побуждают специалистов в области управления к использованию моделей таких систем для адекватного описания динамических процессов, протекающих как внутри организаций, так и при их взаимодействии с внешней средой. Динамический аспект моделирования позволяет формализовать такие явления и процессы, как жизненный цикл организации, непрерывность анализа и синтеза в ходе адаптации к внешней среде, управление качеством продукции, синхронизация производственных процессов, логистика и т.п.

Стр. 114

Неопределенность внешней среды организации, а также погрешности контрольных операций в ее внутренней среде изучаются с применением стохастических моделей различных классов. Исследование аспектов устойчивости, управляемости, наблюдаемости таких систем заметно усложняется при использовании нелинейных моделей, наиболее адекватно описывающих реальные организационные системы и процессы. __Стремление к оптимизации__ управленческих процессов, в том числе процессов управления структурами организаций, требует привлечения методов оптимального оценивания координат состояния системы, оптимального управления, а также распознавания структур собственной системы и ее надсистем. __Наибольшую сложность__ стохастического динамического моделирования организационных систем представляют разнообразные разрывные случайные процессы, не поддающиеся в общем случае формализации посредством дифференциальных или рекуррентных стохастических уравнений. При этом практика управления рыночными организациями в условиях современного высокоподвижного и неопределенного окружения все настойчивее требует отражения в моделях систем аспектов случайности и непредсказуемости смсны явлений и структур (Ackoff и др., 1996; Hannan M. Т. и др., 1984; Weik и др., 1999; Ализес, 2008; Акофф, 2002; Виханский, 1998; Емельянов, 2000; Михненко, 2010; Волкова и др., 2012; Широкова, 2010). __Решение указанных__ и ряда других задач управления оказывается возможным в классе систем со случайной, в том числе скачкообразно изменяющейся структурой. Методическим и математическим аппаратом решения таких задач является теория систем со случайной скачкообразной структурой (ССС).

В аспекте анализа и синтеза социально-экономических организационных систем под структурой понимается совокупность элементов, выстроенных в иерархической последовательности с учетом всех связей и взаимовлияний между ними. Разработка структуры организации базируется на общих стратегических принципах ее функционирования. Однако в процессе развития организации ее структура подвержена изменениям, которые могут носить как инкрементальный, так и радикальный характер. __При кажущейся__ целенаправленности и полной управляемости процессов корректировки или смены типа организационной структуры на практике менеджменту зачастую приходится сталкиваться с неопределенностью знания о характере сложившихся взаимосвязей между элементами организационной системы, уровне их синергии и динамике изменений. __Характерными особенностями__ таких проблем являются структурная неопределенность, т. с, изменение структуры в случайные моменты времени в процессе функционирования организации, а также стохастичность процессов в них. Неточность информации о сложившихся К текущему моменту времени объективных особенностях структуры организации возрастает с ростом ее размеров, сложности, разнообразия целей и задач. Существенный вклад в увеличение структурной неопределенности вносит децентрализация и территориальная разобщенность частей организации.

Случайные перерывы информации, под воздействием внешней среды приводящие к перерывам (или снижению эффективности) управления, также могут трактоваться как смена структуры организации. Эти особенности являются объективными характеристиками функционирования сложных динамических систем (Михненко, 2010). __Под инкрементальными изменениями__ структуры организации будем понимать изменения, не затрагивающие набор элементов, их иерархию и связи в целом, но приводящие к временным или постоянным модификациям, влияющим на характер работы организации. __Примеры таких модификаций:__ __во первых__, формирование внутри организации устойчивых горизонтальных или диагональных неформальных связей как реакция на излишнюю централизацию в рамках многоуровневой структуры управления; __во вторых__, постепенная утрата позитивной роли какого-либо департамента при изменении целей организации или характера взаимоотношений с внешними партнерами; __в третьих__, значительное ухудшение (или наоборот, улучшение) информационного обеспечения управления в результате внедрения на предприятии новой информационной системы и т.п.

Случайность и информационная неопределенность инкрементальных структурных изменений характерна для большинства современных организационных систем. Недостаточная наблюдаемость скрытых свойств и явлений в ряде случаев влечет за собой скачкообразный характер последующих радикальных изменений. Резкое изменение режима функционирования организации в случайные моменты времени может происходить вследствие нарушений каких-либо параметров (производственных, финансовых, социальных и др.) в отдельных структурных единицах, выхода этих параметров из допустимых пределов, а также вследствие ресурсных ограничений. __Перечисленные свойства__ требуют применения для описания и изучения современных организационных систем особого математического и методологического аппарата, учитывающего случайный характер изменения состояния структуры организации. __Системами со случайной__ скачкообразной структурой называется особый класс стохастических динамических систем, структура которых может изменяться случайным образом. При этом смена структуры представляет собой случайный процесс с конечным числом возможных состояний, изменяющихся в случайные моменты времени с некоторой вероятностью (Бухалёв, 1996). В настоящее время прикладная теория систем со случайной скачкообразной структурой находится на стадии развития. Первые труды, посвяшенные анализу систем с переменной и случайной структурой, приналлежат Скляревичу А. Н. (Скляревич, 1976). Методы анализа и синтеза линейных и нелиней-ных систем ССС изложены в трудах Артемьева В. М., Казакова И. Е., Бухалёва В. А. (Артемьев, 1979; Бухалёв, 1992; Бухалёв, 1996; Казаков и др., 1980; Казаков и др., 1993).

В монографии (Бухалёв, 1996) продемонстрированы такие аспекты практического применения теории ССС как оптимальное оценивание состояния, распознавание структуры и оптимальное (в том числе игровое) управление системами, подверженными скрытым и внезапным изменениям. __Здесь под фазовыми координатами__ понимается совокупность экономических, технических, социально-психологических и иных переменных внутренней среды организации, характеризующих ее состояние в каждый момент времени. Существенной отличительной особенностью описания организаций в классе систем ССС является то, что вектор состояния системы помимо фазовых координат  $x \in X$ включает в себя индекс (номер) структуры 111 $s \in S$ , где $X$, $S$ — области допустимых значений $x$ и $s$ соответственно. В общем случае множества $Х$ и $S$ могут зависеть от времени. __Таким образом__, обобщенный вектор состояния организационной системы состоит из двух компонент: континуальной (х) и принадлежащей конечному множеству (s). Благодаря этому свойству процессы управления приобретают характер совместного воздействия как на фазовые координаты, так и на структуру системы.


Модели теории систем ССС предполагают наличие в составе организационной системы элементов (подразделений или лиц), выполняющих функции индикаторов структуры. Индикаторы структуры осушествляют наблюдения за факторами внутренней среды организации, что позволяет косвенно оценивать состояние ее структуры. Фазовые координаты, определяющие состояние организации в рамках существующей структуры, наблюдаются в ходе мониторинга состояния - комплекса информационно-аналитических мероприятий, проводимых менеджментом организации. При возможности количественного описания наблюдаемых процессов понятие «наблюдение» становится тождественным понятию «измерение». __Целью наблюдения__ за фазовыми координатами и состоянием структуры системы является обеспечение как можно более раннего выявления объективных инкрементальных изменений внутри организации, способных привести к значительным изменениям. __Примером перерастания__ незначительных изменений в радикальную трансформацию может служить динамика развития кризиса организации на каждой из сталий ее жизненного цикла (Ализес, 2008; Gireiner, 1972; Lippitt et al., 1967)

Так, в модели Л. Грейнера (Greiner, 1972) на сталии креативности основной целью организации является созлание продукта и выжив ние на рынке; преоблаляют неформальные коммуникации. Основатель ориентированы на технологии или предпринимательство и пренебре гают упракленческой деятельностью. Источник развития организации заключается в творческом и предпринимательском потеициалах ее собственника. На этой сталии жизненного цикла организация неблюрократична и неформальна. __Однако по мере роста организации__, увеличения числа сотрудников и спектра залач начинают появляться признаки недостаточной формализации управленческих и производственных процессов. Отсутствие инструкций и процедур приводит к увеличению времени выполисния простых операций, неэффективной коммуникации, нсупорядоченной координации действий и т.п. Отсутствие насмного менеджмента приводит к перегрузке предпринимателя залачами оперативного управления. Эти незначительные на первый взгляд (и вполне закономерные) проблемы способны быстро перерасти в кризис - исоблодимость формализации управления и создания системы менеджмента. __Неспособность руководителя организации__ своевременно выявить эти изменения может привести к серьезным организационным проблемам и необходимости выработки экстренных антикризисных мер, к тому же далеко не всегла позитивно принимаемых персоналом.

В конечном счете внезапные и скрытые изменения внутренних характеристик организационной системы зависят от интенсивности возлействий на организацию внешней среды. Аппарат теории систем ССС позволяет трактовать малые изменения (скачки) параметров организационной системы как частный случай изменения структуры без нарушения характера взаимных связей между элементами системы. __Удобство моделей__ систем ССС состоит в том, что даже незначительные изменения внутренних факторов организации можно трактовать как появление новой структуры путем введения в множество значений $S$ нового индекса $s$. Такой подход оказывается оправланным в тех случаях, когда требуется описать изменения, не приволяшие к формальной трансформации набора элементов, их иерархии и связей, но затрагивающих характер этих связей. Например, увеличение степени децентрализации управления, формализация линейных отношений, изменение мотивации сотрудника и т. п. __Априорное формирование множества S__ представляет собой достаточно сложную задачу, решаемую на основе всестороннего изучения особенностей организационной системы. Как правило, адекватность множества S ограничивается каким-либо временным периодом исслелования, на котором справедливо предположение о возможном наборе структур и вероятностях их смены (Бухалёв, 1996).

В основе теории систем ССС лежит концепция марковских моделей: модель системы характеризуется вектором текушего состояния, который является марковским процессом (Аоки, 1971; Пугачев и др., 1990; Ярлыков и др., 1993; Тихонов и др., 1977). В отличие от математического аппарата теории массового обслуживания, не учитываошего континуальной составляющей управляемой системы, теория систем ССС позволяет решать задачи, в которых структура системы зависит от ее фазовых координат, и наоборот. __Теория систем ССС__ представляет собой органическое объединение таких научных направлений, как теория марковских процессов и байссовский подход к обработке информации (Бухалёв, 1992). Таким образом, важной отличительной особенностью теории ССС от широко распространенной теории массового обслуживания, изучающей изменение структуры системы, является признание факта зависимости структуры от ее состояния (фазовых координат) системы. __Состояние дискретной динамической системы__, приналлежащей классу систем со случайной скачкообразной структурой, характеризуется вектором $[x_k,s_k]$ где $х_k$ — вектор фазовых координат; $s_k$, — вектор состояний структуры. Взаимозависимость этих двух компонент общего вектора состояния показана на рис. 4.1 (Бухалёв, 1996), где и, — управленческое воздействие на систему.

Рис. 4.1. Схема системы с взаимосвязью структуры и фазовых координат

Взаимозависимость структуры и состояния системы характерна для организационных (социально-экономических) систем. Особенно заметной эта взаимозависимость становится в период организационных изменений, когда реформирование организационной структуры предприятия вызывает изменения социально-психологических и экономических переменных $x_k$, характеризующих состояние организации. В то же время изменения характеристик текущего состояния предприятия, вызванные неструктурными аспектами (повышением квалификации персонала, изменением порядка и правил трудовой мотивации, введением новых стандартов поведения и т.п.), как правило влекут за собой комплекс больших и малых изменений организационной структуры $s_k$, (перераспределение должностных полномочий, изменение структуры ответственности, создание или упразднение структурных единиц и т.п.). __Говоря о трансформациях__ структуры организационной системы, далеко не всегда следует иметь в виду кардинальные изменения типа ОСУ или заметные преобразования в структуре полномочий и ответственности, бизнес-процессов, подходах к департаментализации и пр. Под изменениями структуры организационной системы могут пониматься случайные временные перерывы или нарушения принятых коммуникаций, обусловленные внутренними и внешними факторами.

К внутренним факторам, способным повлиять на целостность и эффективность коммуникационной, информационно-управляюшей, контрольно-учетной и других подсистем предприятия, можно, например, отнести: __во первых__, назначение недостаточно опытного сотрудника на должность, предполагающую функции координации (проектов, залач, информационных потоков); __во вторых__, изменение состава функциональных обязанностей того или иного сотрудника, выполнявшего функции наблюдателя и распределителя информации; __в третьих__, неполадки в технических сетях (телефонная связь, электронная почта и др.) и т.п. __Среди внешних факторов__, способных нарушить целостность и эффективность системы коммуникаций предприятия с элементами его делового окружения, можно отметить изменение организационной структуры или кадрового состава компаний-партнеров, появление новых регламентов ведения переговоров с контрагентами, форсмажорные обстоятельства и т. п. __Случайные скачкообразные процессы__, протекающие в организационных системах, предъявляют определенные требования к ее системе управления, в частности, к информационной и контрольно-учетной подсистемам. Эти требования могут заключаться в необходимости создания системы мониторинга структуры и наблюдения за фазовыми координатами. Мониторинг структуры заключается в отслеживании индикаторной информации, косвенно указывающей на состояние структуры организации. Полнокомпонентное наблюдение за всеми значимыми переменными и параметрами, характеризующими структуру и состояние системы, в реальных организациях, как правило, оказывается невозможным. __Таким образом__, современные организации в процессе адаптации к сложной и неопределенной внешней среде претерпевают случаяные изменения своей структуры, вызванные скрытыми изменениями. Текушее состояние организации, характеризующееся вектором экомомических, технических, социальных и иных переменных, зависит от особенной ее структуры. Адекватное оценивание состояния струкгуры организации и оптимальное управление процессом ее изменения представляет собой актуальную задачу современного менеджмента. Решение этой залачи возможно с применением моделей, построенных в классе систем со случайной скачкообразной структурой.

# 4.2. Принципы математического моделирования процессов организационного развития в классе систем со случайной скачкообразной структурой
Модель организационной системы со случайной скачкообразной структурой имеет следующие характерные черты (Бухалёв, 1996; Казаков и др., 1980): __во первых__, система состоит из элементов однородной или неоднородной сущности, объединенных целенаправленным процессом управления; __во вторых__, система является динамической, т.е. ее состояние (фазовые координаты) изменяется с течением времени; __в третьих__, система является стохастической, так как входные воздействия и ряд внутренних факторов представляют собой случайные процессы; __в четвертых__, структура системы в целом или ее подструктуры могут скачкообразно изменяться по случайному закону в случайные моменты времени, причем эти изменения могут либо зависеть, либо не зависеть от фазовых координат системы; __в пятых__, число структур, в которых может находиться система, конечно. На рис. 4.2 показана общая схема системы со случайной скачкообразной структурой (Бухалёв, 1996). __На схеме приняты следующие обозначения__: $\Theta$ — вектор управления структурой организации; $u$ — вектор управления фазовыми координатами организации; $s$ — индекс (номер) структуры; $x$ - вектор фазовых координат организации; $r$ - вектор показаний индикатора структуры (информания от подразделения, осушествляющего мониторинг структурных параметров); $z$ — вектор наблюдений (информация от подразделения, осушествляющего мониторинг внутренних переменных организации); $р$ - вектор вероятностей возможных структур; $\widehat{x}$ - вектор оценок фазовых координат организации (вектор максимально достоверной, релевантной информации как результат информационно-аналитических мероприятий, направленных на снижение степени неопределенности информации о внутренней среде организации): $\zeta$, $\xi$ - случайные факторы, определяющие стохастический характер смены структуры и изменения фазовых координат соответственно; $v$, $\Theta$ — векторы информационных погрешностей (ошибок), которым подвержен индикатор структуры и измеритель состояния соответственно.

Рис. 4.2. Обобщенная схема организационной системы со случайной скачкообразной структурой

Описание динамики организационной системы рассматривается в дискретном времени ($k = 0, 1, 2, \dots$). Динамика индекса $q_k$, структуры $s_k$, моделируется условной марковской цепью, определяемой условными вероятностями переходов из одной структуры в другую (включая случаи неизменности структуры) $q_k(s_{k+1}|s_k,\Theta_k,x_k,\xi_k)$ при фиксированных значениях переменных, стоящих справа от черты, где индекс $k$ при функции $q$ означает, что в общем случае $q$ зависит от времени. __Например__, при $S = {1, 2, 3}$ возможны девять переходов (рис. 4.3), где $q(1|2)$ - еcть вероятность того, что система, имеющая на $k-м$ этапе структуру $s_k = 2$, на этапе $k + 1$ будет иметь структуру $s_{k+1} = 2$. __Как видно__, вероятность перехода организации из одной структуры в другую зависит от целенаправленного управления этим процессом $\Theta_k$, текущего состояния организации $х_k$ и случайных факторов $\zeta_k$. Очевидно, что сформировавшаяся к моменту времени $k$ структура $s_k$, в свою очередь влияет на текущие фазовые координаты организации $х_k$ (см. первую группу перекрестных связей на рис. 4.2).

Таким образом, согласно теории систем ССС смена структуры организации осуществляется под воздействием трех векторов: __во первых__, целенаправленного управления структурой $\Theta$; __во вторых__, текущих фазовых координат организации $х$ в рамках сушествуюшей структуры; __в третьих__, вектора случайных неуправляемых факторов $\zeta$. Вектор фазовых координат $х$ организации как динамического объекта описывается уравнением $x_{k+1} = \varphi_k(x_k,u_k,s_{k+1},\xi_k)$ где $\varphi_k$ — векторная детерминированная функция случайных аргументов, в общем случае зависящая от $k$. В то же время информация о текушем состоянии организации совместно с результатами мониторинга структуры будет определять вывод менеджмента об особенностих текушей структуры.

Важной особенностью моделей систем ССС является то, что объслиненный вектор состояния системы $[x_k,s_k]$ является марковским, в то время как его компоненты $x_k$, и $s_k$, являются условно-марковскими, а именно: при фиксированном состоянии структуры $s_k$, вектор $x_k$ - марковский процесс, а при фиксированном $х_k$, индекс структуры $s_k$ - марковская цель (Бухалёв, 1996). __Поскольку состояние организации__ в каждый момент времени $k$ характеризуется фазовыми координатами х, (переменными се внутренней среды) и номером структуры $s_k$, обработка информации о текущем состоянии организации должна включать в себя распознавание структуры и оценивание фазовых координат. Под распознаванием структуры понимается определение вероятности того, что текущая структура совпадает с одной из структур, составляющих конечное множество S возможных значений s, (Бухалёв, 1996).

-------------

Информация, формируемая индикатором структуры, представляет собой некоторый признак, косвенно указывающий на то, что к настоящему моменту времени сформировалась та или иная структура, непосредственное наблюдение которой недоступно. Характеристики индикатора структуры моделируются марковской цепью с состояниями, определяемыми вероятностью $\pi_k(r_k,s_k)$, где $r_k$ — признак структуры. __Например__, возможны две структуры организации: $s= 1$ и $s=2$. Индикатор может выдавать значения 1 и 2, косвенно указывающие на номер структуры. Тогда $\pi_k(1|1)$ - есть вероятность того, что индикатор выдаст значение 1 при $s= 1$, а $\pi_k(2|1)$ вероятность того, что при той же структуре индикатор выдаст значение 2. __Вероятность__ $\pi_k(r_k | s_k)$ не является вероятностью той или иной структуры, она представляет собой лишь условную вероятность некоторого контрольного события $r_k$, при гипотезах о номерах структур. Контрольное событие $r_k$, играет роль обновляющего процесса, обеспечивающего процедуру идентификации структуры текущей апостериорной информацией.

Совокупность всех возможных значений $\pi_k(r_k | s_k)$ можно назвать вероятностной характеристикой индикатора, отражающей его способность выполнять свою функцию. Очевидно, что эта характеристика зависит от уровня информационной погрешности $v$ (см. рис. 4.3), с помощью которой моделируется эффективность (точность) работы подразделения или лица, отвечающего за мониторинг структуры. Чем выше информационная погрешность (т.е. чем хуже работает индикатор), тем ниже вероятность того, что в условиях объективно сформировавшей структуры будет зафиксирован признак именно этой структуры. __Измерение фазовых координат__ (наблюдение за характеристиками работы организации) описывается уравнением (Бухалёв, 1996): $z_k=\Psi_k(x_k,z_{k-1},s_k,u_{k-1},\Theta_{k-1},\Theta_k)$ __Зависимость измерения__ фазовых координат от предыдущего измерения $z_{k-1}$, указывает на марковский характер измерений, когда вся предыстория измерений содержится в предыдущем измерении. Такая модель измерений технико-экономических переменных в организационных системах представляется наиболее адекватной. __В целом же задача__ обработки информации в системе ССС состоит в оценивании обобщенного вектора текущего состояния организации $[x_k,s_k]$, который включает в себя вектор фазовых координат $x_k$ и индекс, сформировавшейся к текущему моменту времени структуры $s_k$.

-----------

Для определения оптимальных оценок обобщенного вектора состояния организации необходимо задать критерий в виле функини потерь, отражающей основные экономические и иные требования, предъявлясмые к организации. __В общем виде задача__ оптимальной обработки информации в систенах ССС заключается в определении апостериорной плотности вероятности обобщенного вектора состояния системы (Бухалёв, 199): $\widehat{f}_k(x_k,s_k) = f_k(x_k,s_k|z_{ \overline{0,k} }, r_{\overline{0,k}})$ на основе наблюдения $z_{\overline{0,k}}$, $r_{\overline{0,k}}$ на интервале времени $[0, k]$. __Определение апостериорной плотности__ вероятности обобшенного вектора состояния представляет собой «многоканальный• байесовский алгоритм, где под «каналами» понимаются структуры системы, составляющие множество S (Бухалёв, 1996). Основным достоинством такого алгоритма является совместное распознавание структуры и оценивание фазовых координат. __Как видно__, общая постановка задачи обработки информации в системах ССС предполагает построение плотности вероятности обобщенного вектора состояния на основе всей истории измерений. Однако в абсолютном большинстве практических задач управления социально-экономическими организационными структурами имеют место отсутствие качественных априорных данных, неполнота и недостоверность наблюдений, а также отсутствие данных об истории наблюдений. Использование в этих случаях оптимальных байесовских алгоритмов становится невозможным. Приемлемую точность оценивания могут обеспечить приближенно-оптимальные алгоритмы, основанные, например, на методе двухмоментной параметрической аппроксимации (Бухалёв, 1996; Бухалёв, 1992). В основе метода лежит замена неизвестных условных плотностей $/,(x_i ,s_i)$ известными, зависящими не более чем от двух первых моментов этих условных распределений, при предположении о марковости процесса измерения фазовых координат.

Пример. Пусть промышленное предприятие может выпускать продукцию, используя либо одну (структура $s_1$), либо две (структура $s_2$) производственные линии. Структура $s_1$ является менее затратной по сравнению со структурой $s_2$ ($С_1 < С_2$, где $С_1$, $С_2$ - издержки эксплуатации одной и двух линий соответственно). Управление структурой заключается в выборе з, или з, на каждом шаге управления. __Для упрощения__ в задаче не учитыватся снижение рентабельности предприятия при простое одной из линий. __При прочих равных__ условиях структура $s_1$, обеспечивает меньший объем производства продукции $Q$, чем структура s,. Объем производства $Q$ (в денежном выражении) представляет собой фазовую координату $х$. __Другим способом увеличения объема__ производства является увеличение скорости конвейера производетвенной линии $V$, т,е. скорость движения конвейера представляет собой управление и  $u \in[0, V_{lim}]$, где $V_lim$ — предельно допустимая по нормам условий труда скорость конвейера. Однако увеличение скорости конвейера влечет за собой увеличение издержек за счет стимулирующих выплат персоналу, работающему в сложных условнях. __Таким образом__, излержки и объем производства продукции являются функниями структуры предприятия и скорости конвейера: $С = С(s, V)$, $Q = Q(s, V)$. __Требуется определить оптимальное управление__ производством ${u, 0}_{opt}$ д по критерию максимума функционала $J = Q(s, V) - C(s, V)$.

Очевидно, что управленческое решение и о выборе скорости конвейера будет зависеть от решения $\Theta$ о выборе структуры. В то же премя выбор структуры зависит от выбора скорости. Именно эта взаимосвязь при принятии комплексного решения иллюстрируется на рис. 4.2 третьей парой перекрестных связей. Согласно теории ССС интерес представляет именно комплексная (одновременная) оптимизация вектора $(u, \Theta)$. __В самом общем случае__ синтез оптимального управления организацией как динамической стохастической системой ССС основывается на методе динамического программирования в сочетании с байесовским подходом к обработке информации (Аоки, 1971; Беллман, 1960). __Таким образом__, сложность и многогранность деятельности современных хозяйственных организаций, действующих в условиях неопределенной и высокоподвижной внешней среды, заставляет исследователей и практиков обращаться к разнообразным моделям социально-экономических систем. Модели организаций, основанные на концепции теории систем со случайной скачкообразной структурой, представляют собой наиболее общий способ формализации сложных социально - экономических процессов с учетом их разрывности и структурной вариабельности. Теория систем ССС позволяет синтезировать оптимальные и приближенно-оптимальные алгоритмы управления организационными системами с учетом практической необходимости комплексного управления как состоянием, так и процессом изменения структур организации.

# 4.3. Методологические основы моделирования жизненного цикла организации в классе систем со случайной скачкообразной структурой
Концепция жизенного цикла организации (ЖЦО) возникла и развивается в теории организации в целях объяснения причин, характеристики последствий изменений, происходящих в организациях с течением времени. Согласно этой концепции организационные характеристики организаций меняются на разных стадиях их жизненного цикла, поэтому на разных этапах развития им требуются различные методы управления. Кроме того, для разных стадий должны быть характерны различные подходы к оцениванию организационной эффективности. __По мнению Г. В. Широковой__, «изучение различных управленческих аспектов на разных стадиях ЖЦО предоставляет менеджерам мощный инструмент управления изменениями, который основан на закономермостях перехода от одной практики управления к другой и внутренней логике развития компании» (Широкова, 2008). __Жизненный цикл организации__ — это совокупность последовательно сменяемых друг друга состояний организации, каждое из которых соответствует определенному комплексу организационных характеристик и типичной модели поведения организации. В основе данной концепции лежит аналогия с жизненным циклом развития живого организма. В последние годы огромный интерес к моделям ЖЦО вызван рядом знаковых публикаций на эту тему в научных и деловых журналах. Кроме того, популярность идеи растет благодаря активной лекторской, писательской и консалтинговой деятельности одного из ведущих современных специалистов по управлению жизненным циклом организаций доктора И. Адизеса.

На сегодняшний день в научной литературе описано большое количество моделей ЖЦО, предложенных в разные годы разными авторами. Эти модели различаются как по числу стадий развития организаций, так и по составу и содержанию характеристик (переменных), отражакщих особенности построения и функционирования организаций на этих стадиях. Несмотря на это в отечественных учебных изданиях по менеджменту и теории организации зачастую приводятся некие обобщенные «иллюстративные» модели, не всегда достаточно точно отражающие авторскую идею или вообще лишенные указания на авторство. __Наиболее полный__ и систематизированный сравнительный анализ моделей жизненного цикла организаций сегодня представлен в трудах ученых Высшей школы менеджмента СПбГУ Г. В. Широковой, И. С. Меркурьевой, О. Ю. Серовой и др. (Широкова и др., 2006; Широкова и др., 2006; Широкова, 2008). В настоящей монографии систематизация моделей ЖЦО изложена в соответствии со схемой, предложенной авторами указанных исследований (см. приложение 4). __Названия__ «Становление», «Накопление», «Зрелость», «Диверсификация», •Упадок», используемые в таблице для обозначения типовых стадий ЖЦО, являются условными и отражают структуру пятиэтапной схемы, взятой авторами исследования за основу. В свою очередь, размешение в рамках одной типовой сталии нескольких авторских сталий указывает на их общие признаки (например, соответствие типовой стадии « Упадок», авторских стадий «Аристократизм», «Салем Сити», «Бюрократизация» и «Смерть» модели Адизеса).

Если для ранней литературы, посвященной ЖЦО, более характерным был теоретический характер, то современные авторы делают акцент на эмпрических подходах и анализе прикладного значения этой концепции. __Прикладное значение моделей ЖЦО__, по-видимому, заключается в том, что ситуационный подход к управлению (иначе говоря, ситуационное мышление менеджмента) совместно с системным представленнем о механизмах развития организаций требует от руководителей четкого понимания, на каком этапе жизненного цикла находится я и какие методы управления оптимальны в этой ситуации. __Концепция жизненного цикла__ позволяет выделять характерные, устойчивые особенности состояния организации на каждом из этапов се существования. Однако наиболее важным представляется вывод, который делает Б. 3. Мильнер: «Переходы от одного этапа к другому якляются предсказуемыми, а не случайными» (Мильнер, 2002). Строго говоря, употребление термина «не случайными», на наш взгляд, не вполне оправдано, поскольку стохастический аспект присуш практически любым сложным процессам, к которым, безусловно, относятся и процессы функционирования социально-экономических систем. Тем не менее стохастическая (вероятностная) трактовка смены этапов ЖЦО не отрицает ее предсказуемости, она лишь требует привлечения для анализа методов теории вероятностей и математической статистики.

При этом практика управления современными рыночными структурами, функционирующими в условиях неопределенного и высокоподвижного внешнего окружения, все настойчивее требует отражения в организационных моделях аспектов случайности и непредсказуемости смены явлений и структур. Переходы организашионных систем от одного этапа жизненного цикла к другому как раз относятся к процессам такого типа. __Наиболее общий подход__ к исследованию ЖЦО предполагает детальное изучение природы и структуры всех явлений внутри организации, а также случайных воздействий на нее извне, анализа реакции организационной системы на эти воздействия и выяснения влияния структурных параметров организации на характер этих реакций с последующим обобщением результатов. Очевидно, что решение этих залач требует анализа не только основных закономерностей, определявощих функционирование организации в общих чертах, но случайных процессов и явлений, обусловленных второстепенными факторами.

---------------

К явлениям и процессам, порожлающим стохастичность и неопределенность организационных систем, можно, например, отнести следующие: __во первых__, нечеткость целей на начальном этапе развития организации; __во вторых__, неформальные аспекты коммуникаций и принятия решения, характерные для этапа коллективности; __в третьих__, поиск направлений развития в условиях высокого уровня конкурентного давления; __в четвертых__, слабая предсказуемость изменений внешней среды, обусловленных политическими и экономическими факторами на межлународном уровне; __в пятых__, недостаточная определенность причин снижения основных экономических и иных показателей на этапе спала активности и т.п. __С теоретической точки зрения__ случайные факторы ничем не отличаются от так называемых основных факторов. Однако попытки выявления огромного числа закономерностей неизбежно сталкиваются с проблемой возрастания сложности и громоздкости аналитических процедур, которые к тому же не всегда приводят к практически значимым результатам. При этом вполне определенные и устойчивые закономерности могут быть выявлены при изучении совокупностей однородных явлений.

Уровень стохастичности аспектов деятельности организации зависит от этапа ее жизненного цикла. Те стадии, которые принято называть стадиями зрелости и стабильности (модели Липпита и Шмидта, Каца и Канна, Фламхольца и др.), как правило, характеризуются заметно большей определенностью, т.е. снижением степени величины информационной энтропии за счет формализации управления, совершенствования организационной структуры на основе адаптационных механизмов, институционального характера взаимоотношений с внешней средой и т.п. (Широкова и др., 2006). Для других этапов характерны более высокие значения информационной энтропии, что влечет за собой увеличение степени риска при принятии решений. __Другим важнейшим аспектом ЖЦО__ является фактор внезапности трансформаций организационной системы, обусловленных нарастанием скрытых или плохо наблюдаемых изменений. В ряде случаев при кажущейся целенаправленности и полной управляемости бизнес-процессов менеджменту приходится сталкиваться с неопределенностью знаний об особенностях взаимосвязей между элементами организационной системы, уровне их синергии и динамики изменений. Случайность и, как следствие, высокая информационная неопределенность инкрементальных структурных и процессных изменсний характерна для большинства современных рыночных организаций.

Стр. 130

----------

При этом недостаточная наблюдаемость скрытых явлений в большинстве случаев влечет за собой скачкообразный характер последующих раликальных изменений. __В конечном счете__ большинство изменений внутренних характеристик организации зависит от интенсивности воздействий на нее факторов внешней среды. __Накопление малозаметных изменений__ и перерастание их в раликальные трансформации характеризует развитие кризисов в организации на кажлом из этапов се жизненного цикла. Эти проблемы, порой малозачетные, способиы перерасти в кризис, требующий незамедлительной процедуры формализации упрапления и созлания системы менеджмента. Неспособность руководителя своевременно выявить малые изменения в работе организации может привести к серьезным организационным проблемам и необходимости выработки экстренных антикризисных мер. __Таким образом__, плохо наблюдаемые изменения, происходяшие в организации, могут приводить к качественным, иногда скачкообразным изменениям ее общего состояния и структуры.

Разумеется, деление жизненного цикла организаций на отдельныс этапы является весьма условным, и разнообразие моделей подтвержлает это. Однако возможность выявления и формального описания устойчивых характерных особенностей каждого этапа, помимо всего прочего, имеет существенное методологическое значение: такая «разрывная» интерпретация в целом непрерывного процесса позволяет моделировать ЖЦО в классе систем со случайной, скачкообразно изменяющейся структурой. __В рамках исследования ЖЦО__ под структурой понимается некоторое определенное состояние организации, характерное для конкретного этапа се развития. Другими словами, переход организации от одного этапа жизненного цикла к другому в результате постепенных или скачкообразных изменений может рассматриваться как процесс смены структур. При этом набор структур (набор этапов ЖЦО) считается конечным и заранее известным". __Такая трактовка__ набора этапов в целом соответствует концепини исследовния ЖЦО. Тем не менее в ряде практических залач множество структур, соотвстствующих этапам ЖЦО, может быть не полностью определенным априори. __Важно заметить__, что понятие «структура» в рамках данной модели гораздо шире понятия «организационная структура», употребляемого в традиционном смысле. Здесь под структурой понимается достижение организационной системой состояния динамического равновесия между ее внутренней и внешней средой, характеризующего данный этап ее жизненного цикла.

Стр. 131

-------------------

Тезис о случайности смены структур в случайные моменты времени, на наш взгляд, нисколько не противоречит основным положениям концепции жизненного цикла организации. В частности, вывод о том, для уяснения сущности каждой стадии ЖЦО необходимо понимание того, как характеристики организации изменяются во времени, а также о том, что организация способна возвращаться на более ранние сталии развития (Широкова, 2007), подтверждают необходимость учета динамического аспекта этих процессов, причем именно в стохастической интерпретации. __Структурная стохастичность модели ЖЦО__, образно говоря, позволяст "развязатье" последовательность этапов, устранив необходимость жестко определять их временные периоды, а также направления переходов. При этом такая интерпретация не приводит к утрате выводов о закономерностях развития организации, а лишь переводит их в разряд вероятностных (статистических) оценок. __Важно подчеркнуть__, что термин «скачкообразный», тралиционно используемый для обозначения систем, способных изменять свою структуру за предельно малые промежутки времени, применительно к моделям ЖЦО не обязательно предполагает именно такой характер смены этапов ЖЦО. Скачкообразность в данном случае имеет метолологический смысл, указывающий на то, что в целом плавные процессы организационных изменений рассматриваются как последовательность дискретных этапов, обладающих отличительными свойствами (Михненко, 2010; Михненко, 2008).

Существенной особенностью описания ЖЦО в классе систем ССС является то, что вектор состояния организации, как указывалось выше, помимо набора технико-экономических и социально-психологических характеристик (континуальной части) включает в себя еще и номер структуры (этапа развития). В отличие, например, от математического аппарата теории массового обслуживания, не учитывающего континуальной составляющей характеристик управляемой системы, теория систем ССС позволяет решать задачи, в которых структура системы (в данном случае - этап ЖЦО) зависит от ее характеристик и наоборот, характеристики определяются этапом жизненного цикла, что в полной мере соответствует концепции ЖЦО. Поэтому в наиболее общей постановке задача оценивания состояния организации приобретает характер определения плотности вероятности обобщенного вектора, учитывающего как характеристики организации, так и этап ее жизненного цикла. Необходимость вероятностного подхода к оценке обшего состояния организации обусловлена тем, что на практике менеджмент в подавляющем большинстве случаев не в состоянии однозначно определить, на каком этапе жизненного цикла находится его организация. Наличие же вероятностной оценки этала ЖЦО позволит руководству принимать алекватные решения, используя ситуанионный подход к управлению.

Стр. 132

-------

Модель ЖЦО, построенная в классе систем ССС, не только не отрицаст существующие модели, основанные на выделении классификационных признаков, но и рассматривает их в качестве перпервичных (базовых) инструментов исследования. Применение теории ССС для анализа ЖЦО предполагает выбор на первом этапе одной изи нескольких известных моделей, основанных на системе признаков смены этапов жизненного цикла. В результате выбора первичных молелей формируются конечные множества (наборы) этапов ЖЦО. __В основе модели ЖЦО__ может лежать теорема гипотез - формула Байсса, при этом набор этапов ЖЦО, соответствующий выбранной базовой модели, рассматривается как полная группа несовместных гипотез о равновесных состояниях организации (Вентцель, 2003). __Текущая (апостериорная на каждом шаге анализа)__ вероятность наложасния организации на том или ином этапе жизненного цикла $P[s_i(k)|r(k)]$ определяется как вероятность соответствуюшей гипотезы при условии совершения заранее оговоренного контрольного (индикаторного) события $r(k)$, выступающего в роли «обновляющего процесса»:

$$P[s_i(k)|r(k)]=\frac{P[s_i(k)]\pi[r(k)|s_i(k)]}{\sum\limits_{i}(P[s_i(k)]\pi[r(k)|s_i(k)])}$$

где $k$ - шаг анализа ЖЦО; $s_i(k)$, $і = \overline{1,n}$ — гипотезы о возможных этапах ЖЦО на $k$-м шаге анализа. Начальные (априорные) вероятности гипотез $P[s_i(k)]$ считаются известными на момент начала процедуры динамического анализа ЖЦО.

Индикаторное событие г (k) представляет собой некоторый призак, косвенно указывающий на то, что к настоящему моменту времени сформировалось (или формируется) равновесное состояние организационной системы, соответствующее тому или иному этапу с жизненного цикла. Индикаторные наблюдения должны осуществиться подразделениями организации, ответственными за мониторинг се состояния (жизненного цикла). __Строго говоря__, байссовский подход и основанные на нем модели систем ССС позволяют рассматривать в качестве «обновляющего процесса» не только косвенные признаки конкретного этапа ЖЦО, во и данные многомерного наблюдения (измерения) континуальвой составляющей состояния организации - ее технико-экономических, организационных и социально-психологических характеристик. Однако такой подход представляет собой достаточно сложную задачу, причем точность выводов далеко не всегда возрастает с уведичением числа анализируемых факторов. Поэтому «индикаторный» вректер «обновляющего процесса» представляется в практическом аспекте более адекватным и менее затратным. __Априорной характеристикой__, определяющей степень достоверности исудьтатов мониториига, является известная вероятность $\pi[r(k)|s_i(k)]$, показываюшая, насколько точно работаст «индикатор».

Стр. 133

-----------

Как сказано выше, методология теории систем ССС позволяет расширить байесовскую модель за счет дополнительного параметра - условной вероятности переходов', указывающего на наличие у менеджмента априорной информации о закономерностях смены этапов жизненного цикла в соответствии с выбранной системой признаков. __Расширение байесовской модели__ за счет использования этого параметра позволяет в полной мере учитывать предположения о закономерности смены структур. Тогда выражение для апостериорной вероятности гипотез примет вид

$\widehat{p}(s_i)= \frac{\sum\limits_{j} \pi(r|s_j)q(s_i|s_j)\overline{p}(s_j)}{\sum\limits_{i}\sum\limits_{j}\pi(r|s_j)q(s_i|s_j)\overline{p}(s_j)}$

где $\widehat{р}$ - начальная вероятность нахождения организации в той или иной структуре (априорная информация).

Например, условная вероятность $q(s2|s1)$ - есть заранее известная менеджменту вероятность перехода организации на стадию Nº2 при условии, что в настоящий момент она находится на стадии Nº 1. Очевидно, что вероятности переходов есть не что иное, как формальное отражение знаний, опыта и интуиции руководителя (исследователя). __В ряде случаев__ прикладные модели ЖЦО могут не включать в себя вероятности переходов, указывая на отсутствие у исследователя достаточных оснований предполагать закономерности смены структур. В этом случае вероятности переходов считаются равными единице. Однако в большинстве случаев у менеджмента организации имеются вполне обоснованные предположения как минимум о некоторых переходах, основанные, например, на данных статистических исследований независимых организаций и т.п. 

__Пример__. В качестве примера рассмотрим алгоритм оценивания состояния структуры и фазовых координат организации. Пусть конечное множество структур организации на рассматриваемом этапе ее жизненного цикла равно двум: $S = \{ s_k\}$, $k = 1,2$, где $s_1$ - структура управления, характеризующаяся доминированием бюрократических принципов взаимодействия; s, — органическая структура, отличительной особенностью которой является преобладание неформальных связей и процедур, способствующих эффективной адаптации процессов управления к изменениям внешних условий. Разумеется, категоричное деление особенностей управления на две противоположные структуры является условным и достаточно грубым, в реальных залачах множество S может быть расширено за счет «промежуточных» форм. __Термин «условная вероятность»__ употребляется здесь в традиционном для теории вероятностей смысле, отражая зависимость величины вероятности от выбранного условия (в данном случае — условия, что организация находится на определенном этапе жизненного цикла).

Стр. 134

-------------

При переходе организации со стадии координации к сталии сотрудничества (Greiner, 1972) процесс смены бюрократической структуры органической является закономерным и желательным, однако в большинстве случаев недостаточно наблюдаемым и тем более управляемым. __Допустим__, что анализ дсятельности организации, а также результаты бенчмаркинга позволили менеджменту организации сделать следующие априорные выводы: на данной стадии жизненного цикла вероятность перехода организации из состояния $s_1$, в состояние $s_2$, достаточно велика и составляет $q(s_2|s_1) = 0,7$. Соответственно, $q(s_1|s_1) = 0,3$ — вероятность сохранения бюрократического типа структуры. Вероятность «возврата» в бюрократию мала - $q(s_1|s_2) = 0,1$, при высокой вероятности сохранения органических принципов управления - $q(s_2|s_2) = 0,9$.

Вторая групна априорных данных включает в себя вероятности правильной и ложной работы «индикаторов» - системы мониторинга структуры. При невозможности всестороннего и полного охвата текущих структурных особенностей прямым контролем «индикатор» позволяет реализовать косвенное наблюдение. Пусть в качестве наблюдаемого параметра, косвенно характеризующего состояние структуры, принята частота согласований г оперативных задач на базовом уровне управления (событие р). Очевидно, что высокое значение этого параметра характеризует предельную централизацию управления, косвенно указывая на механистические признаки. Предположим, что работа «индикатора» характеризуется следующими параметрами на каждом шаге анализа: $\pi(r= 1[s_1) = 0,8$ - вероятность того, что в условиях структуры $s_1$ «индикатор» зафиксирует по крайней мере одно согласование $r$ в единицу времени (например, за один месяц); $\pi(r=0|s_1) = 0,2$ - вероятность отсутствия согласования $r$ при $s_1$; $\pi(r=0|s_2)=0,75$ — вероятность отсутствия согласования $r$ при $s_2$; $\pi(r=1|s_2) = 0,25$ - вероятность по крайней мерс одного согласования $r$ при $s_2$.

Очевидно: $\pi(r=1|s_1) = 1$; $\pi(r=0|s_1) + \pi(r=1|s_2) + \pi(r=0|s_2)=1$.

К третьей группе априорных сведений относится начальная вероятность нахождения организации в той или иной структуре:

$\overline{p}(s_1) = 0,85$; $\overline{p}(s_2)=1-\overline{p}(s_1)=0,15$.

Тогда апостериорная вероятность нахождения организации в той или иной структуре определится в соответствии с байссовским принципом (Бухалёв, 1996):

$\widehat{p}(s_1)=\frac{\pi(r|s_1)q(s_1|s_1)\tilde{p}(s_1)+\pi(r|s_2)q(s_1|s_2)\tilde{p}(s_2)}{\sum\limits_{i=1}^2\sum\limits_{j=1}^2\pi(r|s_j)q(s_i|s_j)\tilde{p}(s_j)}$

$\widehat{p}(s_2) = 1 - \widehat{p}(s_1)$

Стр. 135

--------------

Пусть в течение контрольной единицы времени событие р не состо-ялось ни одного раза, тогда апостериорная вероятность сохранения бюрократических принципов управления станет равной

$$\widehat{p}(s_1)=\frac{0,2*0,3*0,85+0,75*0,1*0,15}{(0,2*0,3*0,85+0,75*0,1*0,15)+(0,2*0,7*0,85+0,75*0,9*0,15)}=0,11$$

$$\widehat{p}(s_2)\approx=0,89$$

__Таким образом__, вероятность преобладания в структуре оргинизации адаптивных принципов управления в течение контрольного месяца увеличилась с 0,15 до 0,89 . __Определение вероятности__ нахождения организации на определенном этапе жизненного цикла помимо самостоятельного значения имеет еще и обеспечивающую роль - уточнение оценок плохо наблюдаемых переменных внутренней среды организации. Значения большинства переменных, характеризующих состояние организации, зависят от этапа ее развития. Следовательно, оценка переменной должна рассматриваться как средневзвешенное по вероятностям стадий. __Пусть в качестве фазовой переменной__ $х$ внутренней среды организации в рассматриваемой задаче принята эффективность работы организации (производительность труда). Предположим, что переменная $х$ не наблюдаема. Очевидно, что ее значение будет зависеть от типа структуры ($s_1$ или $s_2)$:

$$x = \begin{cases}
F_1(u,s_1,*) & \text{, при } s =s_1,\\
F_2(u,s_2,*) & \text{, при } s =s_2,\\
\end{cases}$$

При известных функциях $F_1$ и $F_2$ оценкой фазовой переменной $х$ будет величина

$$\widehat{x}=F_1(u,s_1,*)\widehat{p}(s_1)+F_2(u,s_2,*)\widehat{p}(s_2)$$

Тогда оптимальное управление определится как

$$\{u_k,\theta_k\}_{opt}=\arg \min\limits_{u_k \in U_k; \theta_k \in \Theta_k}J,$$

где $J$ - показатель качества (эффективности) управления организацией; (В ряде задач из-за особенностей показателя качества минимизация может заменяться максимизацией функционала.) $U_k$ и $\Theta_k$ - области возможных значений управлений $u_k$ и $\theta_k$ на шаге k. __Полученные результаты__ отражают динамику организационных изменений на базе выбранной первичной модели ЖЦО (в примере - молель Грейнера). При этом значения апостериорных вероятностей гипотез солержательно могут интерпретироваться как «доли» преоблалания в организационной структуре принципов бюрократического и адаптивного управления. __На рис. 4.4__ показаны кривые изменения вероятностей первой и второй стадий при условии, что в течение первых трех периодов наблюления из восьми имело место индикаторное событие.

# Глава 5. Методы нечеткой математики и нечеткого моделирования в исследованиях процессов организационного развития
## 5.1. Подходы к построению нечетких моделей организационных систем
Концепция нечетких множеств, предложенная в 1965 г. профессором Калифорнийского университета Лотфи Заде (Zadeh, 1965), поначалу вызвала неоднозначную реакцию и даже критику в научном сообществе (Пегат, 2012). __С середины 1980-х гг__. представление о прикладном характере этой концепции начало меняться, чему способствовали успехи японских, а затем корейских и европейских высокотехнологичных компаний, производящих разнообразные устройства, работающие на основе нечеткой логики. Позже аппарат нечеткой математики и нечеткой логики нашел применение при решении задач управления поездами метрополитена и другими техническими устройствами. __По мнению ряда исследователей__, к настоящему времени уже можно говорить о вполне сформировавшемся научном направлении, условно именуемом «нечеткой математикой» , которое активно используется при исследовании проблем экономического, социального, политического характера, в психологии, при распознавании и классификации образов, а также в лингвистике и теории языков (Теория систем, 2012).

Долгое время применение нечеткой формализации при решении экономических задач рассматривалось экономистами, по выражению Недосекина А. О. (Недосекин, 2003), «как некоторое новомодное излишество». Однако сегодня, по-видимому, можно говорить об изменении этой научной ситуации. Одним из подтверждений перелома во взглядах на данную концепцию явилась международная конференция «Нечеткие множества и мягкие вычисления в экономике и финансах» (FSSCEF, Санкт-Петербург, 2004).

Строго говоря, нечеткая математика является одним из разделов теории нечетких множеств и нечеткого моделирования (Пегат, 2012).
