В современных условиях большинство предприятий функционирует в условиях жесткой конкурсиции, интернационализации и глобализации экономики, быстро меняющейся внешней среды, а также информационных, технических и технологических инноваций. Недостаточное внимание вопросам адаптации организационной и функциональной структуры компании к этим явлениям порожлает проблему «структурно-функциональной неупорядоченности» и снижения экономической эффективности. Под организационным развитием понимается целенаправленная деятельность менеджмента, направленная на повышение производственной результативности и экономичсской эффективности деятельности организации, как в ближайшей, так и долгосрочной перспективе при соблюдении баланса между состояниями стабильности и развития, отвечающего условию самосохранения. Необходимым условием эффективной реализации программы организационного развития является обеспечение требуемого качества процессов организационных изменений. Под организационными изменениями понимается процесс формирования нового организационного устройства, адекватного характеру изменений внешней среды и представлениям менеджмента организации об этой адекватности.

# Глава 1. Теоретические и методологические основы моделирования процессов организационного развития

### 1.1. Анализ современного состояния теории организационного развития и организационных изменений

Иследование проблем организационного (социального) развития имеет длительную историю. Попытки сформулировать принципы развития общественных систем предпринимали такие мыслители, как Аристотель, Платон, Гегель и др. (Лавриненко и др., 2001; Алексеев и др., 2010). В контексте залач настоящего исследования особое значение приобретают идеи Гегеля об отсутствии «отдельного, не связанного с остальным» и границ между различными формами бытия (Гегель, 1974; Лавриненко и др., 2001). Философские взгляды Гегеля пронизаны идеей развития. По его мнснию, в процессе развития совершастся переход количественных изменений в качественные, а источником развития являются противоречия. __Развитие__ - основной предмет диалектики, а сама диалектика выступает как наука о наиболее общих законах развития природы, общества и мышления. Критерием диалектики служит нацеленность на развитие. Познание законов развития дает возможность управлять процессами развития, изменять мир в соответствии с объективными законами и потребностями человеческой цивилизации (Алексеев и др., 2010) Одной из первых в истории философии была классическая модель диалектики, представленная трудами немецких философов XVIII-XIX вв. - Канта, Фитхе, Шеллинга, Гегеля. В XX в. стихийно-диалектическая концепция развития получила широкое распространение в трудах таких естествоиспытателей, как Дж. Хаксли, Л. Берталанфи, Г. Меллера, Дж. Симпсона, Э. Майра, А. Сент-Дьери.

В философской концепции Гегеля действительность представлена цепью диалектических переходов. Согласно Гегелю, история развивается не как механический процесс, а вследствие совокупных действий отдельных людей, каждый из которых стремится реализовать свои собственные интересы и цели. Однако в результате действий людей, преследующих свои цели, возникает нечто новое, отличающесся от их первоначальных замыслов, с чем в своей дальнейшей деятельности люди вынужлены считаться как с объективной предпосылкой. В этом, по мнению Гегеля, состоит переход случайности в необходимость (Спиркин, 2005). __Термин «организационное развитие»__ (organization development) появился в литературе по менеджменту в конце 50-x — начале 60-х годов ХХ века (Bennis, 1969; Масу и др., 1993; Nisbet, 1969; Lippitt и др., 1967; Starbuck, 1965; Salter, 1968). Однако и в настоящее время все еще не существует не только единого понимания роли и места организационного развития в управлении современными организациями, но и единого определения этого процесса.

__В практическом смысле__ под организационным развитием понимается один из основных видов деятельности менеджмента, исходя из следующего определения этого термина: менеджмент (от англ. management - управление, руководство, администрирование) - это профессиональная деятельность по управлению организациями, обеспечивающая высокую результативность и эффективность их работы в ближайшей и долгосрочной перспективе, а также целенаправленное и устойчивое развитие (Михненко, 2014; Адизес, 2008). __В основе понятия__ «организационное развитие» лежит двойственный, но не дихотомический характер определения организации: организация как процесс или функция - деятельность по упорядочению всех элементов определенного объекта во времени и пространстве; организация как объект - целостный комплекс взаимосвязанных элементов и особое единство с внешним окружением, для которого характерна целенаправленность функционирования и развития.

Применение методов теории систем и системного анализа к исследованию организаций заставляет прибегать к иным определениям, например, организация - свойство систем обнаруживать взаимосвязанное поведение ее частей в рамках целого. Организация системы проявляется прежде всего в ограничении разнообразия поведения ее частей и является инвариантным свойством системы. Разные наблюдатели в зависимости от целей и средств исследования могут обнаруживать в системе различные организации (Волкова и др., 2012). Данное определение весьма удачно при решении задач математического моделирования организаций и процессов организационного развития, в частности на этапах обоснования целей и задач моделирования, выбора типа модели, критериев качества и т.п. В общем случае под организационным развитием понимается регулярная целенаправленная деятельность по совершенствованию средств и способов, с помощью которых организация достигает стоящие перед ней цели.

Стр. 10

--------------------

__В общесистемном смысле__ развитие организации может рассматриваться как стремление к выбору таких форм се существования, которые в наибольшей степени обеспечивают достижение желаемого экономического эффекта в ближайшей и долгосрочной перспективе при соблюдении баланса между состояниями стабильности и развития, отвечающего условию самосохранения. При этом следует полагать, что в аспекте закона самосохранения организационные изменения представляют собой процесс целенаправленного (управляемого) нарушения динамического равновесия организации. __В контексте проблемы__ организационного развития сушность самосохранения организации как целенаправленного управляемого процесса можно выразить следующими словами: организация как система стремится к сохранению своей целостности и качественной определенности, используя для этого весь свой потенциал. Под целостностью в данном случае понимается сохранение системы как целенаправленно упорядоченной совокупности взаимосвязанных элементов. Качественная определенность организации - это рациональное сочетание внутренних переменных организации (структуры, ресурсов, целей, функций и т.п.) и связей с внешней средой, соответствующих главной цели деятельности организации. Другими словами, качественная определенность организации - это ее способность достигать поставленную цель в условиях изменяющейся и не всегда благоприятной внешней среды. 

__При этом важно понимать__, что стремление организации к целостности и качественной определенности (а следовательно, к самосохранению) не означает, что ее деятельность должна быть направлена на фиксацию («замораживание») какого-либо своего состояния и удержание его без изменения длительное время. Самосохранение организации возможно в том случае, когда в ней обеспечивается баланс между двумя противоположными состояниями: стабильности и развития. __Внутренние__ $S_{внутр}(t)$ и __внешние__ $S_{внешн}(t)$ факторы, обеспечивающие стабильность организации (stability), должны уравновешиваться внутренними $D_{внутр}(t)$ и внешними $D_{внешн}(t)$ факторами, способствующими ее развитию (development): $\frac{S_{внутр}(t)+S_{внешн}(t)}{D_{внутр}(t) + D_{внешн}(t)}$

Стр. 11

-----------

При отсутствии внутреннего потенциала стабильности $S_{внутр}(t)$ и стремлении организации к развитию, ее самосохранение возможно лишь в условиях максимально стабильной внешней среды, т.е. когда «вес» внешнего потенциала стабильности оказывается равным «весу» суммы внутреннего и внешнего потенциалов развития: $\frac{S_{внутр}(t)}{D_{внутр}(t) + D_{внешн}(t)}=1$. При отсутствии внутреннего петеициала развития $D_{внутр}(t)$  самосохранение организанки возможно лиши, при наличии внешних сил, побуждающих ее к развитию. Примером может служить государственное унитарное предприятие (или акционерное обшество с госудярством в роли главного акционера), развитие которого обеспечивается лишь за счет госуларственных дотаций или специально созданых льготных условий для реализании выпускаемой продукции: $\frac{S_{внутр}(t)+S_{внешн}(t)}{D_{внутр}(t)}=1$ . __В этой свизи__ интересна связь организационно-экономических аспектов обоснования сушности развитии с философскими взглялами на лянный феномен. Так, В. И, Свидерский отмечает, что развитие честь, такое изменение состояний, которос происходит при условии сокранения их основ, т.с, некоторого исходного состояния, порожлающего новое состояние. Сохранение исходного состояния или основы, только и деласт возможным осушествление закономерногин» (Свидерский, 1985)

В статье «Инновационный менелжмент» (Волкова и др., 2012) Б. А. Баринов предлагает следуюшее опрелеление: «Организационное ризвитие определяется как долгосрочный, тшательный, всеоблемлющий процесс изменения и развития организации и работающих в ней людей. Процесс основывается на обучении всех сотрудников путем прямого взаимолействия и перелячи практического опыта. Цель изменений заключается и одновременном повышении производительности организации и качества труда». __Представляется важным отметить__, что развернутое толкование понятия «организационное развитие» в статье, посвященной иннованионному менеджменту, указываст на неразрывную связь этих понятий в современном практическом менеджменте. Причем формально-логическая связь данных дефиниций осушестиляется посредством обосновании понятия «организационные изменения», подчеркивая его приоритетность по отношению к другим аспектам развития. В рамках концепции организационного развития важно установить. В данном определении прослеживается связь таких ключевых понятий, как «развитие», «изменения», «взаимодействие», «обучение» и «повышение производительности». При этом формальн логический подкод к определению понятий сорганизационное развитие» через термин «развитие» без его формального обоснования трудно признать полностью удачным.

Стр. 12

-------------

Представляется важным отметить, что развернутое толкование понития «организационное развитис» в статье, посвященной инно-национному менеджменту, указывает на неразрывную связь этих понятий в сонременном практическом менеджменте. Причем формально-логическая свизьданных дефиниций осушествляется посредством обоснования понятия «организационные изменения», подчеркивая его приоритетность по отношению к другим аспектам развития: 45 рамках концениии организационного развития важно установить, в каком месте организационной иерархии находится исходный пункт для процесса изменений, который в последующем станет всеохватывающим для социальной системы» (Волкова и др., 2012). __Широкоe использование рядом авторов терминов «стратегия» и «стратегические изменения»__ при обосновании особенностей  процесса организационного развития указывает на прямую связь понятий «стратегичсское управление» и «организационное развитиe» (Weik и др., 1999; Акофф, 2002; Андреева, 2004; Волкова и др., 2012; Широкова, 2003; Акимов, 1997: Карлоф, 1991).


Так, Б. Карлоф (Карлоф, 1991) выделяет пять условий реализации стратегии: __во первых__, способность моделировать ситуацию: __во вторых__, способность выявлять необходимость изменений; __в третьих__, способность разрабатывать стратегию изменений; __в четвертых__, способность использовать в ходе изменений надежные методы; __в пятых__,  умeние воплощать стратегию в жизнь. Важной особенностью идеи Карлофа является указание на необходимость моделирования ситуации как одного из условий успешной реализации стратегии. Организационное развитие, как правило, связано с необходимостью решения задачи организационного проектирования (Коршунова, 2004). При этом автор подчеркивает, что до сих пор отсутствует общепринятый критерий эффективности организационного развития. Вследствие этого при планировании и реализации проектов по организационному развитию невозможно сопоставить количественные результаты оценки с какими-либо показателями, которые могли бы считаться доверительными и быть принятыми в качестве референтной величины. Далее делается вывод о том, что уровень теорстического осмысления и методологического обоснования проблемы организационного развития предприятий не в полной мере соответствует современным требованиям и практическим запросам субъектов российской экономики.

Организационное развитие, так же как и эволюция систем управления, обусловлено реакцией на все усложняющиеся внешною среду и задачи предприятий, прогрессирующее накопление критических факторов успеха (Резник, 2006). Там же дается авторская трактовка понятия «развитие предприятия», под которым понимается динамический процесс решения проблем взаимодействия предприятия как социально-экономической системы с внешней и внутренней средами для обеспечения его соответствия изменчивому спросу и различным рыночным возможностям в целях повышения конкурентоспособности.

Стр. 13 

---------------------

Организационные изменения определены как процессы, протекающие в производственной системе целенаправленно благодаря сознательной деятельности управляющих и направленные на поддержание или нарушение ее динамического равновесия (Казакова, 2006). __В качестве причин__, вызывающих организационные изменения в промышленных экономических системах, автор называет необходимость приспособления к изменениям внешней среды, внутренние процессы и потребности системы, а также задачи повышения эффективности ee функинонирования. __Анализ научных источников__ и практического опыта современных предприятий позволяет сделать вывод, что в наши дни большинство предприятий функционирует в условиях быстро меняюшейся внешней среды, жесткой конкуренции, интернационализации и глобализации спроса и предложения, информационных, технических и технологических инноваций (см. приложения 1-3). Недостаточное внимание вопросам алаптации организационной и функциональной структуры компании к этим явлениям порожлает проблему «структурно-функциональной нeупорядоченности». Эта проблема, по мнению исследователей, проявляется в разрыве связей между базовыми структурообразующими элементами компании: целями, задачами, экономическими условнями, технико-технологическим базисом, функционально-поведенческим укладом и др. (Поздняков, 2005).

Несогласованность основ деятельности любого предприятия влечет за собой проблему нерационального использования ресурсов, возникновение конфликтных ситуаций в различных функциональных сферах организации, экономические спады и банкротство. __В рамках__ данных допущений под целями организационных изменений можно понимать формирование такой структурно-функцио-нальной упорядоченности организационной системы, которая обеспечивала бы в долгосрочной перспективе развитие компании, а в краткосрочной - восстановление ее системной устойчивости как баланса между стабильностью и развитием. На практике эта деятельность подразумевает проектирование и внедрение изменений по таким направлениям, как организационная структура, персонал, функции (бизнес-процессы) и качество их выполнения, корпоративная культура, технологии и др. При этом в контексте его реализации организационное развитие может рассматриваться как система мероприятий в сфере менеджмента, направленных на осушествление в организациях пеленаправленных изменсний. __Философская трактовка__ системного характера изменений указываст на то, что меняется качество системы (в оговоренном значении термина «качество»), что требует фиксации признака «качественного изменения» в определении развития (Алексеев и др., 2010). Очевидно, можно с уверенностью утверждать, что первос десятилетие XXI в., особенно с учетом кризисных явлений глобального характера, окончательно закрепило в сознании мснеджмента представление о том, что способность вовремя измениться является одним из определяющих факторов не только успеха и развития, но и выживания их организаций (Адизес, 2008; Сенге, и др., 2003; Хаммер и др., 2006; Широкова, 2003).

Стр. 14

--------------------

По мнению И. Адизеса, «ошибочным является предположение о том, что для сохранения системы от распада необходимо препятствовать изменениям. Такой подход равносилен самоубийству. Он неизбежно приводит к полному распаду. Другими словами, если вы не берете на себя ответственность за разрушение системы так, как вы этого хотите, и за последующую ее интеграцию в соответствии с более совершенной схемой, то система разрушится сама по себе с гораздо более худшими последствиями» (Адизес, 2008). __При этом__ большинство ученых и практиков однозначно признают, что осуществление организационных изменений представляет собой достаточно сложный и в определенной мере опасный для организации процесс - своего рода «высший пилотаж» в работе руководителей компании, предполагающий знание психологии и технологии процесса изменений, а также опыта организационных изменений других компаний (Широкова, 2010). __В самом общем случае__ под организационными изменениями понимается процесс формирования нового организационного устройства, адекватного характеру изменений внешней среды и представлениям менеджмента организации об этой адекватности. __Таким образом__, опираясь на результаты изучения научных трудов по проблеме организационного развития, можно сделать вывод о мстодологической связи ряда понятий в контексте объяснения данного феномена, центральное место среди которых занимает понятие «организационные изменения» (рис. 1.1).

Стр. 16

-----------

Прототипом всех современных моделей организационных изменений принято считать концепцию немецкого психолога Курта Левина (1890-1947), чьи идеи оказали большое влияние на социальную психологию и другие школы и направления. В основе модели К. Левина лежит понятие стабильности. Он выделил две группы факторов: побуждающие к осуществлению изменений (побуждающие силы) и направленные на сохранение стабильности (ограничивающие силы). При равенстве этих сил сохраняется состояние устойчивости. Изменения предполагают либо возрастание силы побужлающих факторов, либо уменьшение силы ограничивающих. __Модель изменений К. Левина__ представляет собой последовательность трех этапов процесса изменений: __во первых__, «размораживание» - слом существующей практики, отказ от старых методов работы; __во вторых__, «движение» - осуществление запланированных мероприятий по изменению поведения работников или подразделений; __в третьих__, «замораживание» - уточнение и закрепление новой организационной практики, недопущение возврата к прежним методам работы. __Помимо модели К. Левина__ к настоящему времени разработано множество концепций, теорий и метафор, объясняющих причины, движущие силы и содержание процессов изменения организаций. К наиболее популярным из них относятся теория жизненного цикла; телеологическая теория; диалектическая теория; теория эволюции.

Теория жизненного цикла утверждает, что изменение организации является неизбежным процессом. Любая организация, следуя своей «внутренней программе», проходит несколько этапов (стадий) жизненного цикла от момента зарождения до исчезновения (смерти). __По мнению Э. Пенроуза__ (Широкова, 2010), представление о том, что фирмы обладают жизненным циклом, аналогичным жизненному циклу биологических организмов, скрывает в себе следующую идею: существуют «законы», обусловливающие развитие фирм, подобные тем законам, по которым... развиваются живые организмы, причем разные стадии развития являются функцией... возраста. Теория жизненного цикла получила широкое развитие в трудах Грейнера Л. (Greiner, 1972; Greiner, 1967), Ализеса И. (Ализес, 2008), Пенроуза Э., Хэнкса С., Уотсона К. Дж. и др., Липпитта Дж. и Шмидта У. (Lippitt и др., 1967) и других авторов.

Телеологическая теория объясняет изменение и развитие на основе идеи о том, что главная причина движения организации - ее цель (замысел). Именно ради достижения некоего целевого (конечного) состояния и развивается организация как целенаправленная и адаптивная система. Сторонники телеологической теории рассматривают развитие как повторяющуюся последовательность, включающую в себя: а) формулировку цели; б) достижение цели; в) оценку полученного опыта; г) последующее изменение на основе этого опыта или новых целей.

Стр. 17

-----------

Согласно Э. Ван-де-Вену и М. Пулу, в отличие от теории жизненного цикла телеология не предполагает наличия фиксированной последовательности событий или траекторий развития организационной единицы (Широкова, 2010). Телеология не предполагает, что в процессе достижения своих целей организация пребывает в состоянии устойчивости и равновесия. Различные воздействия во внешней среде или в самой организации могут вызывать состояние неустойчивости, которое и продвигает ее к новому этапу развития. __Диалектическая теория__ базируется на утверждении Г. Гегеля о том, что организационная единица существует в мире конфликтующих событий, сил или противоречивых ценностей, борющихся друг с другом за доминирование и контроль. Такое противоборство может иметь как внутренние (противоречие целей, групп людей и т. п.), так и внешние (конкуренция с другими организациями и т.п.) причины. Согласно диалектической теории, баланс между стабильностью и изменениями организации обусловлен балансом противодействующих сил. __По мнению Э. Ван-де-Вена и М. Пула__, «изменение происходит тогда, когда эти противодействующие ценности, силы или события достигают необходимой силы для того, чтобы столкнуться друг с другом и пошатнуть статус-кво» (Широкова, 2010). __Теория звалюции утверждает__, что изменение организации представляет собой непрерывный цикл «генетической» изменчивости, отбора и сохранения. Эволюционный отбор организаций происходит в основном в результате борьбы за ограниченные ресурсы. При этом окружающая среда «отбирает» те организации, которые в наибольшей степени соответствуют ресурсам какой-либо «экологической ниши». «Изменчивость стимулирует отбор новых организационных форм, а сохранение закрепляет устоявшиеся формы и методы» (Широкова, 2010; Андреева, 2008).

Как видно, понятие «организационные изменения» в настоящее время трактуется очень широко, причем это характерно как для научного сообщества, так и для бизнеса. Имеет смысл выделить два типа изменений, различающихся по степени преднамеренности: __во первых__, запланированные - целенаправленные изменения, под которыми, как правило, понимается процесс целенаправленного управлясмого перевода организационной системы из текущего состояния в заданное, в аспекте се конкурентного преимущества перед другими организациями; __во вторых__, незапланированные - спонтанные, в том числе инкрементальные, изменения, осуществляемые без заранее определенных общесистемных целей. __Как отмечает Райченко (Райченко, 2003)__, теоретические исследования и практические наблюдения далеко не всегда в состоянии дать ответ на вопрос о том, что стало результатом самостоятельного развития организации, а что является следствием управленческого воздействия на нес. Помимо степени преднамеренности в качестве основных классификационных признаков организационных изменений рассматриваются (табл. 1.1) следующие: __во первых__, характер появления потребности в изменениях: эволюционные, или постепенные; революционные, или скачкообразные; __во вторных__, источник активности действий: реактивные — реакция на произошедшие события; проактивные — осуществляемые на основе прогноза развития событий.

К наиболее востребованным направлениям осуществления запланированных организационных изменений можно отнести изменения технологий, товаров, услуг, стратегии, организационной структуры, организационной культуры. __Изменения в технологии__ - это изменения процессов производства и оказания услуг, а также знаний и навыков работников организации. Целями таких изменений, как правило, являются повышение эффективности производства, производительности труда сотрудников, увеличение объемов и качества производимой продукции. К технологическим изменениям могут также относиться и изменения процедур (технологий) управления.

Стр. 19

--------

Изменения товаров и услуг - это процесс усовершенствования существующих и создание новых видов продукции и услуг. Осушествление подобных изменений, как правило, обусловлено постановкой таких целей развития, как увеличение доли рынка или освоение новых рынков (групп потребителей).

Изменения стратегии и организационной структуры - это, как правило, наиболее сложный, масштабный и трудоемкий тип изменений, так или иначе затрагивающий все сферы деятельности организации. Изменениям подвергаются политика, организационная структура предприятия, система стратегического планирования, вознаграждения, контроля, отчетности и т. п.

Изменения в культуре затрагивают такие аспекты жизни организации, как ценности, ожидания, поведение, формальные и неформальные коммуникации сотрудников и т.п. (Камерон, Куинн, 2001; Чанько, 2003).

В практическом смысле организационные изменения представляют собой систему мероприятий по управляемому переводу организационной системы в новое состояние в соответствии с представлениями о желаемом будущем. Параметры желаемого состояния организации определяются системой целей, которые, по мнению собственников и менеджмента организации, должны быть достигнуты к определенному моменту времени. Как правило, система целей направлена на увеличение экономического эффекта от деятельности организации в ближайшей и долгосрочной перспективе и повышение уровня конкурентоспособности (см. приложение 3).

При этом под эффективностью организационных изменений понимается способность организации добиваться желаемого экономического эффекта с наименьшими затратами ресурсов и времени, отводимых на осуществление изменений, с наименьшим риском гибели организации.



Наиболее заметным подходом к обоснованию революционного характера изменений существенно влияющим на практическую деятельность компаний можно считать концепцию реинжиниринга хозяйственной деятельности М. Хаммера и Дж. Чампи (Хаммер и др., 2006). По мнению авторов, хозяйственный реинжиниринг представляет собой фундаментальное переосмысление и радикальное перепроектирование предприятия (Как правило, по всей цепи создания стоимости). Результатом является скачкообразное улучшение количественных показателей результативности и экономической эффективности его деятельности.

В рамках эволюционного подхода к изменениям под организационным развитием понимается концепция планирования и осуществления изменений социальных систем с привлечением широкого круга участников.
