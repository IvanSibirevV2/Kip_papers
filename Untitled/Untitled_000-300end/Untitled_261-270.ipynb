# 15.2. Стандарт CGI
Web-сервер обеспечивает поддержку взаимодействия клиента и сервера по протоколу НТТР. Это взаимодействие осуществляется по определенным правилам. __CGI (Common Gateway Interface - интерфейс общего шлюза)__ - набор правил (стандарт интерфейса), предназначенных для создания серверных приложений и определяющих порядок запуска программы на сервере, способы передачи программе параметров и доставки результатов ее выполнения клиенту. Программа, написанная по правилам CGI, называется CGI-сценарием. CGI-программы (шлюзы) запускаются сервером в режиме реального времени. Сервер передает запросы пользователя CGI-программе, которая их обрабатывает и возвращает результат своей работы клиенту. Таким образом, пользователь получает динамическую информацию, которая может изменяться в результате влияния различных событий и факторов. Программа __CGI представляет собой__ обычное консольное приложение, работающее в среде операционной системы web-сервера и осуществляющее обмен данными. CGI загружается в адресное пространство сервера и запускается только по явному запросу пользователя, когда к нему выполняется обращение из HTML-документа. Завершив обработку запроса пользователя, программа CGI выгружается из памяти сервера. Программа CGI работает на сервере как отдельный процесс. В рамках этого процесса она способна, в частности, обращаться к базам данных для выборки или обновления информации. При поступлении очередного НТТР-запроса web-сервер инициирует создание нового процесса и передает ему все необходимые данные НТТР-запроса. После того как этот процесс отработает, он завершается, передав при этом результат обратно web-серверу. С точки зрения операционной системы web-сервер и приложение являются разными процессами, поэтому для обмена информацией между ними используются средства межпроцессного взаимодействия.

Когда к серверу одновременно обращаются несколько десятков клиентов, в памяти для каждого создается отдельный процесс, для размещения кода каждого процесса нужна память. Если запрос содержит указание на файл, находящийся на жестком диске, то сервер возврашает в составе ответа этот файл. Если запрос содержит указание на программу и необходимые для нее аргументы, то сервер исполняет программу и результат ее работы возвращает клиенту.

Рис. 15.4. Формирование динамических документов HTML с помощью CGI

Можно описать работу CGI как последовательность следующих операций (рис. 15.4): __Во первых__, web-сервер получает информацию от клиента. Для передачи данных web-серверу используются формы. Форма задается в HTML-документе при помощи тегов `<form> ... </form>` и состоит из набора полей ввода, отображаемых браузером в виде графических элементов управления: полей ввода, управляющих кнопок и т.д. Когда пользователь заполнит всю форму, данные из полей формы передаются программе CGI; __Во вторых__, данные, извлеченные из HTML-формы, передаются для обработки CGI-программе. Они не всегда могут быть обработаны CGI-программой самостоятельно. Например, они могут содержать запрос к базе данных. В этом случае CGI-программа на основании полученной информации формирует запрос к ядру СУБД, выполняющейся на том же или удаленном компьютере; __В третьих__, после обработки полученной информации CGI-программа создает динамический HTML-документ или формирует ссылку на уже существующий документ и передает результат клиенту. Таким образом, пользователю в ответ на запрос посылается HTML код, который был специально сгенерирован для него.

__В третьих__, содержимое ответа — HTML-документ, являющийся результатом выполнения CGI-программы.

При установке web-сервера один из каталогов специально выделяется для хранения сценариев. Обычно такой каталог получает имя `cgi-bin` (или `Scripts` для веб-сервера IIS). ### 15.4.1. Основы синтаксиса РНР
Одним из языков программирования для создания динамических сайтов является язык РНР. Язык РНР — один из наиболее популярных языков сценария благодаря простоте и богатой функциональности. Его название представляет собой рекурсивную аббревиатуру словосочетания «PHP Hypertext Preprocessor», что переводится как «препроцессор гипертекста». Это простой и в то же время очень мощный язык для создания динамических web-страниц. РНР доступен для большинства операционных систем, таких как MS Windows, Mac OS, Linux, многих модификаций Unix и др. Также в РНР включена поддержка большинства современных web-серверов, таких как Apache, MS IIS, серверов Netscape и других. __При написании кода web-страниц__ в HTML код с помощью специальных операторов подгружаются вставки кода РНР. Файлы со сценариями на языке РНР имеют расширение `.php` и представляют собой HTML-документы, в которые с помощью специальных тегов встраивается РНР-код. Если в HTML-странице изменить расширение с `.html` на `.php`, то она станет РНР-страницей, и выполнять ее уже можно только на сервере, поскольку весь вставленный РНР код предварительно обрабатывается на сервере, и затем готовая web-страница передается браузеру. Поэтому РНР называют серверным языком web-программирования. Операторы РНР-сценария выполняются именно на сервере, где размещен содержащий его HTML-документ. Пользователь, загрузивший Web-страницу, открыв программный код HTML-страницы в браузере, не увидит этих операторов. Пользователь обнаружит обычный HTML-код, сформированный на стороне сервера и являющийся результатом работы РНР-скрипта.

РНР полдерживает многострочные комментарии, ограниченные символами `/* ... */`, и в одну строку, начинающиеся с `//` и идушие до конца строки. __РНР не требуюет указания типа__ при объявлении переменных, так же как и самого объявления переменных, т.е, является языком с динамической типизацией. Преобразование скалярных типов может осуществляться автоматически или явно. Напомним, что к скалярным типам данных относятся целый тип, вещественный тип данных, логический тип, строковый тип и специальный тип NULL. К нескалярным типам относятся «ресурс», массив и объект. Тип NULL предназначен для переменных без определенного значения. Ссылки на внешние ресурсы имеют тип `resource`. Переменные данного типа, как правило, представляют собой дескриптор, позволяющий управлять внешними объектами, такими как файлы, динамические изображения, результирующие таблицы базы данных и т.п. Массивы могут содержать значения любых типов, включая другие массивы. Суперглобальными массивами (`superglobal arrays`) в РНР называются предопределенные массивы, которые видны в любом месте исходного кода без использования ключевого слова global. Массив можно создать двумя способами:

__Во первых__, с помощью конструкции аггау:

    $array name = array ("keyl"=>"valuel","key2"=> "value2");

__Во вторых__, непосредственно задавая значения элементам массива:

    $array_name["key1"] =  value1;

РНР поддерживает объектно-ориентированные возможности. Класс в РНР объявляется с помощью ключевого слова `class`. Методы и поля класса могут быть общедоступными (`public`), защишенными (`protected`) и скрытыми (`private`). РНР поддерживает все три основные механизма ООП - инкапсуляцию, полиморфизм и наследование, а также интерфейсы (ставятся в соответствие с помощью `implements`). Множественное наследование классов не поддерживается, однако класс может реализовывать несколько интерфейсов. Для обращения к методам родительского класса используется ключевое слово `parent`. Экземпляры класса создаются с помощью ключевого слова `new`, обращение к полям и метолам объекта производится с использованием символов `->`. Для доступа к членам класса из его методов используется переменная `$this`. Подробное описание языка можно найти в справочном руководстве по РНР.

# 15.4.2. Функции в РНР
Пользовательские функции, т.е. функции, определяемые разработчиком, в РНР характеризуются тем, что: возвращают любой тип данных; дают возможность изменять переданные функции параметры; можно вызвать одну и ту же функцию с переменным числом параметров. Определение (описание) функции в РНР имеет следующий синтаксис:

    function имя_функции (параметр[=значение], параметр[=значение] ...) 
    { тело функции }

Функцию можно объявить в любом месте программы, но обязательно до момента ее первого применения. Имена функций должны быть уникальными и не содержать пробелов. Регистр символов в данном случае не учитывается. Первым символом в имени функции не может быть `$`. Программный код (тело) функции заключают в фигурные скобки. Они определяют группу выражений, которые относятся к коду именно этой функции. Если требуется, чтобы функция возврашала некоторое значение, то в ее теле предусматривают оператор возврата `return` с указанием справа от него того, что следует возвратить. Возврашаемой величиной может быть любое выражение: простое значение, имя переменной или вычисляемое выражение. Оператор `return` может встречаться в коде функции несколько раз. Если оператор `return` не использовать или не указывать справа от него возвращаемую величину, то функция ничего не будет возвращать. Например, приведенная ниже функция вычисляет и возврашает произведение значений своих параметров:

    function S_rec($w $h)
    {
        $S=$w * $h;
        return $S;
    }
