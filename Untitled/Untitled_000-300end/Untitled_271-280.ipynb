С точки зрения доступности переменные подразделяются на глобальные и локальные. Глобальные переменные доступны из любого места программы, в том числе из функции. Локальные переменные определены внутри функции и доступны только изнутри функции. По умолчанию все используемые внутри функции переменные являются локальными и их невозможно изменить вне тела функции. Кроме локальных и глобальных переменных, в программах РНР можно использовать статические переменные. Если переменную внутри функции объявить статической, ее значения не будут изменяться после вызова функции.     <?php
        function f_count()

            echo $counter;
        }
        for($i=0;$i++<9;)f_count();
    ?>

# 15.4.3. Предопределенные переменные РНР

В PHP имеются предопределенные переменные типа аггау (массив) с глобальной областью видимости. Они доступны из любого сценария на языке РНР. Суперглобальные переменные введены в РНР как место хранения данных системного характера, определяющих работу всего сцеария на системном уровне. Перечислим основные массивы предопределенных переменных, которые формирует PHP при обработке пользовательского запроса:
- __во первых__, `$GLOBALS` - содержит все глобальные переменные. Имена глобальных переменных в массиве `$GLOBALS` являются его символьными ключами (индексами). Поэтому, чтобы получить доступ к глобальной переменной, например, `$my_var`, можно использовать выражение `$GLOBALS[«my_var»]`;
- __во вторых__, `$_GET` - содержит данные, переданные в сценарий на РНР в виде части URL-адреса, а также данные HTML-форм, переданные методом GET (GET-запрос);
- __в третьих__, `$_POST` - содержит данные, переданные в сценарий на РНР из HTML-форм методом POST (POST-запрос);
- __в четвертых__, `$_COOKIE` - содержит данные, переданные в текущий сценарий через механизм cookie;

- __в пятых__, `$_ENV` — содержит переменные окружения, содержимое этого массива зависит от операционной системы;
- __в шестых__, `$_FILES` — содержит имена файлов, загруженных методом POST с помощью браузера;
- __в седьмых__, `$_SERVER` - содержит переменные, установленные web-сервером либо непосредственно связанные с окружением выполнения текущего сценария. Эта информация зависит от того, какой web-cepвер используется;
- __в восьмых__, `$_SESSION` — содержит все переменные сеанса, доступные в текущем сценарии;
- __в девятых__, `$_REQUEST` - содержит все переменные, находящиеся в массивах `$_GET`, `$_POST` и `$_COOKIE`.

Для просмотра содержимого глобальных массивов можно воспользоваться функцией `var_dump()`, например `var_dump($_SERVER)`. Если, например, серверному сценарию из HTML-формы было передано методом POST содержимое элемента с атрибутом `name = «myname»`, то его можно получить в сценарии как `$_POST[' myname ']`.

# 15.4.4. Обработка форм с помощью РНР-сценария

__Ранее уже упоминалось__, что методы `GET` и `POST` используются для передачи данных HTML-форм на сервер. Эти методы отличаются друг от друга способом передачи. Метод `GET` добавляет передаваемые данные к URL-адресу, и, следовательно, эти данные видны в адресной строке браузера. Это не всегда допустимо, поэтому более распространен метод `POST`. Кроме того, методом `POST` можно передавать большие по объему данные. РНР размещает данные, полученные от пользователя, в одном из массивов предопределенных переменных `$_GET` или `$_POST` в зависимости от используемого метода передачи данных. __Стандартным средством для отправки__ на web-сервер данных, введенных пользователем с клавиатуры или сгенерированных клиентским/серверным сценарием, является HTML-форма, создаваемая контейнерным тегом `<FORM>`. В атрибуте `action` тега `<FORM>` указывается URL-адрес файла со сценарием, который должен принять эти данные и каким-либо образом их обработать. В контейнере `<FORM action=URL_адрес>` могут содержаться различные элементы пользовательского интерфейса (поля ввода данных, флажки, переключатели, кнопки) и должна быть, по крайней мере, одна специальная кнопка, создаваемая тегом `<INPUT type-submit>`, для отправки на сервер данных, содержащихся в элементах формы.

Получив данные от клиента, серверный сценарий может просто сохранить их в файле или базе данных, обработать каким-либо образом и, в зависимости от результата отправить клиенту некоторое сообщение, сгенерировать HTML-документ и т.д. Метод передачи данных (`GET` или `POST`) указывают в атрибуте method tera `<FORM>`, например, `<FORM action= "http://localhost/php_form_post.php" method=POST>` . __URL-адрес__, указанный в параметре `action` тега `<FORM>` показывает месторасположение РНР-сценария, который предназначен для обработки данных из этой формы. В данном случае используется локальная обработка данных на компьютере пользователя без выхода в сеть Интернет. Если в тегах элементов формы (`<INPUT>`, `<TEXTAREA>` И т. п.) указано значение атрибута name (имя элемента), то на сервер будут отправлены соответствующие пары вила: `имя_элемента = значение_элемента`.

# 15.4.5. Взаимодействие РНР с базами данных
__PHP__ имеет наборы специальных функций для работы с многими СУБД, такими как Microsoft SQL Server, MySQL, Oracle, PostgreeSQL, Informix, Sybase, IBM DB2. Кроме этого, РНР поддерживает стандарт ODBC (Open Database Connectivity - открытый интерфейс доступа к базам данных), которому соответствуют почти все реляционные СУБД. СУБД устанавливается независимо от РНР. Чтобы работать с ней посредством РНР-сценария, следует установить соответствуюшую данной СУБД библиотеку функций. Наиболее распространено использование СУБД MySQL. __MySQL__ - реляционная система управления базами данных, свободно распространяемая, надежная и простая в использовании, достаточно быстрая. MySQL вполне подходит для небольших проектов. Существует довольно популярный визуальный интерфейс для работы с этой СУБД — PhpMyAdmin, который позволяет значительно упростить работу с базами данных в MySQL. В текстовом режиме работа с базой данных выглядит просто как ввод команд в командную строку. Многие из базовых SQL-функций работы с данными в PhpMyAdmin сведены к интуитивно понятным интерфейсам и действиям. __Взаимодействие сценария РНР с базой данных__ обычно выполняется в следующей последовательности. После запуска сервера базы данных осуществляется подключение к нему. Далее следует передача SQL-запроса, который должен быть выполнен СУБД. SQL-запрос формулируется на языке SQL, оформляется в виде строки и передается специальной РНР-функции в качестве параметра. Если нет ошибок или каких-то других проблем, то СУБД выполнит SQL-запрос. Если SQL-запрос предполагает возврат данных, то они будут получены с помощью специальной РНР-функ-ции. С помощью специальных функций можно получить данные, соответствующие переданному ранее SQL-запросу, и далее их обработать, например, вывести на экран нужный текст. Можно обработать и отформатировать полученные данные, провести анализ, преобразование и т.п. Если не предполагается повторное обращение, то производится отключение от базы данных. Отключаться от базы данных рекомендуется всякий раз, как только прекращается работа с ней на длительный срок. __Подключение к базе данных__. Подключение к базе данных осуществляется с помощью специальных функций РНР, которым в качестве параметров передаются следующие данные: __местоположение (host)__ - для многих серверов БД это доменноеимя или IP-адрес компьютера, на котором расположена БД: если она находится на том же компьютере, что и модуль РНР, то достаточно указать localhost. При необходимости можно указать через двоеточие порт, который прослушивается сервером базы данных; __регистрационное имя (username)__ — строка символов, соответствующая имени пользователя, под которым он может получить доступ к БД; __пароль (password)__ - строка символов, которая вместе с регистрационным именем служит для обеспечения доступа к БД; имя базы данных.

Функция РНР, выполняющая запрос, возвращает указатель на полученную в результате временную таблицу. Если требуется изменить значения полей таблицы, добавить новую или удалить имеющуюся запись, то формируют строку с соответствующим SQL-запросом, а затем передают ее в качестве параметра функции, выполняющей запрос.

В РНР при подключении к базам данных в разных СУБД используются разные функции:

__Во первых__, в системе управления базами данных MS SQL Server:
    
    $connect = mssql_connect (имя_сервера_БД, username, password);
    $db = mssql_select_db (имя_БД, $connect);

__Во вторых__, в системе управления базами данных MySQL: Оншо

    $connect = mysql_connect(host, username, password);
    $db = mysql_select_db($connect, имя_БД).

__Функция__ `<...>_connect()` выполняет соединение с сервером БД и возвращает идентификатор соединения при успехе и `false` - при неудаче. При этом для MySQL в первом параметре этой функции указывают `host` (доменное имя или ІР-адрес компьютера), а для MS SQL Server — имя сервера БД, устанавливаемое при регистрации сервера. Идентификатор соединения, возвращаемый функцией, используется в сценарии, в частности, при выборе базы данных, создании запросов и т.д. __Функция__ `<...>_select()` выбирает конкретную базу данных на сервере, на который указывает идентификатор `$connect`, возвращенный первой функцией. Эта функция возвращает `true` при успешном завершении операции и `false` - в противном случае. B MS SQL Server базу данных можно создать с помощью утилиты Enterprise Manager или с помощью соответствующих команд SQL. B MySQL можно ввести необходимые SQL-выражения в командной строке либо написать соответствующий РНР-сценарий, а затем выполнить его. Отключение от базы данных осуществляет функция, которой передается идентификатор соответствующего подключения. Например, `mysql_close($connect)` - для MySOL, `mssq_close($connect)` — для MS SOL Server.

__Обработка данных в сценарии.__ Выбор нужных данных из одной или нескольких таблиц выполняет SQL-запрос, начинающийся с ключевого слова (оператора) `SELECT`. Критерий выбора данных формулируется в выражении SQL-запроса велед за ключевым словом WHERE. Результатом выполнения такого запроса является временная таблица, возможно, пустая, если нет данных, удовлетворяющих условию запроса. Функция РНР, выполняющая запрос, возвращает указатель на полученную в результате временную таблицу. Этот указатель используется в РНР-сценарии для доступа к данным с помошью специальных функций. Иногда данные из временной таблицы записываются в массив, а затем его обрабатывают. Если требуется изменить значения полей таблицы, добавить новую или удалить имеющуюся запись, то формируют строку с соответствующим SQL-запросом, а затем передают ее в качестве параметра функции, выполняющей запрос (например, `mssql_query()`). В качестве параметров этой функции надо передать SQL-запрос SELECT с соответствуюшими параметрами. РНР-оператор для выборки всей таблицы будет следующий:

    $result-mysql_query("SELECT * FROM staff", $link).

При успешном выполнении запроса функция `mysql_query()` возвращает идентификатор результата запроса, и данный идентификатор можно передавать другим функциям для обработки результата. В этой переменной будет храниться вся таблица в виде матрицы. Для того чтобы извлекать из матрицы отдельные записи, используется функция `mysql_fetch_array()`. В качестве параметра этой функции надо передать имя `$result`. PHP-оператор для выборки первой записи из таблицы будет таким:

    $myrow=mysql_fetch_array($result).

# 15.4.6. Взаимодействие РНР и XML
Язык РНР можно использовать для работы с XML-документами. В РНР для этого существуют два модуля, реализующие два разных стандарта обработки XML-данных: SAX (Simple API for XML) и DOM (Document Object Model). __Стандарт SAX__ не является стандартом WЗC и описывает метод обработки XML-документов для получения из них данных. То есть этот метод обработки XML-документов позволит только прочитать данные из XML-документа, не более того. Создавать и изменять XML-документы с его помощью невозможно. __Другим стандартом для обработки XML-данных является DOM__ - стандарт W3C, спецификация которого имеется на сайте консорциума. В отличие от SAX этот метод позволяет производить любые операции с XML-данными в достаточно удобной форме, представляя XML-документ как дерево объектов. Модуль, реализующий этот стандарт, называется DOM XML. Он не входит в основной набор модулей РНР, но может быть установлен как расширение API (программного пользовательского интерфейса) этого модуля. Модуль DOM XML является мощным и удобным в использовании средством обработки XML-документов. Модуль DOM XML определяет в РНР несколько классов, таких как DomNode, DomDocument, DomElement, Dom Text и DomAttribute, большинство из которых идут из ядра стандарта DOM. Почти для всех классов (в частности, для перечисленных выше) класс DomNode является родительским, поэтому его свойства и методы наследуются всеми остальными классами. В произвольном XML-документе классу Dom Document будет соответствовать сам этот документ, классу DomElement - каждый XML-тег, классу DomAttribute - атрибуты тегов, а классу DomText - содержание XML-элементов. В то же время классу DomNode будет соответствовать каждый из перечисленных элементов XML-документа.

# 15.4.7. Сеансы в РНР
__Сеанс (сессия) в РНР__ - механизм, который позволяет обеспечить доступность данных, созданных на различных серверных страницах в рамках одного сайта, для других страниц того же сайта. __Технология сеансов__ перекрывает возможности и преодолевает ограничення `соокіс` по обмену данными между web-страницами. Сеанс создается и закрывается, когда это необходимо. В интервале времени между этими событиями все РНР-сценарии, расположенные в различных файлах и подключенные к данному сеансу, имеют доступ к переменным, созданным в других сценариях этого же сайта. Чтобы воспользоваться возможностями, предоставляемыми механизмом сеанса, его необходимо создать или подключиться к нему. В пределах существующего сеанса можно установить сеансовые переменные. Для этого служит суперглобальный массив `$_SESSION`. __Каждое web-приложение__, как правило, состоит из нескольких страниц, содержащих сценарии. При загрузке каждой страницы переменные в этих сценариях очищаются, и для обмена данными между страницами можно использовать механизм передачи параметров, базу данных либо запись информации в cookie-файлах. Однако наиболее удобным методом является использование сеансов. __При открытии сеанса__ для каждого пользователя на сервере формируется определенная среда, заданная совокупностью параметров. относящихся к данному пользователю. Например, там могут сохраняться какие-то настройки интерфейса, конфигурации отдельных сервисов и т.п. Соответствующие данные хранятся в суперглобальном массиве `$_SESSION`. __Каждому сеансу присваивается__ уникальный идентификатор, который может использоваться при повторном открытии сеанса. Этот идентификатор можно сохранить, например, в сооіе-файле, после чего при повторном посещении сайта уже не потребуется заново настраивать работу с его интерфейсом и сервисами. Для открытия сеанса используется функция `session_start()`, не имеющая пара-метров. После этого становится доступным массив `$_SESSION`, и в нем можно сохранять свои данные, которые будут доступны из любой части web-приложения до завершения сеанса.
