# 13.3. Расширяемый язык разметки XML
__XML (eXtensible Markup Language)__ - текстовый формат, предназначенный для хранения структурированных данных и для обмена данными между разными системами обработки информации. XML рекомендован консорциумом Всемирной паутины W3C язык разметки и фактически представляет собой свод общих синтаксических правил. XML является упрощенным подмножеством языка SG ML и характеризуется следующими признаками: __во вервых__, XML — язык гипертекстовой разметки, ориентированный на описание структурированных данных; __во вторых__, позволяет использовать не только предопределенные теги, но и вводить свои; __в третьих__, не зависит от платформы. __Язык XML__ был разработан рабочей группой XML Working Group консорциума W3C специально для размещения информации в WWW, аналогично языку гипертекстовой разметки HTML, который изначально стал стандартным языком создания web-страниц. Это расширяемый язык разметки данных, но не каких-то конкретных, а любых структурированных данных, которые можно представить в текстовом виде. На основе XML можно создавать различные расширения - специализированные языки разметки. Стандарт XM L. регламентирует лишь то, как должны описываться структурные элементы документа (теги) и связанные с ними атрибуты. Иными словами, XML. определяет обобщенный синтаксис разметки документа. XML позволяет структурировать информацию разного типа, используя для этого произвольный набор инструкций. XML может использоваться для создания документов какого-то определенного типа и структурой необходимой для конкретного приложения. Однако если сфера применения языка оказывается достаточно широкой и он начинает представлять интерес для большого числа разработчиков, то его спецификация вполне может быть представлена на рассмотрение в W3C и после согласования всеми заинтересованными сторонами утверждена консорциумом в качестве официальной рекомендации.

К известным недостаткам языка можно отнести следующее: __во первых__, размер XML-документа существенно больше, чем размер документа в альтернативных текстовых форматах передачи данных и особенно в форматах данных, оптимизированных для конкретного случая использования. Избыточность XML может повлиять на эффективность приложения. Возрастает стоимость хранения, обработки и передачи данных; __во вторых__, для большого количества задач не нужна вся мощь синтаксиса XML, и можно использовать значительно более простые и производительные решения; __в третьих__, пространства имен XML сложно использовать и их сложно реализовывать в XML-парсерах. XML-парсер — это программа, предназначенная для анализа содержимого текстового документа, который соответствует спецификации XML; __в четвертых__, XML не содержит встроенной в язык поддержки типов данных. В нем нет понятий целых чисел, строк, дат, булевых значений и т.д. __Важным отличием XML__ от HTML является более строгий контроль за соблюдением правил языка при разметке документов. Документ XML считается правильно построенным, если он соответствует всем синтаксическим правилам XML, а именно: __во первых__, XML-документ содержит один и только один корневой элемент, содержащий все остальные элементы; __во вторых__, дочерние элементы, содержащиеся в корневом элементе, должны быть правильно вложены. __Имена элементов__ подчиняются простым правилам: имя начинается с буквы, знака подчеркивания или двоеточия; после первого символа в имени могут быть буквы, цифры, знаки переноса, подчеркивания, точка или двоеточие; имена не могут начинаться с буквосочетания XML.

# 14.1. Объектная модель документа
Объектная модель документа (Document Object Model — DOM) является стандартом, предложенным консорциумом W3C, и регламентирует способ представления содержимого web-страницы в виде набора объектов. Фактически это интерфейс, позволяющий программам и сценариям динамически обращаться и обновлять содержимое, структуру и стиль документа. Под содержимым понимается все, что может находиться на web-странице: текст, рисунки, ссылки, абзацы И Т.Д. __В отличие от__ объектной модели браузера (ВОМ), которая уникальна для каждого браузера, объектная модель документа является стандартом и должна поддерживаться всеми браузерами. DOM может применяться не только в web-страницах, но и в любых других документах. В частности, она может использоваться с любыми словарями XML, причем одним из таких словарей является XHTML. __В DOM документ представляется в виде__ древовидной структуры (рис. 14.1), являющейся одной из наиболее употребительных структур в программировании. Это обеспечивает унифицированный способ навигации по документу. Структура DOM представляет собойиерархию объектов, похожую на объектную иерархию любого объектно-ориентированного языка. DOM предлагает свой полезный API, не привязанный к какому-либо конкретному языку программирования.

    </html>

Согласно модели DOM: __во первых__, весь документ представляется узлом документа; __во вторых__, каждый HTML тег является узлом элемента; __в третьих__, текст внутри HTML элементов представляется текстовыми уз-лами; __в четвертых__, каждому HTML атрибуту соответствует узел атрибута; __в пятых__, комментарии являются узлами комментариев. В приведенном выше примере корневым узлом является тег `<html>`. Все остальные узлы содержатся внутри `<html>`. У этого узла имеются два дочерних узла: `<head>` и `<body>`. Узел `<head>` содержит узел `<title>`, а узел `<body>` содержит узлы `<h1>` и `<p>`.

Все узлы HTML-документа могут быть доступны посредством дерева, при этом их содержимое может быть изменено или удалено, а также можно добавить новые элементы. Все узлы дерева находятся в иерархических отношениях между собой. Для описания этих отношений используются термины: родитель, дочерний элемент и потомок. Родительские узлы имеют дочерние узлы, а дочерние элементы одного уровня называются потомками (братьями или сестрами). В отношении узлов дерева соблюдаются следующие правила: __во первых__, существует корневой узел — самый верхний узел дерева; __во вторых__, каждый узел, за исключением корневого, имеет один родитель-ский узел; __в третьих__, узел может иметь любое число дочерних узлов; __в четвертых__, конечный узел дерева не имеет дочерних узлов; __в пятых__, потомки имеют общего родителя. __DOM__ - это спецификация универсального доступа, независимого от платформы и программ, к содержимому документов. DOM является своеобразным АРІ (прикладным программным интерфейсом) для документов. DOM предоставляет стандартный способ построения объектной модели любого HTML- или XML-документа, при помощи которой можно производить поиск нужных фрагментов, создавать, удалять и модифицировать его элементы.

Программный интерфейс DOM включает в себя набор стандартных свойств и методов. Свойства представляют некоторые сущности, а методы - действия над ними. Для описания интерфейсов доступа к содержимому документов в спецификации DOM применяется язык IDL и для использования их необходимо «перевести» на какой-то конкретный язык программирования. Однако этим занимаются создатели самих анализаторов, прикладным программистам можно ничего не знать о способе реализации интерфейсов — с точки зрения разработчиков прикладных программ, DOM выглядит как набор объектов с определенными методами и свойствами. __Объектная модель__ документа делает все элементы страницы программируемыми объектами. С ее помощью через языки сценариев можно получить доступ и управлять всеми элементами документа. Каждый элемент HTML доступен как индивидуальный объект, следовательно, можно изменять значение любого параметра любого тега HTML-страницы, и, как следствие, документ действительно становится динамическим. Любое действие пользователя объектной моделью документа трактуется как событие, которое может быть перехвачено и обработано процедурой сценария. DOM обеспечивает для HTML-документа возможность динамического изменения своего содержимого без перезагрузки страницы. __Динамический HTML (Dynamic HTML или DHTML)__ не является языком разметки страниц. Это термин, применяемый для обозначения HTML-страниц с динамически изменяемым содержимым. Реализация DHTML основывается на HTML, каскадных таблицах стилей (Cascade Style Sheets - CSS) и языке сценариев (JavaScript или VBScript). Эти три компонента DHTML связаны между собой объектной моделью документа DOM.

# 14.2. HTML5
Стандарт пятой версии HTML был завершен в 2014 г., тогда вышла рекомендованная версия к использованию. Цель разработки пятой версии HTML — улучшение уровня поддержки мультимедиа-технологий. HTML5 - это, можно сказать, новая технология в развитии web-индустрии. Рассмотрим основные моменты, которым уделялось внимание при создании спецификации языка HTML5. __Одной из важных составляющих HTML5__ является семантика, каждый тег обладает своей смысловой нагрузкой. Программы, которые будут анализировать сайт, построенный на НTML5, должны будут понимать, какие данные заключены между тегами, какова их суть и какова их значимость. Появились новые теги, например для семантического разделения страницы сайта на верхнюю часть, нижнюю, боковую, основной контент (`<section>`, `<article>`, `<header>`, `<nav>`). Были исключены некоторые устаревшие элементы, которые можно было использовать в HTML 4.01, включая чисто оформительские элементы, такие как `<font>` и `<center>`, так как эффекты могут выполняться с помощью каскадных таблиц стилей (CSS). Heкоторые элементы, например `<а>`, `<menu>` и `<cite>`, были изменены, переопределены или стандартизированы. Но при этом HTMLS сохраняет полную совместимость с предыдущими языками разметки XHTML, HTML4 и 3. __Большое внимание уделяется__ независимости языка от плагинов (динамически подключаемых модулей). Музыка, видео, анимация, игры - со всем этим можно работать, не загружая на компьютер и в браузер сторонние программы. Элементы `<video>`, `<audio>` и `<canvas>`, а также возможность использования SVG и математических формул разработаны для упрощения создания и управления графическими и мультимедийными объектами в сети, без необходимости использования сторонних АРІ.
